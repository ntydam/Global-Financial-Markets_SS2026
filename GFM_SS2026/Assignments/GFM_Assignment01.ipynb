# Assignment: Measuring & Managing Currency Risk in a Multinational Firm

---

## Learning Objectives

By completing this assignment, you will be able to:

- Quantify exchange rate volatility using real data  
- Analyze co-movement across currencies (diversification effects)  
- Measure total FX risk using a portfolio approach
- Evaluate and design hedging strategies under uncertainty 
- Communicate risk management decisions to senior management  

---



## Assignment Structure (Group-Based: 3–5 Students)

You will act as the **FX Risk Management Team** of a multinational corporation (MNC).

Each group is assigned:
- A **set of currency pairs**
- A **fixed exposure profile**

You must use your assigned setup for all analysis.

---

# Step 1: Exchange Rate Dynamics & Volatility
---

## Task
Download exchange rate data from Yahoo Finance for your assigned currency pairs (check group assignment part).

## Requirements

1. Use **daily or weekly data (≥ 2 years)**  
2. Compute returns
3. Compute for each currency:
- Mean return  
- Standard deviation (volatility)  
- Annualized volatility

## Analysis Questions

- Which currency is most volatile? Why?  
- Are there differences between developed vs emerging currencies?  
- Does volatility appear stable over time?  

---

# Step 2: Correlation & Diversification
---
## Task
Analyze relationships between currencies.

## Requirements

1. Compute correlation matrix
2. Visualize (heatmap recommended)

## Analysis Questions

- Which currencies provide diversification benefits?  
- Which are highly correlated? Why?  
- Economic reasoning (commodities, geography, trade)

---

# Step 3: MNC Exposure & Portfolio FX Risk
---
## Requirements

1. Convert exposures to USD using current rates  
2. Compute weights:

$$
w_i = \frac{\text{USD exposure}_i}{\text{Total absolute exposure}}
$$

3. Compute portfolio risk (portfolio having more than 2 currencies)


## Analysis Questions
- Which currencies drive risk the most?  
- Does diversification reduce risk?  
- Is risk driven more by volatility or correlation?

---

# Step 4: Hedging Strategy Design
---
## Task
Propose **TWO hedging strategies**

Examples:
- Forward contracts  
- Options  
- Natural hedging  
- Money market hedge  

## Requirements

For each strategy:
- What exposure is hedged?  
- How it works  
- Cost vs benefit  
- Residual risk  

## Critical Thinking

- Should firms hedge all exposures?  
- What determines optimal hedging level?

---

# Step 5: Executive Recommendation
---
You **cannot hedge 100% of exposures**
Deliverable: 5–7 slides 

Must Include
- Key findings  
- Quantitative support  
- Trade-offs
