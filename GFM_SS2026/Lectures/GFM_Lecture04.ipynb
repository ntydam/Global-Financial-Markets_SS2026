# 1. International Parity Conditions

---

Managers of MNEs, international portfolio investors, importers, exporters, and government officials confront a few core questions every day:

- What determines exchange rates (ERs)?
- Are ER changes predictable?

The economic theories that connect ERs, price levels, and interest rates (IRs) are known as **international parity conditions**. These conditions form the backbone of the financial theory that is distinctive to international finance.

In practice:
- parity relationships do not always appear to “hold” when compared with real-world outcomes. 
- the gap is not a failure of the theory itself, but a matter of how the theory is interpreted and applied—including assumptions, timing, market frictions, and measurement choices.

# 2. Exchange‑Rate Fundamentals </span>

---


## <span style="color:#4A86C5;">2.1. Nominal Exchange Rates & Real Exchange Rates</span>

---
### <span style="color:#5b9bd5;">Nominal Exchange Rates (NER)
- NER is the price of one currency in terms of another.
- It is the most commonly quoted exchange rate and is used for most transactions.

### <span style="color:#5b9bd5;">Real Exchange Rates (RER)
- The RER measures how many units of domestic goods are needed to buy a foreign good, adjusting for differences in price levels.

$$
\boxed{\text{RER} = e \times \left(\frac{P^{*}}{P}\right)}
$$
 
Where:
- $e$ = Nominal ER (domestic currency per unit of foreign currency)
- $P^{*}$ = Foreign price level
- $P$ = Domestic price level

Interpretation
- RER > 1 → Domestic goods are relatively expensive → Currency may be **overvalued  
- RER < 1 → Domestic goods are relatively cheap → Currency may be undervalued

---

## <span style="color:#4A86C5;">2.2. Exchange Rate Indices: NEER and REER</span>

Economists and policymakers use **ER indices** to:

- assess the overall external value of a country’s currency
- evaluate whether the currency may be overvalued or undervalued

These indices are constructed as **trade-weighted averages** of a home currency’s ERs against those of its major trading partners, with the weights based on bilateral trade shares.

Why trade-weighted?
- larger trading partners have a greater influence on a country’s external position
- a single bilateral ER may not capture the full picture of a currency’s international competitiveness

---



### <span style="color:#5b9bd5;">Nominal Effective Exchange Rate (NEER)</span>

- NEER is a trade-weighted average of a home currency’s **nominal ERs** against its trading partners
- the weights reflect bilateral trade shares
- it does not adjust for differences in price levels or inflation across countries

### <span style="color:#5b9bd5;">Real Effective Exchange Rate (REER)</span>

- REER adjusts the NEER for differences in inflations between the home country and its trading partners
- REER extends RER from a single bilateral rate to a trade-weighted basket of currencies
- REER provides a better measure of a country’s international price competitiveness

$$
REER = NEER \times \frac{C_F}{C_H}
$$

where:
- $C^F$ = Foreign price level (trade-weighted)
- $C^H $= Domestic price level  

**Example: Where to collect NEER and REER?**
- Bank for International Settlements (BIS) – Real Effective ERs: The BIS publishes nominal and real effective exchange rate indices for over 60 economies.  These indices are used to assess a currency’s value against a basket of trading partners.  Data portal: [Click to view how the data is presented on the BIS website](https://data.bis.org/topics/EER/tables-and-dashboards)

> <span style="color:#9E86C5;">**Question:**</span><br>
> **What is going to happen if domestic inflation rises faster than foreign inflation?**


> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> - **NEER** = trade-weighted average of NERs
> - **REER** = NEER adjusted for relative price levels

# 3. Law of One Price & Purchasing Power Parity

---


## <span style="color:#4A86C5;">3.1. Law of One Price</span>

- A primary principle of competitive markets is that prices will equalize across markets if frictions do not exist.
- To be clear: Identical goods sold in different countries should have the same price.  

Assumptions for LOP to hold:
- no trade barriers (e.g., tariffs, quotas)
- no transportation frictions
- perfect competition and arbitrage (traders exploit price quickly, forcing convergence)
- goods are tradable and identical

## <span style="color:#4A86C5;">3.2. Absolute Purchasing Power Parity (PPP)</span>
Purchasing Power Parity (PPP) draws a comparison between different currencies through a market **basket of goods** approach.

If the LOP holds, the spot exchange rate (implied PPP rate) equals the ratio of general price levels in the two countries.  

\begin{aligned}
\text{Implied PPP Rate} &= \frac{P_\text{Burger, home}}{P_\text{Burger, foreign}} \\
\text{ }\\
\end{aligned} 

The Big Mac Index compares the price of a Big Mac in different countries to compute implied PPP rates and assess currency over‑ or undervaluation.

> **What does the Big Mac Index suggest about the valuation of the British pound in July 2017?**
> - Given:
>   - Price of Big Mac in Britain: 3.29 GBP
>   - Price of Big Mac in the United States: 5.661 USD
>   - Actual ER: 1.3489 USD/GBP

<span style="color:#9e86C5;">**Solution:**</span><br>
  - The implied PPP rate was:
\begin{aligned}
\text{Implied PPP Rate} &= \frac{5.66\,\text{USD}}{3.29\,\text{GBP}} = 1.7204\,\text{USD/GBP}
\end{aligned}
  
  - Since the actual ER < the implied PPP (1.3489 < 1.7204), GBP is overvalued to the USD, by:

\begin{aligned}
\text{Overvaluation (\%)} &= \frac{1.7204 - 1.3489}{1.3489} \times 100\% = 27.5\%
\end{aligned}

## <span style="color:#4A86C5;">3.3. Relative PPP & Empirical Evidence</span>

When absolute PPP’s assumptions are relaxed, only changes in price levels (inflation) matter.

Relative PPP states that changes in the ER between two countries equal the inflation differential between them.  

$$
\boxed{\frac{\Delta S}{S} \approx \pi_H - \pi_F}
$$ 

Implication:
  - Currencies of high-inflation countries tend to depreciate
  - Currencies of low-inflation countries tend to appreciate.

**Note:**
  - Relative PPP is not used to determine the current spot ER level.
  - Instead, it predicts the change in ERs over time driven by inflation differences
  - Macro factors (e.g., trade barriers, productivity shocks) and asset market dynamics often dominate ER fluctuations.

**Empirical Evidence:** 
  - PPP works well over very long horizons but fails to predict short‑term ER movements. 
  - PPP performs better for economies with persistently high inflation or underdeveloped financial markets.  

---


### <span style="color:#5b9bd5;">Summary</span>

> - **LOP**: If arbitrage is possible, prices of tradable goods converge across borders.
> - **Absolute PPP**: Extends this idea to a basket of goods: currencies should adjust so that a representative bundle costs the same everywhere. 
> - **Deviations**: Arise because many goods are non‑tradable (e.g., services, labor), trade barriers and taxes exist, and price indices differ across countries.
> - **Relative PPP**: Links currency depreciation to inflation differentials. 
> - **Real Purchasing Power**: Even if levels of prices differ across countries, currencies tend to adjust so that real purchasing power remains comparable.
> - **Persistent Inflation Gaps**: Create long‑run trends in ERs, but short‑term deviations are common due to speculative capital flows, government interventions and shifts in risk premia.

# 4. Pass-through & Policy Trade-offs

---

## <span style="color:#4A86C5;">4.1. Exchange Rate Pass-Through</span>

**Exchange rate pass-through (ERPT)** measures the response of import and export prices to changes in ER.

### <span style="color:#5b9bd5;"> Complete pass-through
In theory, if prices fully adjust to ER movements, changes in the ER should be completely reflected in traded-goods prices.

**Example: Volvo Car Price Change**

Suppose a Volvo is priced at **€50,000** in Europe.

$$
P_{\text{Volvo}}^{\$} = P_{\text{Volvo}}^{€} \times S^{\$/€}
$$

If the euro appreciates from $1.00/€ to $1.20/€, then the U.S. dollar price rises from:

$$
\$50{,}000 \to \$60{,}000
$$

$$
\frac{P^{\$}_{\text{Volvo},2}}{P^{\$}_{\text{Volvo},1}}
=
\frac{60{,}000}{50{,}000}
=
1.20
$$

This is a 20% increase, which matches the 20% appreciation of the euro.  

---



### <span style="color:#5b9bd5;">Partial pass-through
- In practice, pass-through is often partial (ER changes do not always translate one-for-one into the prices), because firms may:
    - absorb part of the ER movement in their profit margins, 
    - face contractual rigidities, or 
    - adjust prices gradually over time.


**Example: Volvo Car Price Change**

If Volvo chooses to raise the U.S. price only to **$58,000** instead of **$60,000**, then:

$$
\frac{P^{\$}_{\text{Volvo},2}}{P^{\$}_{\text{Volvo},1}}
=
\frac{58{,}000}{50{,}000}
=
1.16
$$

This is only a 16% increase, even though the ER changed by 20%.  

---


## <span style="color:#4A86C5;">4.2. Why Pass-Through Is Often Partial</span>

In real markets, firms often do not pass through the full effect of ER changes because:

- raising prices too much may reduce sales volume
- demand may be sensitive to price changes
- firms may absorb ER changes through profit margins
- imported inputs may change production costs
- contracts and menu costs can delay price adjustment

As a result, ER changes may affect prices gradually rather than immediately.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> ERPT depends not only on the ER itself, but also on firm pricing strategy, market structure, and consumer responsiveness.

---


## <span style="color:#4A86C5;">4.3. Price Elasticity of Demand and Pass-Through</span>

**Price elasticity of demand** is useful when determining the desired rate of pass-through.

$$
\boxed{\varepsilon_p = \frac{\%\Delta Q_d}{\%\Delta P}}
$$

Where:

- $\epsilon_p$ = price elasticity of demand  
- $Q_d$ = quantity demanded  
- $P$ = product price  

**Interpretation:**
- If $|epsilon_p| < 1$, demand is **inelastic**  
  - Consumers are not very price sensitive → quantity demanded changes little.  
  - firms are more able to pass ER changes into prices  

- If $|epsilon_p| > 1$, demand is **elastic**  
  - Consumers are very price sensitive → quantity demanded drops sharply if price rises.  
  - firms are less able to pass ER changes into prices  

**Example:**
- Volvo cars have very price-elastic demand in the U.S., Volvo may avoid fully increasing its U.S. price after a euro appreciation. In that case, pass-through will likely be partial.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> Elastic demand → Partial pass-through
> Inelastic demand → Higher / near-complete pass-through

---


```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Elasticity cases
elasticities = [0.5, 1.0, 1.5]
colors = ["blue", "black", "red"]

def plot_quantity_demand_change(price_change_pct=0.0):
    """
    price_change_pct: user input in percent, from -20 to +20
    Formula:
        %ΔQd = elasticity * ( - %ΔP )
    """
    q_change_results = []

    for e in elasticities:
        q_change = -e * price_change_pct
        q_change_results.append(q_change)

    # Build chart data: show how %ΔQ varies across the full slider range
    x = np.linspace(-20, 20, 200)

    plt.figure(figsize=(9, 6))
    for e, c in zip(elasticities, colors):
        y = -e * x
        plt.plot(x, y, color=c, linewidth=2, label=f"Elasticity = {e}")

    # Highlight current slider-selected point for each elasticity
    for e, c, q_change in zip(elasticities, colors, q_change_results):
        plt.scatter(price_change_pct, q_change, color=c, s=60)

    plt.axhline(0, color="gray", linewidth=1, linestyle="--")
    plt.axvline(0, color="gray", linewidth=1, linestyle="--")

    plt.title("Change in Quantity Demanded for Different Elasticities")
    plt.xlabel("% Price Change")
    plt.ylabel("% Change in Quantity Demanded")
    plt.xlim(-20, 20)
    plt.grid(True, alpha=0.3)
    plt.legend()
    plt.show()

    # Print calculated outputs
    print(f"Input % Price Change: {price_change_pct:.1f}%\n")
    for e, q_change in zip(elasticities, q_change_results):
        print(f"Elasticity = {e:<3}: % Change in Quantity Demanded = {q_change:+.1f}%")

interact(
    plot_quantity_demand_change,
    price_change_pct=FloatSlider(
        value=0.0,
        min=-20.0,
        max=20.0,
        step=1.0,
        description="%ΔPrice"
    )
);
```

## <span style="color:#4A86C5;">4.4. Pass-Through, PPP, and the REER</span>

If PPP held perfectly:
- ER changes would be fully offset by relative price changes, and 
- REER would remain close to its equilibrium value.

However, when pass-through is partial:
- relative prices do not fully adjust → PPP may not hold in the short run
- REER can deviate from its benchmark or base-year value
- competitiveness effects may persist over time

This helps explain why real ER movements are often more persistent than simple PPP theory would suggest.

---



## <span style="color:#4A86C5;">4.5. Policy Trade-Offs and Emerging Markets</span>

In Emerging market economies (EMs), ER changes often have a strong effect on domestic inflation because:
- Heavy reliance on imports (energy, technology, etc.).
- Strong ERPT (lower pricing power, more volatile currencies → cannot absorb the cost shock)
- Less anchored inflation expectations (higher $\pi$ expectation implifies $\pi$ impact).

### <span style="color:#5b9bd5;">The Impossible Trinity
Many EMs face the classic policy trade-off described by the impossible trinity (or policy trilemma):
- A country cannot simultaneously maintain all three of the following:
    - fixed ER
    - independent monetary policy
    - free capital mobility

Countries must choose which two of the three objectives.

In order to attract capital inflows, many EM countries have:
- moved away from pegged ERs, and

This is because:
- Flexible ER helps absorb external shocks (Currency depreciation can restore competitiveness).

Policy dilemma:
- more flexible ERs can increase ERPT

## <span style="color:#4A86C5;">5.1. The Fisher and International Effects</span>

---


### <span style="color:#5b9bd5;">5.1.1. The Domestic View: The Fisher Effect
Named after economist Irving Fisher.

This theory explains how inflation eats into your investment returns domestically.

It states that the nominal IR is simply a combination of the real IR (your actual growth in purchasing power) and expected inflation.

**The Formula:**
$$\boxed{i \approx r + \pi^e}$$

**Where:**
* $i$ = Nominal rate of interest (e.g., your bank's advertised yield)
* $r$ = Real rate of interest (actual increase in purchasing power)
* $\pi^e$ = Expected inflation

> <span style="color:#9e86C5;">**Homework:**</span><br> 
> Check the exact derivation of this equation in Appendix 1.

**Example:**
If the nominal rate on a bank deposit is **4%** and expected inflation is **3%**, your real rate of return is roughly **1%**. 
* *Insight:* Empirical studies using short-maturity government bonds show this relationship holds reasonably well in the real world. A smaller real rate implies a slower growth of your actual purchasing power.



---



### <span style="color:#5b9bd5;">5.1.2. The Global View: International Fisher Effect (IFE)
The IFE takes the domestic Fisher Effect and applies it to foreign exchange.

It relates changes in **spot ERs** to **IR differentials** between two countries:
- Global investors demand the same *real* return regardless of where they put their money
- A country with a higher nominal IR doesn't actually offer a better real return—it just has higher expected inflation.

Therefore, that country's currency must eventually depreciate to offset the higher IR.

**The Formula (Approximation):**
$$\boxed{\frac{S_{t+1} - S_t}{S_t} \approx i_H - i_F}$$

**Where:**
* $S_t$ = Current spot ER
* $S_{t+1}$ = Expected future spot ER
* $i_H$ = Home country nominal IR
* $i_F$ = Foreign country nominal IR

**Example:**
- If the US nominal IR ($i_H$) is **5%** and the Japanese nominal rate ($i_F$) is **1%**, the US rate is **4%** higher. 
- According to the IFE, the US Dollar should depreciate by roughly **4%** against the Japanese Yen to eliminate any arbitrage opportunity.

---



### <span style="color:#5b9bd5;">5.1.3. The Real Interest Rate Parity (Real-IRP)
Why does the IFE work? It relies on **Real Interest Rate Parity**. 

- If Purchasing Power Parity (PPP) holds and global capital markets are fully integrated (meaning money flows freely across borders), real IRs will converge across countries ($r_H \approx r_F$). 
- If real rates are identical everywhere, then any difference in *nominal* IRs is purely driven by differences in *expected inflation*. 

**The Mathematical Connection:**
- If $r_{Home} = r_{Foreign}$, we can substitute the Fisher equation ($i = r + \pi^e$) for both countries.
- This reveals that the difference in nominal rates equals the difference in expected inflation:
$$i_H - i_F \approx \pi^e_H - \pi^e_F$$

---

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> Nominal IRs compensate lenders for two things: 
> - the time value of money (real return)
> - the loss of purchasing power (expected inflation). 
>
> Internationally:
> - If capital flows freely, real returns align across countries. 
> - Therefore, currencies with higher nominal IRs are simply masking higher expected inflation. 
> - To compensate, those high-interest currencies tend to depreciate against low-interest currencies.

## <span style="color:#4A86C5;">5.2. Linking Parity Conditions in Equilibrium</span>

---

- In perfectly efficient, open global markets, money cannot make "free" riskless profit simply by moving across borders. 
- All the major economic levers—inflation, IRs, spot ERs, and forward ERs—must stay in equilibrium. 
- The chart below visually maps this interconnected web.

### <span style="color:#5b9bd5;">The Baseline Scenario (The Setup)
To explain the chart, we use a real-world numerical example comparing the USD and JPY:
* Expected Inflation: 5% in the US vs. 1% in Japan. *(Japan has 4% less expected inflation).*
* Nominal IRs (1-year): 8% in the US vs. 4% in Japan. *(Japan's rate is 4% lower).*
* Current Spot Rate ($S_1$): JPY 104/USD.
* 1-Year Forward Rate ($F$): JPY 100/USD.

### <span style="color:#5b9bd5;"> Breaking Down the 5 Relationships (A - E)

#### Relation A: Purchasing Power Parity (PPP)
* **What it connects:** Expected Inflation $\rightarrow$ Forecasted Spot Rate.
* **The Concept:** Currencies with higher inflation lose purchasing power faster, so their currency must depreciate against currencies with lower inflation. 
* **The Math:** Japan has 4% less inflation than the US (1% vs. 5%). Therefore, the JPY must strengthen by 4% over the next year to maintain purchasing power parity. 
$$S_2 = S_1 \times \frac{1 + \pi^{\text{JPY}}}{1 + \pi^{\text{USD}}} = 104 \times \frac{1.01}{1.05} = \text{JPY 100/USD}$$

#### Relation B: The Fisher Effect
* **What it connects:** Expected Inflation $\rightarrow$ Nominal IRs.
* **The Concept:** Real returns (growth in purchasing power) should be equal across currencies assuming efficient and open markets. Therefore, the difference in nominal IRs must perfectly equal the difference in expected inflation.
* **The Math:** * USD Real Rate: $8\% \text{ nominal} - 5\% \text{ expected inflation} = 3\%$.
    * JPY Real Rate: $4\% \text{ nominal} - 1\% \text{ expected inflation} = 3\%$. 
    * The nominal IR differential (-4%) mirrors the inflation differential (-4%).

#### Relation C: International Fisher Effect (IFE)
* **What it connects:** Nominal IRs $\rightarrow$ Forecasted Spot Rate.
* **The Concept:** The forecast change in the spot ER is equal to, but opposite in sign to, the differential between nominal IRs. 
* **The Math:** Japan's interest rate is 4% lower than the US ($4\% - 8\% = -4\%$). To compensate, the JPY is forecasted to strengthen by exactly 4%.
$$\frac{S_1 - S_2}{S_2} \times 100 = i^{\text{JPY}} - i^{\text{USD}} = -4\%$$

#### Relation D: Interest Rate Parity (IRP)
* **What it connects:** Nominal Interest Rates $\rightarrow$ Forward ER.
* **The Concept:** According to the theory of interest rate parity, the difference in nominal interest rates is equal to, but opposite in sign to, the forward premium. 
* **The Math:** Because Japan's nominal interest rate is 4% less than the US rate ($i^{\text{JPY}} - i^{\text{USD}} = 4\% - 8\% = -4\%$), the 1-year forward market prices the JPY at a positive 4% premium to the USD.
$$f^{\text{JPY}} = \frac{S_1 - F}{F} \times 100 = \frac{104 - 100}{100} \times 100 = 4\%$$

#### Relation E: Forward Rate as an Unbiased Predictor
* **What it connects:** Forward ER $\rightarrow$ Forecasted Spot Rate.
* **The Concept:** If assumed to be an unbiased predictor, the 1-year forward rate contracted today (JPY 100/USD) should be our best guess of what the future spot rate will be one year from now (JPY 100/USD).

---


### <span style="color:#5b9bd5;">Parity Conditions Summary</span>

$$
\begin{aligned}
\text{PPP:}&\quad S_t = \frac{P_H}{P_F}, \quad \frac{\Delta S}{S} \approx \pi_H - \pi_F \\ 
\text{IRP:}&\quad F_t = S_t \times \frac{1 + i_H}{1 + i_F},\quad \frac{F_t - S_t}{S_t} \approx i_H - i_F \\ 
\text{Fisher effect:}&\quad i = r + \pi,\quad \text{IFE:}\; \frac{\Delta S}{S} \approx i_H - i_F \\ 
\text{Real interest parity:}&\quad r_H = r_F
\end{aligned}
$$


> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> - The international parity conditions form a consistent framework linking goods prices, IRs and ERs.  
> - When markets are frictionless, these relationships align and arbitrage opportunities vanish.  
> - In practice, frictions and expectations cause persistent but predictable deviations, offering opportunities for risk management and speculation.

## <span style="color:#4A86C5;">Appendix 1: Derivation of the Fisher Equation

The Fisher effect is derived by looking at how an investment grows in nominal terms versus how prices grow due to inflation. 

**Step 1: Define the basic variables**
- Let's assume you have $1 to invest today.
    * $i$ = Nominal IR
    * $r$ = Real IR
    * $\pi^e$ = Expected inflation rate

**Step 2: Calculate your nominal wealth next year**
- If you invest your $1 at the nominal IR, your money grows. Next year, you will have:
$$\text{Nominal Wealth} = 1 + i$$

**Step 3: Calculate the new price of goods next year**
- If a basket of goods costs $1 today, inflation means that the same basket will cost more next year. The new price will be:
$$\text{Future Price} = 1 + \pi^e$$

**Step 4: Find your real purchasing power**
- Your *real* return ($1 + r$) is determined by how much your new wealth can actually buy at the new prices.
- We find this by dividing your nominal wealth by the future price of goods:
$$1 + r = \frac{1 + i}{1 + \pi^e}$$

**Step 5: Rearrange the equation**
- Multiply both sides by the denominator $(1 + \pi^e)$ to get it out of the fraction:
$$(1 + r)(1 + \pi^e) = 1 + i$$

**Step 6: Expand the left side**
- Multiply the terms inside the parentheses together:
$$1 + r + \pi^e + r\pi^e = 1 + i$$

**Step 7: Simplify**
- Subtract 1 from both sides of the equation to isolate the nominal IR:
$$i = r + \pi^e + r\pi^e$$

**Step 8: The Approximation**
- In the real world, both the real IR ($r$) and expected inflation ($\pi^e$) are usually small percentages (decimals). 
- When you multiply two small decimals together (the $r\pi^e$ term), the result is so tiny that it is practically negligible. 

*For example, if $r$ is 2% (0.02) and $\pi^e$ is 3% (0.03), their product is 0.0006.*

Because the cross-product $r\pi^e$ is so close to zero, economists drop it to form the standard, simplified Fisher equation:
$$i \approx r + \pi^e$$

Many countries have experienced high inflation and currency misalignment in history. Once high inflation becomes embedded in the expectation of the society, it is very hard to stop.

Argentina struggled with high inflation, and the government used price controls to slow rising prices (E.g., setting price control and limiting wage increases).

The **Big Mac** became politically sensitive because it is used in The Economist’s Big Mac Index, a simple indicator of purchasing power parity (PPP) and relative prices around the world.

In 2012, other burgers on the McDonald’s menu were priced around 33–35 pesos, while the Big Mac was only 20 pesos.  
This unusually cheap price suggested that there was unofficial pressure to keep the Big Mac price low. However, because the Big Mac was artificially cheap, it was often not widely promoted or sometimes unavailable in stores.

Keeping the price low helped make Argentina’s prices—and therefore its currency—appear less misaligned in international comparisons.

### <span style="color:#4A86C5;">Case 2: United States, 1929–1933

#### A Short Story Behind the Table

Imagine being a shop owner in the United States in 1929. At first, business slows a little. But each year things get worse. Customers lose their jobs, factories close, and prices keep falling. By 1931 and 1932, unemployment is extremely high and the economy is shrinking. 

IRs appear low, but because prices are falling (deflation), the **real cost of borrowing money actually becomes very high**. This makes it harder for businesses and households to borrow and spend. As a result, economic activity continues to decline. 

This table shows how unemployment rose dramatically while output collapsed between 1929 and 1933. At the same time, deflation increased real IRs, contributing to what economists call a **deflation trap** during the Great Depression.

*Source: Adapted from Olivier Blanchard, Macroeconomics, “Deflation in the Great Depression.”*
