# 1. Introduction to FX Exposures

---

## <span style="color:#4A86C5;"> 1.1. FX Risk and Firm Value
---
Exchange rate (ER) movements directly affect a firm's performance by influencing:
- Revenues (e.g., export sales)
- Costs (e.g., imported inputs)
- Profit margins

FX exposure is highly relevant for MNCs:
- It is often explicitly disclosed in financial statements because ER movements can affect reported performance and firm value

- **Example:**
    - *“In general, we are a net receiver of currencies other than the U.S. dollar.  Accordingly, changes in exchange rates, and in particular a strengthening of the U.S. dollar, will negatively affect our revenue and other operating results as expressed in U.S. dollars.”*  **- Facebook (Meta)**

Since firm value is determined by the ${{PV}^{e}_{CF}}$, any ER-driven change in revenues, costs, or margins can ultimately affect the firm's value.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> - FX risk is volatility in CFs that creates uncertainty in firm value  
> - FX risk is not just volatility, it’s valuation risk.

---


## <span style="color:#4A86C5;">1.2. Types of FX Exposure
---
FX exposure is categorized into three main types:

| Type         | Nature                        | Time Horizon | Effects      |
|--------------|-------------------------------|--------------|--------------|
| Transaction  | Contractual CFs               | Short-term   | Transaction exposure affects contracted foreign-currency CFs, e.g., receivables and payables  |
| Translation  | Accounting (balance sheet)    | Periodic     | Translation exposure affects reported financial statements when foreign operations are consolidated into the parent’s reporting currency |
| Economic     | Competitive / strategic       | Long-term    | Economic exposure affects future CFs, competitiveness, and firm value |

# 2. Transaction Exposure

---


## <span style="color:#4A86C5;"> 2.1. Definition
---

Transaction exposure refers to the risk that ER movements affect contracted foreign-currency (FC) CFs.

It arises when a firm has **known future payments or receipts** denominated in a FC.
- Exporter → foreign-currency receivable 
- Importer → foreign-currency payable

**Example**
- Intel invoices many exports in USD, which reduces transaction exposure.  
- However, if the EUR weakens, European customers may switch to cheaper local suppliers.
This affects **future (uncontracted) CFs**, not existing contracts.

> <span style="color:#9e86C5;">**Question:**</span><br> 
> How can we distinguish between transaction exposure and operating exposure if both arise from ER changes that affect CFs?

> <span style="color:#9e86C5;">**Answer:**</span><br> 
- The key distinction is the nature of the cash flows:
  - Transaction exposure relates to **contracted** foreign-currency cash flows (e.g., receivables, payables).  
  - Operating (economic) exposure relates to **expected future CFs** that are not yet contracted and depend on competitiveness, prices, and demand.

---


## <span style="color:#4A86C5;"> 2.2. Direction of Risk
---

The impact of ER changes depends on whether the firm holds a receivable or payable.

| Position   | If FC Appreciates | If FC Depreciates |
|------------|------------------|------------------|
| Receivable | Gain             | Loss             |
| Payable    | Loss             | Gain             |

This is critical for hedging decisions

---


### <span style="color:#4A86C5;"> 2.2.1.  Foreign Currency Depreciation

When a FC depreciates against the home currency:

#### Effects on cash inflows
- FC revenues convert into **fewer home-currency units**  
- Export demand may fall if competitiveness declines  
- Foreign income (interest, dividends) decreases in value  

#### Effects on cash outflows
- Imports priced in foreign currency become **cheaper**  
- FC debt-payments decline in home-currency terms  

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Depreciation benefits firms with FC **costs**, but hurts firms with FC **revenues**.

---


### <span style="color:#4A86C5;"> 2.2.2. Foreign Currency Appreciation

When a FC appreciates:

#### Effects on cash inflows
- Foreign revenues convert into **more home-currency units**  
- Export demand may increase  
- Foreign investment income rises  

#### Effects on cash outflows
- Imports priced in foreign currency become **more expensive**  
- Foreign-currency debt payments increase  

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Appreciation benefits firms with FC **revenues**, but hurts firms with FC **costs**.

---


## <span style="color:#4A86C5;"> 2.3. Measuring Transaction Exposure
---

To measure transaction exposure, firms follow a simple process:

### <span style="color:#4A86C5;">Step 1. Map Foreign-Currency Cash Flows</span>

Identify all expected receipts and payments:
- By currency  
- By amount  
- By timing  

### <span style="color:#4A86C5;">Step 2. Calculate Net Positions</span>

$$\boxed{\text{Net Exposure} = \sum \text{Receivables} - \sum \text{Payables}}$$

- Firms focus on **net exposure**, not gross exposure  
- This avoids unnecessary hedging

### <span style="color:#4A86C5;">Step 3. Convert to Home Currency</span>

$$ \boxed{\text{Home-currency exposure = Net FC position × Expected future spot rate}} $$

- Allows comparison across currencies  
- Identifies which currencies create the largest risk  

---

### <span style="color:#4A86C5;"> Example:

A U.S. firm expects the following next-quarter positions:

- GBP: net inflow of GBP 10 million, expected rate USD 1.50/GBP
- CAD: net inflow of CAD 10 million, expected rate USD 0.80/CAD  
- SEK: net outflow of SEK 100 million, expected rate USD 0.15/SEK  

$
\text{Home currency exposure} = (\text{GBP}\ 10 \text{M} \times 1.50) + (\text{CAD}\ 10\text{M} \times 0.80) - (\text{SEK}\ 100\text{M} \times 0.15)
$

$
\text{Home currency exposure} = 15 + 8 - 15 = \text{USD}\ 8\ \text{million}
$


> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Transaction exposure is:
> - Short-term
> - Contract-based
> - Directly measurable <br>
>
> It forms the foundation for:
> - Hedging decisions
> - More advanced measurement (VaR, Sensitivity Analysis)

# 3. Hedging Transaction Exposure
---

## <span style="color:#4A86C5;">3.1 Forward Hedge
---
The firm signs a contract to exchange a specific amount of currency at a predetermined forward rate on a future date.
- Receivable: Sell the FC forward.
- Payable: Buy the FC forward.

> <span style="color:#9e86C5;">**Question:**</span><br>
> A U.S. exporter expects to receive **GBP 20 million** in 90 days. To hedge transaction exposure, should it buy or sell pounds forward? What does the contract lock in?

> <span style="color:#9e86C5;">**Answer:**</span><br>
The exporter is **long GBP** (will receive GBP). To hedge, it must take an offsetting short position: **sell GBP 20 m forward** at today's forward rate. This locks in the USD value of the receivable regardless of where GBP/USD moves over the next 90 days.

---


## <span style="color:#4A86C5;">3.2. Money Market Hedge
---
### <span style="color:#4A86C5;">3.2.1. Procedure</span>
A money market hedge creates a **synthetic forward hedge** by combining:
- borrowing,
- spot conversion, and 
- investment.

| Step | Hedging a Receivable | Hedging a Payable |
|---|---|---|
| Borrow today | Borrow the foreign currency | Borrow the home currency |
| Convert now | Exchange it at today’s spot rate into home currency | Exchange it at today’s spot rate into foreign currency |
| Invest today | Invest the home currency until maturity | Invest the foreign currency until maturity |
| At maturity | Customer payment repays the foreign-currency loan | Matured foreign-currency investment is used to pay the supplier |
| What is locked in? | A known home-currency inflow | A known home-currency outflow |

---

### <span style="color:#4A86C5;">3.2.2. Example: Money Market Hedge for a Receivable

A U.S. firm expects to receive EUR 1,000,000 in 90 days.
- Spot rate: 1.10 USD/EUR  
- EUR interest rate (90 days): 2%  
- USD interest rate (90 days): 1%  

#### The Procedure:

**Step 1:** Borrow  $\text{PV}_\text{receivable}$ at 2% interest rate (90 days)

$$
\text{EUR borrowed today} = \frac{1,000,000}{1.02} = 980,392 \text{ EUR}
$$

**Step 2:** Convert EUR to USD

$$
\text{USD-amount} = \frac{980,392}{1.10 \text{USD/EUR}} = +1,078,431 \text{ USD}
$$

**Step 3:** Invest USD at 1% interest rate (90 days)

$$
\text{USD-received in 90 days} = +1,078,431 \text{ USD} \times {1.01} = +1,089,215 \text{ USD}
$$

**Step 4:** Use the receivable of EUR 1,000,000 to payback the borrowed amount at day 90.


> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> - The firm locks in an inflow of 1,089,215 USD
> - No exposure remains to ER movements

## <span style="color:#4A86C5;">3.3. Option Hedge
---


### <span style="color:#4A86C5;">3.3.1. Definition
* A currency option gives the firm the right, but not the obligation, to exchange currency at a predetermined strike price, $K$

* The firm can let the option expire if market conditions turn out to be more favorable

* An option hedge therefore provides downside protection while preserving upside potential

* The main cost of this flexibility is the option premium, which must be paid upfront

---


### <span style="color:#4A86C5;">3.3.2. How the hedge works:

#### 1. Payable Hedging with a Call Option
- A firm with a FC payable is concerned that the FC may appreciate
- Buying a **call option** sets a maximum purchase price:

$$
\boxed{\text{Maximum effective cost per unit} = K + \text{Premium}}
$$

Decision rule:
- If $S_T < K$, the firm does **not** exercise the option and instead buys at the spot rate $S_T$
- If $S_T \geq K$, the firm **exercises** the option and buys at the strike price $K$

---


#### 2. Receivable Hedging with a Put Option
- A firm with a FC receivable is concerned that the FC may depreciate
- A put option provides a minimum guaranteed ER (floor)

$$\boxed{
\text{Net inflow per unit} =
\begin{cases}
K - \text{Premium}, & \text{if } S_T < K \\
S_T - \text{Premium}, & \text{if } S_T \ge K
\end{cases}}$$

**Decision Rule (Receivables)**
- If $S_T < K$ → Exercise the option (sell at $K$)  
- If $S_T \geq K$ → Let it expire (sell at the market rate $S_T$)

---

**Example: Viner Co.**
- Expected receivable: CHF 200,000 in 6 months
- Put option details:
  - Strike price ($K$): USD 0.72/CHF  
  - Premium: USD 0.02/CHF  
  - The option guarantees a minimum net rate of USD 0.70/CHF
- If the currency strengthens, the firm still benefits from the higher spot rate
- The premium is the cost of flexibility

| Scenario | $S_T$ |  | $K$ | Premium | Decision | Net per Unit (USD) | Total Inflow (USD) |
|---|---:|:---:|---:|---:|---|---|---:|
| 1 | 0.71 | < | 0.72 | 0.02 | Exercise | 0.72 - 0.02 = 0.70 | 140,000 |
| 2 | 0.74 | > | 0.72 | 0.02 | Expire | 0.74 - 0.02 = 0.72 | 144,000 |
| 3 | 0.76 | > | 0.72 | 0.02 | Expire | 0.76 - 0.02 = 0.74 | 148,000 |

---

> - Premium = $0.04
>
> **Questions:**
> 1. Spot = $0.74, exercise or not?

---

### <span style="color:#4A86C5;">3.3.3. Contingency Graph - Call Option </span>

**Example: Call option hedge on a payable**
- A firm buys a call option on EUR with:
  - $K$ = $1.20/€
  - Premium = $0.03/€

**Case 1: Euro stays below the strike price**
- Suppose the future spot rate is $1.16/€
- The firm lets the option expire and buys euros in the spot market
- Total dollar cash outflow per euro: $\boxed{1.16 + 0.03 = 1.19}$

**Case 2: Euro rises above the strike price**
- Suppose the future spot rate is $ 1.26/€$
- The firm exercises the option and buys euros at $ 1.20/€
- Total dollar cash outflow per euro: $\boxed{1.20 + 0.03 = 1.23}$

### <span style="color:#4A86C5;">3.3.4. Comparison with a forward hedge</span>
- A forward hedge locks in one fixed price and must be executed

- An option hedge is more flexible:
  - if $S_T$ is favorable, the firm can let the option expire
  - if $S_T$ is unfavorable, the firm exercises the option

- The trade-off is that the option hedge requires paying a premium

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> A put option creates a **floor (downside protection)** while preserving **upside potential**, unlike a forward hedge which locks in a single outcome.

---

### <span style="color:#4A86C5;">Expected Value of Hedged Position</span>

An MNC may incorporate its own ER forecasts when estimating CFs to be received from a put option hedge.

Assume probabilities:
- Scenario 1: 30%  
- Scenario 2: 40%  
- Scenario 3: 30%  

$
E(V) = (140,000 \times 0.30) + (144,000 \times 0.40) + (148,000 \times 0.30) = \boxed{144,000}
$

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> A put option creates a **floor (downside protection)** while preserving **upside potential**, unlike a forward hedge which locks in a single outcome.

# 4. Choosing a Hedging Strategy
---


## <span style="color:#4A86C5;">4.1 Decision Factors
---


> Many firms hedge selectively. For example: <br>
> ***“Decisions regarding whether or not to hedge a given commitment are made on a case-by-case basis by taking into consideration the amount and duration of the exposure, market volatility, and economic trends.” — DuPont Co.***

### <span style="color:#4A86C5;">Certainty of cash flow
- If the FC CF is **known and fixed**, a firm is more likely to use a full hedge.
- If the cash flow is **uncertain**, the firm may prefer a more flexible hedge, such as an option, or hedge only part of the exposure.
- Overhedging is possible when the actual payment or receipt later turns out to be smaller than expected.

### <span style="color:#4A86C5;">Risk tolerance
- More risk-averse firms often hedge most of their exposure, even if hedging does not always maximize expected CF.
- Less risk-averse firms may hedge selectively, especially if they believe diversification across currencies already reduces total risk.

### <span style="color:#4A86C5;">Cost of hedge
- Firms compare the expected CFs under each hedge.
- Forward and money market hedges usually produce **known** home-currency outcomes.
- Option hedges require an upfront **premium**, but provide flexibility because they do not have to be exercised.

### <span style="color:#4A86C5;">Market expectations
- The optimal hedge may vary over time depending on:
  - the forward rate,
  - interest rates,
  - option premium, and
  - the firm’s forecast of the future spot rate.

---



## <span style="color:#4A86C5;">4.2. When to Use Each
---



Firms usually compare the home-currency CFs from the available hedging techniques and then choose the one that best fits the transaction.

1. Compare **forward/futures** and **money market hedge**, since both produce known outcomes.  
2. Then evaluate the **option hedge**, whose outcome depends on the future spot rate and must be assessed using scenario analysis or an expected value. 

| Situation | Best Tool | Why |
|---|---|---|
| Certain CF and desire for certainty | Forward hedge | Locks in a known exchange rate and known home-currency cash flow |
| Certain CF but no suitable forward advantage | Money market hedge | Creates a synthetic forward using borrowing and investing |
| Uncertain CF or desire to keep upside potential | Option hedge | Protects against adverse ER movements while preserving favorable outcomes |
| No forward or option market available | Money market hedge | Uses spot and money markets instead of derivatives |
| Forward and money market give similar results | Either one | Under IRP and no transaction costs, they should be very close |

---

**Example: Coleman Co. must pay EUR 100,000 in one year.**

**Market data:**
- Forward rate = **USD 1.20/EUR**
- Spot rate today = **USD 1.18/EUR**
- EUR deposit rate = **5%**
- USD borrowing rate = **8%**
- Call option strike price = **USD 1.20/EUR**
- Call option premium = **USD 0.03/EUR**

**Forecast for the option hedge:**
- Scenario 1: \(S_T = 1.16\), probability = 20%
- Scenario 2: \(S_T = 1.22\), probability = 70%
- Scenario 3: \(S_T = 1.24\), probability = 10%

| Hedge Tool | Calculation | USD Cash Outflow |
|---|---|---:|
| **Forward hedge** | $100{,}000 \times 1.20 $| 120,000 |
| **Money market hedge** | Step 1: Find the EUR needed today $\frac{100{,}000}{1.05} = 95{,}238\ \text{EUR}$ <br>Step 2: Convert that EUR amount into USD today $95{,}238 \times 1.18 = 112{,}381\ \text{USD}$ <br>Step 3: Borrow USD today to lock in the cost, repay it in one year $112{,}381 \times 1.08 = 121{,}371\ \text{USD}$ | 121,371 |
| **Call option hedge** | Scenario 1: if $S_T = 1.16 < K$, let option expire  $(1.16 + 0.03)\times 100{,}000 = 119{,}000$  <br> Scenario 2: if $S_T = 1.22 > K$, exercise option $(1.20 + 0.03)\times 100{,}000 = 123{,}000 $ <br> Scenario 3: if $S_T = 1.24 > K$, exercise option $(1.20 + 0.03)\times 100{,}000 = 123{,}000$  <br><br> Expected outflow: $(119{,}000 \times 0.20) + (123{,}000 \times 0.70) + (123{,}000 \times 0.10)$ | 122,200 |


In that case, the **forward hedge** is optimal because it gives the lowest expected dollar cost among the hedging alternatives.

---


## <span style="color:#4A86C5;">4.3 Speculation vs Hedging
---

### <span style="color:#4A86C5;">Comparison
A hedge is designed to reduce the uncertainty of future home-currency CFs.
- Hedging most exposure allows firms to forecast future CFs more accurately.
- This can improve financing, liquidity planning, and budgeting decisions.


A speculative position is taken because the firm believes ERs will move in its favor.
- A firm that remains unhedged is accepting ER risk.
- It may do so because it expects a better outcome than the hedge provides.

---



### <span style="color:#4A86C5;">Hedge vs No Hedge
Firms should compare:
- the **known** CF from hedging, and
- the **expected** CF from not hedging.

| Choice | Cash flow outcome | Risk level |
|---|---|---|
| Hedge | More predictable | Lower |
| No hedge | Potentially better or worse | Higher |



**Example: Coleman Co. (payables)**
For Coleman’s EUR payables:
- Forward hedge cost = **USD 120,000**
- Expected unhedged cost = **USD 121,000**

In addition, there was an **80% probability** that remaining unhedged would be more expensive than using the forward hedge. Therefore, Coleman chose to hedge.

> <span style="color:#9e86C5;">**Question:**</span><br>
> - If the known cost of hedging is about the same as the expected cost of remaining unhedged, will firms hedge?

> <span style="color:#9e86C5;">**Answer:**</span><br>
Yes, most firms will hedge, because hedging removes the possibility of a large adverse ER movement.

# 5. Netting vs Natural Hedge
---


## <span style="color:#4A86C5;">5.1 Netting</span>
---

**Netting**: Combining all expected inflows and outflows in the **same currency** over a given horizon before deciding how much exposure to hedge.

A firm may have large FC receivables and payables in the same currency, but it is exposed only to the **net position**.

$$\boxed{\text{Net Exposure} = \sum \text{Receivables} - \sum \text{Payables}}$$

**Example: Net exposure in EUR**

| Position Type | Amount (EUR) |
|--------------|-------------:|
| Receivables  | +10,000,000 |
| Payables     | −7,000,000  |
| **Net Exposure** | **+3,000,000** |


Once net positions are estimated, they should be converted into home-currency terms so that management can compare exposures across currencies on a common basis.

$$ \boxed{\text{Home-currency exposure} = \text{Net foreign-currency position} \times \text{Expected future spot rate}}$$


**Why net exposure matters:**
- avoid unnecessary hedging  
- lower transaction costs, e.g., bid-ask spreads and bank fees  
- reduce administrative complexity  
- concentrate on the currencies that create the largest true risk  

---



## <span style="color:#4A86C5;">5.2. Natural Hedging</span>
---

- Netting is a **measurement step** that aggregates FC receivables and payables
- Natural hedging is an **operational decision** that reduces exposure internally before using derivatives.

A firm creates a natural hedge by aligning FC inflows and outflows so that ER movements offset each other automatically.

Common ways to create a natural hedge:
- earning and spending in the same currency  
- borrowing in the same currency as revenues  
- sourcing inputs in the same market where products are sold  
- diversifying production and sourcing across currencies  
- using risk-sharing clauses in contracts  


**Example:**

| Currency | Inflows | Outflows | Net Exposure |
|---|---:|---:|---:|
| EUR | 5,000,000 | 5,000,000 | 0 |
| JPY | 8,000,000 | 3,000,000 | +5,000,000 |

- The **EUR position is naturally hedged**
- Only the **JPY net exposure** remains to be managed financially

**Netting vs Natural Hedging**

| Concept | What it does | Nature |
|---|---|---|
| **Netting** | Combines receivables and payables in the same currency to measure the true exposure | Treasury / measurement tool |
| **Natural hedging** | Changes the firm’s operations or financing so inflows and outflows offset each other automatically | Operational / internal hedge |

### <span style="color:#5b9bd5;">Question 5: Comprehensive Hedging Decision

**4. Hedge priority:**

### <span style="color:#5b9bd5;">Solution 2: Forward Hedge vs. Money Market Hedge

**1. Forward hedge:**

$$\text{JPY cost} = \$5{,}000{,}000 \times 147.20 = ¥736{,}000{,}000$$

**2. Money market hedge (hedging a payable in USD):**

**Step 1:** Calculate the PV of the USD payable (discount at USD interest rate for 180 days):
$$\text{USD to invest today} = \frac{5{,}000{,}000}{1 + 0.05/2} = \frac{5{,}000{,}000}{1.025} = \$4{,}878{,}048.78$$

**Step 2:** Convert USD to JPY at today's spot rate:
$$\text{JPY needed today} = 4{,}878{,}048.78 \times 148.50 = ¥724{,}390{,}243.90$$

**Step 3:** The firm must borrow JPY today (or use JPY cash). Borrow ¥724,390,243.90 at 0.4% annualized (0.2% for 180 days):
$$\text{JPY repayment in 180 days} = 724{,}390{,}243.90 \times 1.002 = ¥725{,}839{,}024.38$$

**Step 4:** The USD investment matures to exactly $5,000,000, which pays the supplier. Total JPY cost = **¥725,839,024**.

**3. Comparison:**
- Forward hedge: ¥736,000,000
- Money market hedge: ¥725,839,024
- **Money market hedge is cheaper by ¥10,160,976** (approximately ¥10.2 million)

**4. CIP-implied forward rate:**

$$F = S_0 \times \frac{1 + i_{JPY}}{1 + i_{USD}} = 148.50 \times \frac{1 + 0.004/2}{1 + 0.05/2} = 148.50 \times \frac{1.002}{1.025} = 148.50 \times 0.97756 = ¥145.17/\text{USD}$$

The CIP-implied forward rate is **¥145.17/USD**, but the quoted forward is **¥147.20/USD**. The quoted forward overprices the USD relative to CIP, making it more expensive than the money market hedge. This discrepancy could arise from: transaction costs, credit risk premia in the forward market, market frictions, or temporary supply-demand imbalances.

### <span style="color:#5b9bd5;">Solution 3: Option Hedge — Put Option for a Receivable

**1. Exercise decision and net GBP inflow per scenario:**

The put option gives the right to sell USD at $K$ = GBP 0.7850/USD. Exercise if $S_T < K$.

FV of premium (at GBP 90-day rate = 4.8%/4 = 1.2%):
$$\text{FV premium} = 0.0090 \times 1.012 = \text{GBP } 0.009108 \text{ per USD}$$

| Scenario | $S_T$ | Exercise? | Gross rate | − FV Premium | Net rate | × $8m | Net GBP |
|---|---|---|---|---|---|---|---:|
| USD strengthens | 0.8000 | No ($S_T > K$) | 0.8000 | 0.009108 | 0.790892 | × 8m | **£6,327,136** |
| Stable | 0.7870 | No ($S_T > K$) | 0.7870 | 0.009108 | 0.777892 | × 8m | **£6,223,136** |
| USD weakens | 0.7600 | Yes ($S_T < K$) | 0.7850 | 0.009108 | 0.775892 | × 8m | **£6,207,136** |
| USD crashes | 0.7300 | Yes ($S_T < K$) | 0.7850 | 0.009108 | 0.775892 | × 8m | **£6,207,136** |

**2. Expected value:**

$$E(V) = 0.20(6{,}327{,}136) + 0.45(6{,}223{,}136) + 0.25(6{,}207{,}136) + 0.10(6{,}207{,}136)$$
$$= 1{,}265{,}427 + 2{,}800{,}411 + 1{,}551{,}784 + 620{,}714 = \textbf{£6,238,336}$$

**3. Forward hedge:**

$$\text{GBP proceeds} = \$8{,}000{,}000 \times 0.7890 = \textbf{£6,312,000}$$

**4. Comparison:**

| Strategy | GBP proceeds | Risk |
|---|---|---|
| Forward | £6,312,000 | Zero — locked in |
| Option (expected) | £6,238,336 | Variable (£6,207,136 – £6,327,136) |

The forward hedge offers **higher expected value** and **zero risk**. The option would be preferred if:
- The USD receivable is **uncertain** (e.g., the sale might be cancelled) — then the option avoids the obligation to deliver USD
- The firm has a **strong conviction** that USD will appreciate significantly beyond 0.8000, making the unlimited upside valuable
- The firm's **risk tolerance** allows for the premium cost in exchange for flexibility

In this case, with only a 20% chance of the best scenario and 35% combined probability of USD weakness, the forward is clearly superior for a **certain** cash flow.

### <span style="color:#5b9bd5;">Solution 4: Multi-Currency Netting

**1. Consolidated net exposure per currency:**

**EUR** (Germany + France combined):
- Receivables: €6,000,000 + €1,500,000 = €7,500,000
- Payables: €2,500,000 + €4,000,000 = €6,500,000
- **Net: +€1,000,000** (net long)

**GBP:**
- Net: £5,000,000 − £1,200,000 = **+£3,800,000** (net long)

**JPY:**
- Net: ¥300,000,000 − ¥650,000,000 = **−¥350,000,000** (net short / payable)

**2. Convert to USD:**

| Currency | Net FC | × Spot | USD equivalent |
|---|---:|---:|---:|
| EUR | +1,000,000 | × 1.09 | **+$1,090,000** |
| GBP | +3,800,000 | × 1.28 | **+$4,864,000** |
| JPY | −350,000,000 | × 0.0067 | **−$2,345,000** |
| | | **Total net** | **+$3,609,000** |

The firm has a total net long position of approximately **$3.6 million**.

**3. Gross vs. net hedging volume:**

**Gross hedging** (hedge every receivable and payable separately):

| Currency | Gross receivables (USD) | Gross payables (USD) | Total gross |
|---|---:|---:|---:|
| EUR | 7,500,000 × 1.09 = $8,175,000 | 6,500,000 × 1.09 = $7,085,000 | $15,260,000 |
| GBP | 5,000,000 × 1.28 = $6,400,000 | 1,200,000 × 1.28 = $1,536,000 | $7,936,000 |
| JPY | 300m × 0.0067 = $2,010,000 | 650m × 0.0067 = $4,355,000 | $6,365,000 |
| **Total** | | | **$29,561,000** |

**Net hedging volume:** $|1,090,000| + |4,864,000| + |2,345,000| = **$8,299,000**

$$\text{Reduction} = \frac{29{,}561{,}000 - 8{,}299{,}000}{29{,}561{,}000} = \frac{21{,}262{,}000}{29{,}561{,}000} = \textbf{71.9\%}$$

Netting eliminates approximately **72% of hedging volume**, significantly reducing transaction costs and bank fees.

**4. Natural hedging for JPY:**
The firm is net short ¥350 million (payables exceed receivables). To create a natural hedge:
- **Increase JPY revenue:** shift some sales to Japanese customers, invoicing in JPY
- **Source JPY inputs:** procure components or raw materials from Japanese suppliers, creating JPY payables that offset existing JPY receivables from future sales
- **Borrow in JPY:** take a JPY-denominated loan, creating a JPY outflow (debt service) that offsets future JPY inflows if the firm expands Japanese sales
- **Relocate production:** if feasible, open a small production facility in Japan so that costs naturally match the currency of future revenue growth

### <span style="color:#5b9bd5;">Solution 5: Comprehensive Hedging Decision

**1. Forward hedge:**

$$\text{EUR proceeds} = \$20{,}000{,}000 \times 0.9260 = \textbf{€18,520,000}$$

**2. Money market hedge:**

**Step 1:** Borrow the PV of USD receivable at USD 120-day rate (5.2% annualized = 1.7333% for 120 days):
$$\text{USD borrowed today} = \frac{20{,}000{,}000}{1 + 0.052 \times (120/360)} = \frac{20{,}000{,}000}{1.01733} = \$19{,}659{,}370.44$$

**Step 2:** Convert to EUR at spot:
$$\text{EUR today} = 19{,}659{,}370.44 \times 0.9200 = €18{,}086{,}620.80$$

**Step 3:** Invest EUR at 3.4% annualized (= 1.1333% for 120 days):
$$\text{EUR in 120 days} = 18{,}086{,}620.80 \times 1.01133 = \textbf{€18,291,616.19}$$

**Step 4:** When USD 20m arrives, repay the USD loan. EUR proceeds locked in = **€18,291,616**.

**3. ATM put option** ($K$ = 0.9200, premium = €0.0105/USD):

FV of premium (at EUR 120-day rate): €0.0105 × 1.01133 = €0.010619/USD
Total FV premium: $20m × €0.010619 = €212,380

| Scenario | $S_T$ | Exercise? | Gross EUR ($20m ×) | − FV Premium | Net EUR |
|---|---|---|---:|---:|---:|
| Strong USD | 0.9400 | No | €18,800,000 | €212,380 | **€18,587,620** |
| Mild strength | 0.9300 | No | €18,600,000 | €212,380 | **€18,387,620** |
| Stable | 0.9200 | Indifferent | €18,400,000 | €212,380 | **€18,187,620** |
| Mild weakness | 0.9050 | Yes ($K$) | €18,400,000 | €212,380 | **€18,187,620** |
| Sharp weakness | 0.8700 | Yes ($K$) | €18,400,000 | €212,380 | **€18,187,620** |

$$E(V)_{\text{ATM}} = 0.15(18{,}587{,}620) + 0.30(18{,}387{,}620) + 0.25(18{,}187{,}620) + 0.20(18{,}187{,}620) + 0.10(18{,}187{,}620)$$
$$= 2{,}788{,}143 + 5{,}516{,}286 + 4{,}546{,}905 + 3{,}637{,}524 + 1{,}818{,}762 = \textbf{€18,307,620}$$

**4. OTM put option** ($K$ = 0.9100, premium = €0.0055/USD):

FV of premium: €0.0055 × 1.01133 = €0.005562/USD
Total FV premium: $20m × €0.005562 = €111,246

| Scenario | $S_T$ | Exercise? | Gross EUR ($20m ×) | − FV Premium | Net EUR |
|---|---|---|---:|---:|---:|
| Strong USD | 0.9400 | No | €18,800,000 | €111,246 | **€18,688,754** |
| Mild strength | 0.9300 | No | €18,600,000 | €111,246 | **€18,488,754** |
| Stable | 0.9200 | No | €18,400,000 | €111,246 | **€18,288,754** |
| Mild weakness | 0.9050 | No | €18,100,000 | €111,246 | **€17,988,754** |
| Sharp weakness | 0.8700 | Yes ($K$) | €18,200,000 | €111,246 | **€18,088,754** |

$$E(V)_{\text{OTM}} = 0.15(18{,}688{,}754) + 0.30(18{,}488{,}754) + 0.25(18{,}288{,}754) + 0.20(17{,}988{,}754) + 0.10(18{,}088{,}754)$$
$$= 2{,}803{,}313 + 5{,}546{,}626 + 4{,}572{,}189 + 3{,}597{,}751 + 1{,}808{,}875 = \textbf{€18,328,754}$$

**5. Ranking and recommendation:**

| Rank | Strategy | Expected EUR | Risk |
|---:|---|---:|---|
| 1 | **Forward hedge** | **€18,520,000** | Zero |
| 2 | OTM option ($K$ = 0.91) | €18,328,754 | Variable (€17,988,754 – €18,688,754) |
| 3 | ATM option ($K$ = 0.92) | €18,307,620 | Variable (€18,187,620 – €18,587,620) |
| 4 | Money market hedge | €18,291,616 | Zero |

**Recommendation:** The **forward hedge** is optimal. It offers the highest expected proceeds (€18,520,000) with zero risk. The money market hedge is risk-free but delivers €228,384 less due to the interest-rate differential. The OTM option offers more upside potential than the ATM option (lower premium cost) but exposes the firm to more downside. The ATM option has a tighter range but the premium cost is high.

The forward would only be suboptimal if the firm believed with high confidence that the USD would strengthen well beyond 0.9400 — but even then, the option's premium erodes much of the upside gain.
