# 1. Introduction and Capital Budgeting Refresher</span>

## <span style="color:#4A86C5;">1.1. Why multinational capital budgeting matters</span>
---
- Objective remains the same: determine whether a project creates value for shareholders.
- Multinational capital budgeting extends to an international setting.

- Compared with domestic projects, multinational projects require firms to consider:
    - exchange rate (ER) movements,
    - remittance restrictions,
    - tax differences, and
    - country risk.

## <span style="color:#4A86C5;">1.2. Core capital budgeting principles</span>
---
International projects add more complications, but they are still evaluated using the same core principles.

| Principle | What it means | Why it still matters in multinational projects | Mini example |
|---|---|---|---|
| Incremental cash flows | Include only CFs that occur because the project is undertaken | We care only about the project’s additional effect on firm value | New factory generates $1M revenue → include |
| After-tax cash flows | Use after-tax CFs, not before tax | Taxes determine the actual cash received by the firm | $100 profit with 30% tax → CF = $70 |
| Ignore sunk costs | Exclude costs that have already been incurred | Past spending cannot be changed and should not affect decisions | $2M already spent on R&D → ignore |
| Include opportunity costs | Include the value of the next best alternative use of resources | Using existing assets still has an economic cost | Land could be sold for $5M → include $5M as cost |
| Exclude interest expense | Do not deduct financing costs from project CFs | Financing is already reflected in the discount rate | Do not subtract loan interest when calculating project CFs |

These principles ensure that project evaluation remains focused on
- relevant CFs, 
- true economic costs, and
- value creation.

In multinational capital budgeting, the challenge is applying these rules correctly when exchange rates, taxes, remittances, and country risk make the CFs more difficult to estimate.

---


## <span style="color:#4A86C5;">1.3. Operating cash flow
---
A useful starting point is:

$$\boxed{OCF = (Rev - Costs - Depreciation)(1-T_c) + Depreciation}$$


Where:

- $Rev$ = revenue  
- $Costs$ = cash operating costs  
- $Depreciation$ = depreciation  
- $T_c$ = tax rate  

This formula still applies in multinational settings, but we must now also ask:

- In **which currency** are these CFs measured?
- Who receives them first: the **subsidiary** or the **parent**?
- Can all of them actually be **remitted** to the parent firm?

# 2. Multinational Capital Budgeting Framework</span>
---



## <span style="color:#4A86C5;">2.1. Core Decision Rule</span>
---

- The key rule is to evaluate using **NPV from the parent’s perspective**.

- Accept if:
$$
\boxed{NPV_{parent} > 0}
$$

- A positive NPV means the project is expected to add value to the MNC.

**Example**
- A foreign subsidiary may appear profitable in local currency.
- However, once taxes, ER changes, and remittance restrictions are included, the NPV to the parent may be much lower or even negative.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> In multinational finance, the relevant question is ***“Does the project create value for the parent company?”***

---

## <span style="color:#4A86C5;">2.2. Parent vs Subsidiary Cash Flows</span>
---

- A central idea in multinational capital budgeting is:

> Local project value does not necessarily equal parent value

Key adjustments include:
- Host-country taxes
- Withholding taxes on remittances
- Blocked funds or remittance restrictions
- Transfer pricing rules
- Management fees or royalties
- ER movements
- Home-country taxes

---

> <span style="color:#9e86C5;"><strong>**Question**:</strong></span><br>
> **Scenario:**
>   - A US parent is considering a one-year project in the UK.
>   - The project will generate **£1,000,000** in profit. 
>   - The UK corporate tax rate is **25%**.
>   - The UK government also imposes a **10% withholding tax** on any funds remitted to the US.
>   - The current ER is **$1.30/£**.
>
> Calculate the net CF received by the US parent in USD.
<br>
</br>

> <span style="color:#9e86C5;"><strong>**Answer**:</strong></span><br>
- Calculate after-tax profit in the UK 
    $$£1{,}000{,}000 \times (1 - 0.25) = £750{,}000$$

- Apply withholding tax on remittance  
    $$£750{,}000 \times (1 - 0.10) = £675{,}000$$

- Net CF received by the US parent in USD 
    $$£675{,}000 \times 1.30 = \boxed{\$877{,}500}$$

---


## <span style="color:#4A86C5;">2.3. MNC's Project Evaluation Process
---
The evaluation process can be written as:

1. Estimate incremental OCFs in the host-country currency
2. Adjust for taxes, working capital, and terminal value
3. Estimate what portion can actually be remitted
4. Convert remittable CFs into the parent currency
5. Discount them at an appropriate RRR
6. Calculate NPV

# 3. Exchange Rate Effects in Project Valuation</span>
---


## <span style="color:#4A86C5;">3.1. Why exchange rates matter

A multinational project often generates CFs in the host-country currency, while the parent firm values the project in its own home currency.

This means the project’s value depends not only on operating performance, but also on how ERs evolve.

ER changes can affect:

- OCFs: inflows & outflows
- remittances to the parent
- terminal value on disposal
- overall project NPV

> <span style="color:#4A86C5;">**Key intuition:**</span>
> The parent firm is interested in the amount of home currency it will receive.
---


## <span style="color:#4A86C5;">3.2. Forecasting exchange rates</span>

- In practice, firms do not know future ERs with certainty. 
- As a result, ER forecasts are always subject to error and should be treated as assumptions rather than precise predictions.

Several approaches are commonly used:

| Approach | Main idea | Usefulness |
|---|---|---|
| Spot rate | Assume the current exchange rate continues | Simple benchmark, but often unrealistic |
| Forward rate | Use market-implied future exchange rate | Widely used in practice |
| Purchasing Power Parity (PPP) | Exchange rates adjust based on inflation differences | Useful for long-run intuition |

---

### <span style="color:#4A86C5;">3.2.1. Purchasing Power Parity (PPP)

- PPP suggests that currencies with higher inflation tend to **depreciate over time** relative to currencies with lower inflation.
$$
\boxed{\% \Delta S \approx \pi_{home}​ - \pi_{foreign}}​$$

- This provides a useful guide, but PPP is not a reliable short-term forecasting tool. 
- In capital budgeting, it should be used to support intuition rather than as a precise estimate.

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - ER assumptions directly affect projected CFs and can change NPV.
> - A project that appears profitable under one ER assumption may become unattractive under another. 
> - Therefore, ER forecasts should always be tested using **sensitivity or scenario analysis**.

---

# 4. Risk Analysis in Multinational Capital Budgeting</span>
---


## <span style="color:#4A86C5;">4.1. Why risk analysis is essential
---
An NPV estimate depends on assumptions, and those assumptions are uncertain. 

For multinational projects, uncertainty exists in more dimensions:

- sales volume
- input costs
- ERs
- inflation
- tax rules
- political intervention
- remittance restrictions

Therefore, managers should not rely on a single estimated NPV.

---



## <span style="color:#4A86C5;">4.2. Sensitivity analysis
---
Sensitivity analysis examines how NPV changes when **one variable changes**, holding the others constant.

This is useful for identifying the most important project driver.

For multinational projects, common variables include:

- exchange rate
- sales volume
- local operating costs
- salvage value
- tax rate

**Example:**

- Suppose a project is expected to generate EUR 1,000 next year under the base case.
- We examine how ER changes affect the value to the parent.

| Exchange rate (AUD/EUR) | Parent cash flow (AUD) | Change vs base |
|---|---:|---:|
| 1.60 (Favorable) | 1,600 | +200 |
| 1.40 (Base case) | 1,400 | 0 |
| 1.20 (Unfavorable) | 1,200 | −200 |

- Even though the subsidiary generates the same EUR CF, the value to the parent changes materially.
- The project is highly sensitive to ERs.

---



## <span style="color:#4A86C5;">4.3. Scenario analysis
---
Scenario analysis changes **multiple assumptions at the same time**.

This is often more realistic for multinational projects because variables are usually related. For example:

- a weak economy may reduce sales,
- weaken the local currency,
- and increase country risk simultaneously.

**Example:**

| Scenario | Sales | Exchange rate | Country conditions | NPV |
|---|---|---|---|---:|
| Optimistic | High | Favourable | Stable | 700,000 |
| Base | Expected | Expected | Normal | 300,000 |
| Pessimistic | Low | Unfavourable | Tightened controls | -250,000 |

Scenario analysis helps show that project failure often comes from a **combination** of adverse conditions, not from one isolated variable.

---


## <span style="color:#4A86C5;">4.4. Break-even analysis</span>

Break-even analysis asks:

> At what level of a key input does the project’s NPV become zero?

Applications in multinational projects
- Break-even ER
- Break-even sales volume
- Break-even remittance proportion

**Example:**
- US parent company has a subsidiary in Australia that is expected to generate AUD CFs, which will be converted into USD for the parent.
- Suppose the break-even ER is 1.30 AUD/USD:
    - **Current ER = 1.50 AUD/USD**→ comfortable margin  
        - The parent receives stronger USD-equivalent cash inflows  
        - The project remains positive NPV even if the AUD weakens somewhat

    - **Current ER = 1.32 AUD/USD** → very risky  
        - The expected ER is only slightly above the break-even level  
        - A small depreciation of the AUD could push the project into negative NPV

In Excel, break-even values can be found using:
- **Goal Seek**
- **Solver**

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - If the actual value is far from the break-even point → project is relatively safe  
> - If it is close to the break-even point → project is highly risky  
---


## <span style="color:#4A86C5;">4.5. Country risk analysis
---
A multinational project may face risks that do not normally arise in domestic projects:

- political instability
- regulatory change
- expropriation risk
- tax changes
- currency controls
- restrictions on dividend remittance

These risks matter because they can alter either:

- expected CFs, or
- the required return used to discount them.

---


## <span style="color:#4A86C5;">4.6. Practical treatment of country risk</span>

Risks can be incorporated in two main ways:

<div style="width: 48%; padding: 18px; border-left: 6px solid #4A86C5; background-color: #F7FBFF; border-radius: 8px;">
<h3 style="margin-top: 0; color: #4A86C5;">Cash-flow Adjustment</h3>

<p style="margin-bottom: 8px;"><b>Managers ask:</b></p>
<ul>
  <li>How do shocks affect project CFs?</li>
  <li>How sensitive is value to key assumptions?</li>
</ul>

<p style="margin-bottom: 8px;"><b>Tools:</b></p>
<ul>
  <li><b>Scenario analysis</b> → Base / Stress</li>
  <li><b>Sensitivity analysis</b> → Change one variable at a time</li>
</ul>

<p><b>Focus:</b> Adjust projected CFs</p>
</div>

<div style="width: 48%; padding: 18px; border-left: 6px solid #4A86C5; background-color: #FCFCFC; border-radius: 8px;">
<h3 style="margin-top: 0; color: #4A86C5;">Discount-rate Adjustment</h3>

<p style="margin-bottom: 8px;"><b>Managers ask:</b></p>
<ul>
  <li>Is this project riskier than normal?</li>
  <li>What return compensates for that risk?</li>
</ul>

<p style="margin-bottom: 8px;"><b>Approach:</b></p>
<ul>
  <li>Increase the required return</li>
  <li>Apply a higher discount rate to future CFs</li>
</ul>

<p><b>Focus:</b> Adjust required return</p>
</div>

</div>

**Example:**  
If managers are worried about exchange rate depreciation, they may:
- reduce expected CFs under a downside scenario, or
- increase the DR to reflect higher risk.

> <span style="color:#4A86C5;">**Key intuition:**</span>  
> Country risk can be reflected in either **cash flows** or the **discount rate**, depending on how the risk is best modelled.

**Important:** Do not adjust risk in both CFs and DRs without clear justification. Otherwise risk may be **double counted**.

# 5. Cost of Capital in Multinational Projects</span>
---


## <span style="color:#4A86C5;">5.1. Role of cost of capital
---
The cost of capital is the required return used to discount project cash flows. Its purpose is to reflect the project’s risk.

The key principle remains:

> Cash flows should be discounted at a rate consistent with their risk.

---



## <span style="color:#4A86C5;">5.2. Currency consistency
---
An important rule in multinational valuation is that the currency of the cash flows must be consistent with the discount rate.

| Cash flows measured in | Discount rate should be in |
|---|---|
| AUD | AUD-based required return |
| USD | USD-based required return |
| EUR | EUR-based required return |

Mixing currencies and discount rates creates valuation errors.

---


## <span style="color:#4A86C5;">5.3. WACC framework
---
A useful benchmark remains the standard WACC framework:

$$
\boxed{WACC = \frac{E}{V}R_e + \frac{D}{V}R_d(1-t_c)}
$$

Where:

- $R_e$ = cost of equity
- $R_d$ = cost of debt
- $t_c$ = tax rate

Interpretation:
- If a firm increases the proportion of low-cost debt, WACC may fall.
- But if leverage becomes too high, both debt and equity investors may demand higher returns.

WACC framework still applies for multinational projects, but modifications may be required due to:
- different debt costs in different countries
- different tax systems
- different inflation and currency environments
- different political and sovereign risks
- different market integration assumptions

Therefore, the relevant RRR for a foreign project may not simply equal the parent's domestic WACC.

---


```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# === Dropdown inputs ===
cost_of_debt_dd = widgets.Dropdown(
    options=[(f"{i}%", i/100) for i in range(2, 11)],  # 2%–10%
    value=0.05,
    description='Cost of Debt:'
)

cost_of_equity_dd = widgets.Dropdown(
    options=[(f"{i}%", i/100) for i in range(6, 21)],  # 6%–20%
    value=0.10,
    description='Cost of Equity:'
)

tax_rate_dd = widgets.Dropdown(
    options=[(f"{i}%", i/100) for i in range(10, 41, 5)],  # 10%–40%
    value=0.25,
    description='Tax Rate:'
)

# === Plot function ===
def plot_wacc(cost_of_debt, cost_of_equity, tax_rate):
    debt_ratio = np.linspace(0, 1, 101)
    equity_ratio = 1 - debt_ratio
    
    after_tax_kd = cost_of_debt * (1 - tax_rate)
    wacc = debt_ratio * after_tax_kd + equity_ratio * cost_of_equity
    
    kd_line = np.full_like(debt_ratio, after_tax_kd)
    ke_line = np.full_like(debt_ratio, cost_of_equity)
    
    plt.figure(figsize=(10, 6))
    plt.plot(debt_ratio * 100, ke_line * 100, label="Cost of Equity")
    plt.plot(debt_ratio * 100, kd_line * 100, label="After-Tax Cost of Debt")
    plt.plot(debt_ratio * 100, wacc * 100, label="WACC", linewidth=2)
    
    plt.xlabel("Debt Ratio (%)")
    plt.ylabel("Cost of Capital (%)")
    plt.title("WACC vs. Capital Structure")
    plt.grid(True)
    plt.legend()
    plt.show()

# === Interactive output ===
ui = widgets.VBox([cost_of_debt_dd, cost_of_equity_dd, tax_rate_dd])

out = widgets.interactive_output(
    plot_wacc,
    {
        'cost_of_debt': cost_of_debt_dd,
        'cost_of_equity': cost_of_equity_dd,
        'tax_rate': tax_rate_dd
    }
)

display(ui, out)
```

## <span style="color:#4A86C5;">5.4. Approaches to applying cost of capital
---

| Approach | Description | Strength | Limitation |
|---|---|---|---|
| Parent WACC | Use parent firm’s standard required return | Simple | May ignore project-specific country risk |
| Local required return | Use host-country risk environment | Reflects local conditions | May be difficult to estimate consistently |
| Adjusted WACC | Start with parent WACC and add a risk premium | Practical compromise | Requires judgment |

**Example:**

Suppose:

- parent WACC = 10%
- estimated additional country risk premium = 3%

Then a possible adjusted required return is:


$$\text{Adjusted required return} = 13\%$$


This does not mean every project in that country should always use 13%, but it illustrates the logic of incorporating country-specific risk into the discount rate.

---


## <span style="color:#4A86C5;">5.5. Separation of investment and financing decisions
---
Project cash flows should be evaluated independently from financing cash flows.

Therefore:

- **interest expense should not be included** in project operating cash flows
- **loan repayments should not be deducted** from project cash flows for NPV purposes

Why?

Because financing costs are already reflected in the discount rate.

Including them again would understate project value through double counting.

**Example:** common mistake

Suppose a foreign project is partly funded with local debt. If interest payments are subtracted from project cash flows and then those reduced cash flows are discounted using WACC, the project is being penalized twice for financing.

That is incorrect.

Key insight
A project should be judged on whether its **assets and operating cash flows create value**. Financing affects how the project is funded, but not whether the project itself is economically worthwhile.


---


### Integrated example

Assume:

- Initial investment = **AUD 1,000**
- Expected annual subsidiary cash flow = **USD 400** for 3 years
- 100% can be remitted
- Expected exchange rate = **1 USD = 1.50 AUD**
- Required return = **10%**

#### Step 1: Convert annual cash flow

$$\text{USD}\ 400 \times 1.50 = \text{AUD}\ 600$$

#### Step 2: Discount annual cash flows

Present value of 3 annual cash flows of AUD 600 at 10%:

$$PV \approx 1,492$$

#### Step 3: Compute NPV

$$NPV = 1,492 - 1,000 = 492$$

So the project appears acceptable.

---

### Now add exchange rate risk

If the exchange rate falls from **1.50** to **1.20 AUD per USD**, annual parent cash flow becomes:

$$\text{USD}\ 400 \times 1.20 = \text{AUD}\ 480$$

Now the project’s NPV is much lower.

This example shows that:

- operating performance may be unchanged,
- but value to the parent can fall substantially due to exchange rate changes.

## <span style="color:#4A86C5;">Solutions

---

### <span style="color:#5b9bd5;">Solution 1: Parent vs Subsidiary Cash Flow

**1.** After-tax BRL: $10{,}000{,}000 \times (1 - 0.30) = 7{,}000{,}000$ BRL.

**2.** After withholding: $7{,}000{,}000 \times (1 - 0.15) = 5{,}950{,}000$ BRL → in USD: $5{,}950{,}000 / 5.0 = \$1{,}190{,}000$.

**3.** The subsidiary appears to earn BRL 10 m, but only **\$1.19 m** reaches the parent — a 40% reduction. This shows that **subsidiary-level profitability** can systematically overstate the value to the parent. The relevant NPV is always from the **parent's perspective**, after taxes and remittance frictions.

---

### <span style="color:#5b9bd5;">Solution 2: Exchange-Rate Sensitivity

Annuity factor at 10%, 3 years: $\dfrac{1 - 1.10^{-3}}{0.10} = 2.4869$.

**1. Base case** ($1.60$ AUD/EUR):
Annual AUD CF = 1.60 m AUD. PV = $1.60 \times 2.4869 = 3.979$ m. **NPV = 3.979 − 1.50 = +AUD 2.479 m**.

**2. Unfavourable** ($1.40$):
Annual AUD CF = 1.40 m. PV = $1.40 \times 2.4869 = 3.482$ m. **NPV = +AUD 1.982 m**.

**3. Favourable** ($1.80$):
Annual AUD CF = 1.80 m. PV = $1.80 \times 2.4869 = 4.476$ m. **NPV = +AUD 2.976 m**.

**4.** A 12.5% change in the EUR/AUD rate (1.60 → 1.40 or 1.80) moves NPV by ~AUD 0.5 m (~20%). **NPV is highly sensitive to exchange-rate assumptions** — the project should be hedged or stress-tested.

---

### <span style="color:#5b9bd5;">Solution 3: Break-Even Exchange Rate

Annuity factor at 12%, 4 years: $\dfrac{1 - 1.12^{-4}}{0.12} = 3.0373$.

Set NPV = 0:

$$0 = -8 + 50 \cdot S \cdot 3.0373$$

$$S = \dfrac{8}{50 \times 3.0373} = \dfrac{8}{151.87} = 0.0527 \ \text{USD/MXN}$$

**1.** Break-even rate ≈ **0.0527 USD/MXN**.

**2.** Current spot = 0.055; margin = $\frac{0.055 - 0.0527}{0.0527} \approx 4.4\%$. The project barely clears: a small MXN depreciation could turn it negative. **High risk; consider hedging.**

---

### <span style="color:#5b9bd5;">Solution 4: WACC and Country Risk

**1.** Using 9% (parent WACC) on a higher-risk foreign project is **inappropriate** — it ignores the additional country risk that the parent's domestic projects do not face. NPV would be **overstated**.

**2.** Adjusted required return = 9% + 4% = **13%**.

**3.** If political risk has already been built into a **lower expected CF** (e.g., probability-weighted), adding a higher discount rate captures the same risk twice → systematic **underestimation** of project value. Choose **one** lens (CF adjustment or DR adjustment) per identified risk.
