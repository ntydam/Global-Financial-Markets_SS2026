## <span style="color:#4A86C5;">1.3.Core Building Blocks of Global Financial Markets

---
Financial assets represent claims on future cash flows (CFs), they include:

- **Debt securities**  
  - Government bonds and corporate bonds that promise fixed or variable payments and are often used as benchmarks for pricing other assets.

- **Equity securities**  
  - Shares of stock that represent ownership in firms and claims on residual profits.

- **Derivatives**  
  - Financial contracts whose values are derived from underlying assets such as bonds, equities, interest rates, or exchange rates (ERs), 
  - Commonly used for risk management and speculation.

Together, these assets form the backbone of the global financial system and support the operation of international investment and multinational business activity.

# 2. Define a Multinational Corporation (MNC)</span>

---
## <span style="color:#4A86C5;">2.1. Definition of an MNC</span>
---

An MNC is a firm that:
- engages in international business activities and
- generates CFs linked to more than one country.

### <span style="color:#5b9bd5;">Multinational Activities:
An MNC typically has at least one of the following:
- **Foreign sales** (exporting or selling through overseas distributors)
- **Foreign sourcing** (importing inputs or outsourcing production abroad)
- **Foreign assets** (subsidiaries, plants, offices, or acquisitions in other countries)
- **Cross-border financing** (borrowing/lending in foreign currency markets)
- **Foreign direct investment (FDI)** (ownership/control of foreign operations)

> <span style="color:#4A86C5;"> **Key intuition:**</span><br>
> - “Multinational” is about **cross-border CFs and risk**
> - It is not just branding or having customers overseas.

---

### <span style="color:#5b9bd6;">MNC vs. Domestic Firm — Comparison
| Dimension | Purely Domestic Firm | Multinational Corporation |
|---|---|---|
| **Revenue drivers** | Driven primarily by domestic demand and the domestic business cycle. | Driven by both domestic and foreign demand, reflecting multiple business cycles across countries. |
| **Cost structure** | Costs determined by domestic input prices (labor, materials, regulation). | Costs influenced by domestic and foreign inputs, including wages, commodities, logistics, tariffs, and foreign regulations. |
| **Currency of CFs** | Revenues and costs denominated and tied to one economy. | Revenues and costs occur in multiple currencies and linked to multiple economies. |
| **Financial exposure** | Exposed mainly to <br> <i> - domestic inflation, <br> <i> - domestic interest rates, and <br> <i> - local competition. | Exposed to: <br> <i> - foreign exchange (FX) risk, <br> <i> - cross-country interest rate differentials, <br> <i> - country risk (political, legal, regulatory), and <br> <i> - international competitive effects. |
| **Financing options** | Relies on domestic capital markets and domestic currency funding. | Accesses both domestic and international capital markets, with multi-currency financing options. |

**Example:**
- Two firms both expect the equivalent of £100 million in operating profit next year.
    - Firm A earns it entirely in the UK. 
    - Firm B earns one-third in the UK, one-third in euros, and one-third in pesos. 
- On paper they look similar. 
- In finance they are not.
    - Firm B's profit depends not only on operations, but also on exchange rates, local inflation, transfer rules, and country risk.

## <span style="color:#4A86C5;">2.2. Core exposures by the MNC

---
### <span style="color:#4A86C5;">Foreign exchange exposure
- FX risk arises because MNCs generate revenues, incur costs, and hold assets and liabilities in **multiple currencies**. 
- When ERs (ERs) fluctuate, the **home-currency value of CFs** and **firm value** changes, even if operations remain unchanged.

- FX risk can:
    - alter realized CFs from international transactions,
    - affect long-term competitiveness across markets, and
    - distort reported financial performance through accounting translation.

- As a result, FX exposure is a **core strategic and financial concern** for MNCs, influencing pricing, sourcing, financing, and hedging decisions.

- Kinds of exposure:
    - Transaction exposure: Invoice
    - Economic exposure: Business activity 
    - Translation exposure: Accounting

---

### <span style="color:#4A86C5;">Interest rate exposure

- MNCs operate across countries with **different monetary policies and interest rate cycles**. 
- Changes in interest rates directly affect:
  - borrowing costs, 
  - discount rates, and 
  - investment incentives.

- IR movements can:
  - change the cost of servicing debt and raising new capital,
  - affect project valuation through discounting, and
  - influence cross-border financing and investment location decisions.

> For MNCs, IR risk is often **interlinked** with ER risk, since funding, investing, and repatriating cash typically involve multiple currencies and capital markets.

---

### <span style="color:#4A86C5;">Country risk exposure
- Country risk (geopolitical, legal, regulatory, and tax) affects both investor appetite and investor outcomes.
- When political conditions deteriorate, investors demand a higher return (premium) and may avoid new projects because the probability of capital controls, forced divestment, and contract/regulatory shocks increases.
- Even if a foreign firm can operate and generate profits, transfer and exit restrictions can prevent (or severely discount) converting ruble cashflows into foreign currency and sending them home. 
- In practice, many foreign exits occurred under government approval regimes, with mandatory discounts and additional fiscal “contributions”, leading to large realised losses for global firms.

#### Sovereign CDS spreads and ratings
- **Sovereign CDS spread**:
    - CDS = Credit Default Swap
    - It is a financial derivative contract.
    - The buyer of a CDS pays periodic premiums to the seller.
    - In return, the seller agrees to compensate the buyer if a borrower (the sovereign reference entity) defaults/restructures on its debt.
    - A higher CDS spread reflects a higher sovereign credit risk.
- **Credit ratings** (S&P / Moody’s / Fitch)
    - Ratings provides benchmarks for credit risk and default likelihood.
    - Expressed on ordinal scales (e.g., AAA/Aaa highest quality down to default).
    - If the country's credit rating is suddenly lowered, for example, then its required rate of return will increase.

    - external challenges that affect all businesses across markets, such as:
        - inflation,

## <span style="color:#4A86C5;">2.3. Managing the MNC

---

### <span style="color:#4A86C5;">Objective of the MNC
- The **primary objective** is to maximize shareholder wealth.  
    - For publicly traded firms, this objective is commonly proxied by maximizing the firm’s market value (stock price).
    - Stock price reflects investors’ expectations about:
        - future CFs  
        - the risk associated with those CFs
- **Other objectives** are secondary to shareholder wealth maximization.
    - maximizing sales
    - market share
    - employee welfare

### <span style="color:#4A86C5;">Shareholders' confidence
In countries where managers are required to act in the interests of shareholders:
- managerial decisions are more closely aligned with value maximization, and
- investor confidence is higher, leading to more active participation in financial markets.

As a result, **stock market participation and trading activity** tend to be higher in such environments. 

For example, in FY24, U.S. equity markets represent **49.1%** of the $126.7 trillion in global equity market cap.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Strong institutions and high transparency reduce agency problems, increase investor trust, and ultimately support deeper and more liquid capital markets.

## <span style="color:#4A86C5;">2.4. Agency Problems

---

### <span style="color:#4A86C5;">Definition:

* An agency problem arises when managers’ decisions **do not fully align** with shareholders’ interests.

* Because shareholders cannot pefectly observe or control managers, managers may pursue private benefits (e.g., pay, prestige) rather than firm value. <br></br>

* Agency problems are often more severe in MNCs because:

  * Monitoring is harder (distance + different legal/cultural/reporting systems).
  * Information asymmetry is larger (subsidiaries know more than headquarter).
  * Organizational complexity increases (more units and cross-border CFs).
  * Local objectives may diverge from shareholder value.

* Example:
  * Expanding into a new market mainly to raise the CEO’s prestige despite weak expected returns.
  * A subsidiary manager prioritizing local goals over firm profitability (e.g., low-quality inputs).

### <span style="color:#4A86C5;"> Responses to agency problems and related costs

* MNCs reduce agency problems using:
  * stronger reporting and internal controls,
  * tighter performance evaluation/benchmarking,
  * value-aligned incentives (stock/options, performance bonuses),
  * clear strategy from HQ to subsidiaries,
  * replacing managers in poorly performing subsidiaries.

* Agency costs are paid to limit misalignment, typically higher in MNCs:
  - monitoring costs (audits, oversight),
  - bonding costs (contracts, incentives),
  - residual losses (remaining value loss).

#### <span style="color:#4A86C5;">How SOX Strengthens Corporate Governance</span>

**1. Governance principles imposed by SOX**
- SOX improves the credibility of financial reporting by mandating:
    - robust internal control systems over financial reporting,
    - direct executive accountability for the accuracy of financial statements,
    - stronger oversight and audit independence.

**2. SOX implementation in MNCs**

    - centralize financial information in shared, firm-wide databases,

- SOX controls help to:
    - limit managers’ ability to conceal poor performance,
    - reduce opportunities for manipulation in complex multinational reporting,

### <span style="color:#4A86C5;"> Management Structure of an MNC

The **magnitude of agency costs** varies with the MNC’s **management style**.

The structure you choose determines:
- who has decision rights (parent vs. subsidiaries), 
- how information flows,
- and how effectively the parent can monitor foreign operations.

**Structure types:**
- Centralized MNC financial management
- Decentralized MNC financial management
- Hybrid

---
#### <span style="color:#4A86C5;">Centralized vs. Decentralized Multinational Financial Management</span>

| <span style="color:#4A86C5;">Dimension</span> | <span style="color:#4A86C5;">Centralized MNC<br>(Parent-led control)</span> | <span style="color:#4A86C5;">Decentralized MNC<br>(Subsidiary-led autonomy)</span> |
|---|---|---|
| **What it is** | Key financial decisions are controlled by the parent firm’s financial managers. | Key financial decisions are made by subsidiary managers. |
| **Agency-cost implication** | Parent control:<br>- reduces subsidiary managerial discretion, <br>- limits self-serving behavior and agency costs. | Subsidiary autonomy:<br>- increases the risk that subsidiary managers pursue local objectives rather than firm-wide value. |
| **Policy consistency** | High consistency across countries in cash management, funding, and capital-expenditure discipline. | Policies may differ across subsidiaries, reflecting local priorities and conditions. |
| **Main advantage** | Stronger alignment with MNC value maximization. | Better-informed and faster decisions due to local knowledge. |
| **Main trade-off / risk** | Headquarters may make suboptimal local decisions due to limited knowledge of local markets, regulations, and operating conditions. | Higher agency costs unless incentives and monitoring are carefully designed. |
| **When it works best** | When global coordination and control are critical and information asymmetry is high. | When subsidiary incentives are explicitly aligned with overall MNC value and monitoring is effective. |

# 3. Explain Why Firms Become Multinational

---
## <span style="color:#4A86C5;">3.1. Comparative Advantage Theory</span>
---
### <span style="color:#4A86C5;">Definition:
> The theory explains: <br>
> **Why countries trade with each other and how international trade can make all trading countries better off, <br> even when one country is more efficient at producing everything.**

- It is developed in a simplified world that assumes:
    - free trade (no tariffs or quotas),
    - perfect competition,
    - no uncertainty,
    - costless and perfect information, and
    - no government interference.

---

### <span style="color:#4A86C5;"> Core mechanism: specialization and trade
- Countries should:
    - **specialize** in producing goods for which they have a comparative advantage (lower opportunity costs, given its factors of endowments, such as land, labor, capital, and technology), and
    - **trade** for goods where they have a comparative disadvantage.
- By this, countries can consume beyond their production possibilities frontier, leading to higher welfare because:
    - Resources are allocated to their most efficient uses in each country,
    - The total combined output of Country A and Country B is maximized.
    - Trade enables countries to access a wider variety of goods and services at lower costs than if they produced everything domestically.

---

### <span style="color:#4A86C5;">Why the model does not fully fit the modern world
In reality, the assumptions of the classical model often do not hold due to **market imperfections**:
- Market frictions (e.g., tariffs, quotas, subsidies),
- Uncertainty and information costs.
- Imperfect competition (e.g., monopolies, oligopolies),
- Government intervention (e.g., regulations, trade policies).

---

### <span style="color:#4A86C5;">Why comparative advantage still matters today
Even though the real world is imperfect, comparative advantage remains a useful baseline for understanding trade.

It explains why countries specialize and trade, including in services and digital supply chains, when they are relatively more efficient than others.

Comparative advantage depends on:
- labor skills,
- access to capital,
- technology,
- institutional quality.

It also evolves as economies develop, acquire new capabilities, and adopt new technologies.

---

<div style="border:2px solid #4A86C5; border-radius:10px; padding:18px; background-color:#F4F9FF;">

### <span style="color:#4A86C5;">Example: Competitive Advantage vs. Comparative Advantage

<strong> 1. Comparative advantage (country level)</strong>  
Think of Vietnam and garment manufacturing.

- Vietnam has a comparative advantage because it has abundant labor, which makes clothing production relatively cheaper than in many other countries.  
- This explains why Vietnam specializes in garments and exports them to the world.


<strong> 2. Competitive advantage (firm level)</strong>  
- Now look inside Vietnam.

- Not all garment firms perform equally. Some firms outperform others because they have a competitive advantage, such as:
    <ul>
      <li>better quality control (fewer defects),</li>
      <li>faster delivery to buyers,</li>
      <li>stronger compliance with global standards.</li>
    </ul>

Because of these firm-specific strengths, these companies perform better than other garment producers, even though they operate in the same country.

</div>

<br>

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> - Comparative advantage explains why a country specializes in an industry.
> - Competitive advantage explains why some firms succeed more than others within that industry.

## <span style="color:#4A86C5;">3.2. Imperfect Markets Theory</span>
---

### <span style="color:#4A86C5;">Two Extremes and the Real World</span>

To understand **market imperfections**, it is useful to compare two **theoretical extremes**.  
**Neither** extreme exists in reality, but the gap between them explains why international business occurs.

---

| **Extreme 1: Completely Closed Markets** | **Extreme 2: Completely Perfect (Fully Open) Markets** |
|---|---|
| **Definition**  <br>Countries are economically isolated. No trade, no capital flows, no labor mobility. | **Definition**  <br>All factors of production (labor, capital, technology) move freely across borders at zero cost. |
| **Key Features**  <br>- No imports or exports  <br>- No FDI  <br>- No cross-border labor movement | **Key Features**  <br>- No barriers to trade or investment  <br>- Instant movement of labor and capital  <br>- Identical information everywhere |
| **Economic Consequences**  <br>- Each country must rely solely on domestic resources  <br>- Inefficient production  <br>- Lower overall welfare | **Economic Consequences**  <br>- Resources flow to highest-return locations  <br>- Costs and returns equalize globally  <br>- Comparative advantage disappears |
| **Implication for International Business**  <br>❌ No international business possible | **Implication for International Business**  <br>❌ No incentive for international business |

---

### <span style="color:#4A86C5;">The Real World: Imperfect Markets</span>

In reality, markets lie **between** these two extremes.

- Labor is not fully mobile (due to immigration laws, language barriers, relocation costs).
- Capital faces regulatory, tax, and political constraints.
- Resources such as energy, land, and technology differ across countries.
- Information is asymmetric and costly.

These differences create economic rationale for international trade and FDI, and allow **MNCs** to:
- locate production where labor is cheaper,
- invest where specialized resources are available,
- operate where energy, land, or regulation lowers total cost.

---


### <span style="color:#4A86C5;">Examples of Imperfect Markets</span>

> **Example 1: Apparel manufacturing and Labor**
- Labor cannot freely move from low-income to high-income countries due to:
    - immigration laws and work-visa restrictions,
    - language and cultural barriers,
    - relocation and family costs.  
As a result, firms such as Nike or Gap produce garments in countries with lower labor costs and export finished products to higher-income markets.

> **Example 2: Data Centers and Energy**
- Energy prices differ across countries due to differences in natural resources,regulation, and infrastructure. Thus, production costs and returns differ across countries, creating incentives for FDI.
- Technology companies—often invest in countries with **lower energy costs or reliable access** to energy when locating data centers.
    - Real-world Example: **Iceland attracts data centers due to its:**
        - abundant renewable energy (its two primary sources of generating renewable electricity are geothermal and hydroelectric power, which producer very low-cost, low-carbon, and stable power) and
        - cool climate, which reduce cooling costs.


> <span style="color:#4A86C5;">**Key Intuition:** </span><br>
> International business exists precisely because markets are imperfect. If markets were either fully closed or perfectly open, multinational activity would not be economically justified.

---

#### <span style="color:#4A86C5;">**Stage 1: Domestic Introduction**</span>

- The firm introduces a **new product** in its **home market**.
- It has an **initial advantage**, such as:
  - unmet local demand,
  - superior technology,
  - better design or innovation.
- Home markets are preferred because:
  - information about customers and competitors is clearer,
  - legal and business environments are familiar.

**Key point:** Firms almost always establish themselves **domestically first**.

</div>

<div style="
border:2px solid #4A86C5;

">

#### <span style="color:#4A86C5;">**Stage 2: Exporting to Foreign Markets**</span>

- As the product succeeds at home, **foreign demand emerges**.
- The firm initially serves foreign customers by **exporting** because:
  - it requires lower commitment,
  - logistics are simpler,
  - foreign production risks are avoided.

**Key point:** International expansion typically begins through **exports**, not FDI.

# 4. Identify Major Modes of International Business

## <span style="color:#4A86C5;">4.1. International Trade</span>

---

<div style="
border:2px solid #4A86C5;
border-radius:8px;
padding:16px;
background-color:#F2F6FC;
margin-bottom:16px;
">

### <span style="color:#4A86C5;">Definition:
International trade is a **conservative and low-risk approach** that firms use to:
- enter foreign markets through exporting, or  
- reduce production costs by importing lower-cost inputs.

Compared with FDI, international trade involves **minimal commitment and high flexibility**:
- Firms do not need to commit capital abroad, reducing exposure to:
  - political risk,
  - regulatory uncertainty,
  - ER risk.
- If trade volumes decline, firms can:
  - scale back exports or imports, or  
  - exit trade relationships  
  at relatively low cost.
- Operations can be adjusted quickly without closing factories or selling foreign assets.
- Therefore, international trade is often the **first stage of international expansion** before firms progress to higher-risk strategies such as FDI.

</div>

### <span style="color:#4A86C5;"> How the Internet Facilitates International Trade

The Internet has dramatically **reduced the cost and complexity of international trade**.

#### Product Information & Marketing
- Firms use websites to:
  - list products and prices,
  - advertise to potential importers worldwide,
  - update product lines or prices instantly.
- Importers can check exporters’ websites instead of relying on physical brochures.

#### Ordering & Delivery
- Websites allow firms to:
  - accept orders online,
  - communicate with importers via email,
  - provide shipping and delivery updates.
- Digital products (e.g., software) can be downloaded instantly, eliminating shipping altogether.

#### Logistics & Inventory Management
- Internet-based systems enable:
  - real-time shipment tracking,
  - automated inventory monitoring,
  - automatic reordering when inventory falls below a threshold.
- For firms with multiple warehouses, the Internet allows them to:
  - operate as an integrated network,
  - fulfill orders from alternative locations.

> <span style="color:#4A86C5;"> **Key Intuition:** </span><br>
> International trade offers firms a flexible, low-risk entry into global markets, and modern nternet technologies have made exporting and importing faster, cheaper, and more efficient.

### <span style="color:#4A86C5;">Licensing</span>

**Licensing**: one firm (the licensor) allows another firm (the licensee) to use its:
- copyrights,
- patents,
- trademarks, or
- trade names  

in exchange for fees or other compensation.

**Key characteristics:**
- The licensor does not establish production facilities abroad.
- No need to transport physical goods across borders.
- Particularly common in software, media, and technology industries.

**Example:**
- Many software producers allow foreign firms to use their software in exchange for licensing fees, generating international revenue without physical foreign operations.

**Key Intuition:** Licensing enables firms to earn foreign income with minimal capital commitment and low risk.

### <span style="color:#4A86C5;">Franchising</span>

**Franchising**: a firm provides:
- a specialized sales or service strategy,
- brand name and operating standards,
- training and support assistance,  
- and sometimesm, an initial investment,  
in exchange for periodic franchise fees.

Under this arrangement:
- Local residents own and manage the individual outlets.
- Franchisees must follow the franchisor’s strict operating standards.

- **Example: McDonald’s model**
    - McDonald’s typically purchases the land and builds the restaurant.
    - It then leases the facility to a franchisee.
    - The franchisee operates the business for a fixed period (e.g., 20 years) while complying with McDonald’s global standards.

**Key Intuition:**  Because franchising by an MNC often involves direct investment in foreign operations, it is classified as a form of **FDI**.

### <span style="color:#4A86C5;">Definition:

Joint ventures allow firms to **enter foreign markets with shared ownership, shared risk, and combined strengths**.

JVs are often used to enter foreign markets where:
  - local knowledge is crucial,
  - regulatory barriers exist,
  - or where shared investment reduces exposure to risk.

Common reasons for forming JVs:

Unlike wholly owned subsidiaries:
  - ownership and control are shared,

## <span style="color:#4A86C5;">4.4. Acquisitions of Existing Operations</span>

---

<div style="
border:2px solid #4A86C5;
border-radius:8px;
padding:16px;
background-color:#F2F6FC;
">

### <span style="color:#4A86C5;">Definition</span>

Firms frequently acquire existing firms in foreign countries as a way to penetrate foreign markets quickly. This is because:
  - It allows the MNC to:
    - enter the market immediately,
    - acquire existing customers, assets, and capabilities, rather than build them from scratch.
    - obtain a large foreign market share in a short time.

**Example:**
  - In 2006, Google acquired several foreign firms to expand its global presence and strengthen its technology base.
  - Google has acquired firms in Australia, Brazil, China (search engines), Canada (mobile browser), Finland (micro-blogging), Germany (mobile software), Russia (online advertising),South Korea (weblog software), Spain (photo sharing),Sweden (videoconferencing).

### <span style="color:#4A86C5;">Full Acquisitions
Full acquisitions → high control, fast market entry, high risk.
International acquisitions involve high risk:
- They require a large upfront investment.
- If foreign operations perform poorly:
  - losses can be substantial,
  - it may be difficult to sell the acquired business at a reasonable price.
- Integration challenges (culture, management, systems) can further reduce value.

### <span style="color:#4A86C5;"> Partial Acquisitions </span>
Partial acquisitions → lower risk, learning opportunity, reduced control.
- Some firms choose to acquire only a partial stake in foreign companies because:
  - They require a smaller investment than full acquisitions.
  - They expose the firm to less financial risk.
  - They allow firms to:
    - establish a toehold in a foreign market,
    - learn about local operations before committing fully.
- Trade-off: Strategic decisions must be shared with other owners.

**Key Intuition:** Because the acquiring firm directly purchases foreign operations, international acquisitions are a form of FDI.

## <span style="color:#4A86C5;">4.5. Establishing New Foreign Subsidiaries (Greenfield Investments)</span>

---

<div style="
border:2px solid #4A86C5;
border-radius:8px;
padding:16px;
background-color:#F2F6FC;
">

### <span style="color:#4A86C5;">Definition</span>
Firms can enter foreign markets by establishing new operations in foreign countries to produce and sell their products.  

This approach is known as a greenfield investment because the firm builds its operations from the ground up on previously undeveloped land.

### <span style="color:#4A86C5;">Why Firms Choose Greenfield Investment</span>

- New subsidiaries can be designed and structured exactly to the firm needs, including:
  - production processes,
  - technology,
  - management systems,
  - corporate culture.
- In some cases, the initial investment may be smaller than the cost of purchasing an existing firm.
- Avoids integration problems often associated with acquisitions (e.g., cultural or managerial conflicts).

### <span style="color:#4A86C5;"> Key Limitations

The firm does not earn immediate returns.

Time is required to:
  - build facilities,
  - hire and train employees,
  - establish supplier relationships,
  - develop a customer base.

Market entry is slower compared with acquisitions.

Greenfield investment offers maximum control and customization, but involves high upfront costs and delayed rewards => most suitable for firms with a long-term strategic perspective.

**Key Intuition:** Greenfield investments require a large FDI.

# 5. Apply a Basic Valuation Framework for MNCs
---

## <span style="color:#4A86C5;">5.1. Valuation Models
---

### <span style="color:#4A86C5;">Purely Domestic Firm 
All CFs are generated in one currency, so there is no ER risk.  
Firm value equals the present value of expected future CFs discounted at the cost of capital.

$$
V = \sum_{t=1}^{n} \frac{E(CF_t)}{(1+k)^t}
$$

Here:
- $E(CF_t)$ is the expected CF in period \(t\)
- $k$ is the firm’s cost of capital

Higher expected CFs increase firm value, while a higher discount rate reduces it.

### <span style="color:#4A86C5;"> A Multinational Corporation
An MNC earns CFs in multiple currencies.  
Before valuation, each foreign-currency CF must be **converted into the home currency** using the expected ER.

For a single period, expected home-currency CF is:

$$
E(CF_t^{H}) = \sum_{j=1}^{m} E(CF_{j,t}) \times E(S_{j,t})
$$

Where:
- $CF_{j,t}$ = CF in currency \(j\) at time \(t\)
- $S_{j,t}$ = ER (home currency per unit of currency \(j\))

Once all CFs are expressed in the home currency, valuation follows the same logic as for a domestic firm:

$$
V = \sum_{t=1}^{n}
\frac{\sum_{j=1}^{m} E(CF_{j,t}) \times E(S_{j,t})}{(1+k)^t}
$$

**Example:**
- Suppose entering Country X adds £20 million of expected annual cash flow. That sounds attractive. 
- But if the cash flows are volatile, exposed to depreciation, and hard to repatriate, investors may raise the discount rate enough that the present-value gain becomes small or even negative.

---

## <span style="color:#4A86C5;"> 5.2. Uncertainties Surrounding an MNC's Cash Flows:

---
MNCs face greater uncertainty than domestic firms because both their CFs and ERs are uncertain, due to **foreign economic risk, political risk, and ER risk, etc**. These uncertainties directly affect firm value.

These risks reduce firm value in two ways:
1. Actual CFs may be lower than expected  
2. Investors require a higher return for bearing additional risk: Higher discount rate $k$ lowers the PV of the MNC.

### <span style="color:#4A86C5;">Case 1: “China+1”?

**The Origins (2000s – 2017)**
China+1 began as a cost hedge in the early 2000s.

* Rising China wages and demographics pushed low-margin assembly toward Southeast Asia (e.g., Vietnam, Thailand).
* Shift was gradual: China stayed the “mother ship” for complex components, while simpler work drifted out.

**The Development (2018 – 2024)**
It became strategic during the U.S.–China tariff cycle and COVID disruptions. “Vietnam-refining” spread:

* Firms kept core China production but added final assembly in Vietnam/India/Mexico.
* Some used minimal value-add steps to claim non-China origin and reduce tariff exposure.

**The 2025 “Default” (Policy Shock)**
Rules tightened as the U.S. broadened tariffs and increased scrutiny of re-routing.

* “Plus One” countries faced more reciprocal pressure.
* Rules-of-origin enforcement strengthened: low local value-added (e.g., mostly Chinese parts + minor assembly) risked being treated as China-origin for tariffs.

**The 2026 Reality**
China+1 didn’t disappear—it became costlier and more structural.

* Final assembly alone is often insufficient; firms must relocate deeper supplier tiers to prove origin.
* Strategy splits: “In China, for China” plus parallel supply chains for the U.S./EU.
* Trade-off: resilience rises, but duplicated networks can lift global prices (often cited ~15–20%).

    * C. Split strategy: “In China, for China” + separate supply chain for U.S./EU
