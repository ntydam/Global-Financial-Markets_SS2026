# 1. The Evolution of the International Monetary System</span>

---
## <span style="color:#4A86C5;">1.1. Definition:</span>

---
The international monetary system (IMS) consists of rules, institutions, and conventions that govern exchange rates (ERs), international payments, and cross-border capital flows.

It is managed by institutions and central banks, with the International Monetary Fund (IMF) oversight and occasional intervention.

U.S. dollar is used as the dominant reserve and transaction currency.

The IMS answers three fundamental questions:
- How is the value of a currency determined?
- Can governments influence that value?
- How are international imbalances resolved?

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
The IMS is the macroeconomic environment within which Global Financial Markets operate.
---

## <span style="color:#4A86C5;">1.2. The Gold Standard (1870s-1930s)</span>
---
- Each country **fixed** its currency to a specific quantity of gold and committed to exchange bank notes for gold on demand. 
- Money supply was therefore constrained by gold reserves and ERs were effectively fixed.

- Implications:
    - Gold convertibility created strong exchange‑rate stability but left little room for independent monetary policy.
    - Domestic economic conditions were subordinated to external balance, and recessions were transmitted across countries.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
Stability was achieved by **sacrificing policy autonomy**.

---

## <span style="color:#4A86C5;">1.3. Bretton Woods System (1945–1973)</span>
---
- After World War II, the global economy was severely disrupted. Countries urgently needed a stable monetary system to rebuild their economies and revive international trade.

- The system designed at Bretton Woods, and later overseen by the IMF, addressed this need by establishing a framework of fixed ERs:
    - Currencies were **pegged** to the U.S. dollar, and the dollar itself was convertible into gold.
    - The IMF supervised the system and provided temporary financing to countries facing BOP deficits.
    - This arrangement significantly reduced exchange‑rate uncertainty and facilitated post‑war reconstruction. Capital controls limited destabilising flows and the U.S. dollar served as a credible anchor.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
Stability was prioritized to rebuild trust in the global economy.

---

### <span style="color:#5b9bd5;">The Role of the U.S. Dollar</span>
At the core of Bretton Woods was the U.S. dollar (a reserve currency).

As global trade and investment expanded:
- Foreign central banks, firms, and investors demanded more dollars for:
    - reserves, 
    - transactions, and 
    - financial markets.

To supply these dollars, the U.S. ran persistent **BOP deficits**, sending large quantities of dollars abroad through capital outflows.

Eventually, the amount of dollars held outside the U.S. far exceeded the country’s ability to back them with gold. Confidence in the dollar’s convertibility eroded, and the system became unsustainable.

### <span style="color:#5b9bd5;">The System's Breakdown</span>
Over time, countries pursued divergent fiscal and monetary policies. Inflation rates differed and repeated economic shocks strained fixed rates. Maintaining pegs required high domestic costs, including recessions and interest‑rate hikes. 

By the early 1970s the system collapsed; currencies began to float and the world moved to more flexible regimes.

---

## <span style="color:#4A86C5;">1.4. Modern Floating System (1973–Present)</span>

---

### <span style="color:#5b9bd5;"> Floating Exchange Rates

After the collapse of the Bretton Woods system, many major currencies began to float, allowing market forces to determine ERs.
- Major currencies (e.g., USD, EUR, JPY, and GBP) operate under floating ER regimes.
- Floating ERs provide **greater monetary policy autonomy**, allowing countries to respond to domestic economic conditions.

### <span style="color:#5b9bd5;"> Managed Floats and Pegged Exchange Rates

Many countries do not allow their currencies to float freely and instead adopt hybrid ER regimes to balance ER stability with policy flexibility.
- Common arrangements include managed floats and pegged regimes (e.g., CNY, SGD, MYR, INR).
- In practice, de facto ER behavior often differs from de jure (officially stated) regimes.
- IMF evidence shows that while a majority of countries classify their regimes as floating, only a minority allow their currencies to float freely without regular intervention.

Example — **Managed Floats in Practice:**
- **Vietnam** manages the VND through *discretionary intervention*, reflecting priorities of export competitiveness and limits excessive volatility.
    - Each business day, the State Bank of Vietnam (SBV) announces a central reference ER for USD/VND. The VND is allowed to trade within a ±5% band around this reference rate.
- **Singapore** manages the SGD using a *transparent, rules-based exchange-rate framework*, reflecting its extreme trade openness, deep financial markets, and fully open capital account.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Even countries that claim to float often intervene actively in foreign exchange markets.

---


## <span style="color:#4A86C5;">1.5. Policy Trilemma</span>

---

ER regimes differ by **how much flexibility is allowed** and **how actively authorities intervene.**

| **ER Regime**          | **Key Feature**                                                         | **Degree of Flexibility** | **Central Bank Role**                         | **Typical Examples**                            |
| --------------------------------- | ------------------------------------------------------------------------------------ | ------------------------- | --------------------------------------------- | ----------------------------------------------- |
| **Fixed** ER    | Currency value is fixed to a commodity or another currency at a predetermined rate   | None                      | Full commitment to maintain the fixed rate    | Gold Standard (1879–1934)                       |
| **Pegged** ER   | Currency is pegged to a major currency (often USD) at a fixed rate                   | Very low                  | Intervenes to maintain the peg                | Bretton Woods (1944–1970s). DKK, HKD |
| **Managed** ER  | Currency allowed to move within a band or managed relative to a basket of currencies | Medium                    | Active intervention to smooth fluctuations    | CNY, SGD, MYR, VND             |
| **Floating** ER | ER determined by market demand and supply                                 | High                      | Limited intervention (“dirty float” possible) | USD, GBP, JPY, AUD                              |

**Trilemma**: 
Objectives of an ER regime:
- (1) exchange‑rate stability, 
- (2) monetary independence and 
- (3) full capital mobility

A country cannot simultaneously achieve all three, only two of the three objectives can be attained:

- Fixed rate + capital mobility ⇒ monetary independence is impossible because interest rates must align with those of the anchor currency.

- Monetary independence + capital mobility ⇒ ER must float. This is the approach of the United States, the eurozone and Japan.

- Fixed rate + monetary independence ⇒ capital controls are needed to prevent outflows/inflows that would undermine the peg.

# 2. Balance of Payments</span>

---
## <span style="color:#4A86C5;">2.1. Definition:</span>

---

The BOP is a record of all economic transactions between a country’s residents (households, firms, and government) and the rest of the world over a specified period.

These transactions include:
- Trade in goods and services (exports and imports),
- Financial flows, such as investment and borrowing,
- Transfer payments, including foreign aid and remittances.

For economies engaged in international trade and finance, the BOP is a **key macroeconomic indicator** because it summarizes how funds move into and out of the country through trade, investment, and cross-border transfers.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> The BOP records flows of funds over time, not a country’s stock of wealth.

---


## <span style="color:#4A86C5;">2.2. Main Components of the BOP</span>
---

The BOP records all economic transactions into three main accounts: 
- the Current Account (CA),
- the Capital Account (KA), and 
- the Financial Account (FA).

---
### <span style="color:#5b9bd5;"> Current Account:
The CA measures flows of goods, services, income, and transfers between a country and non-residents.

Sub-components:
- Goods: Exports(X) and imports (M) of physical products (e.g., machinery, food, electronics).
- Services: X and M of services (e.g., tourism, banking, transport).
- Primary Income: Earnings from foreign investments (interest, dividends) minus payments to foreign investors.
- Secondary Income: One-way transfers such as remittances sent by migrants or foreign aid.

Example:
- A country exporting US$500 billion worth of goods and importing US$450 billion would have a goods trade surplus of US$50 billion.
- Net income from FDI is US$30 billion and net transfers are −US$10 billion.
- Its CA Balance would be:
$$
\begin{aligned}
\text{Goods balance}      &\quad +50 \\
\text{Net income}         &\quad +30 \\
\text{Net transfers}      &\quad -10 \\
\hline
\textbf{Current account balance} &\quad \textbf{+70 (surplus)}
\end{aligned}
$$


><span style="color:#4A86C5;">**Key Intuition:**</span><br>
> A country runs a CA surplus when earnings from trade and FDI exceed payments to the rest of the world.
---

### <span style="color:#5b9bd5;"> Capital Account:
The KA records capital transfers and transactions in non-produced, non-financial assets. These include:
- Debt forgiveness or cancellation,
- Transfers of ownership of physical or intangible assets (e.g., land rights, patents).

Compared with FA or CA, the KA is usually **smaller in volume**.

---

### <span style="color:#5b9bd5;"> Financial Account
The FA captures transactions in financial assets and liabilities between residents and non-residents.

Key categories:
- Direct investment: Cross-border acquisition of lasting interests (e.g., a MNC building a factory abroad).
- Portfolio investment: Cross-border purchases of stocks and bonds.
- Other investment: Loans, deposits, and banking flows.
- Reserve assets: Changes in official reserve holdings held by the central bank.

Example:

$$
\begin{aligned}
\text{FDI inflow}           &\quad +100 \\
\text{Net portfolio inflow} &\quad +50 \\
\hline
\textbf{Financial account balance} &\quad \textbf{+150 (surplus)}
\end{aligned}
$$

---

## <span style="color:#4A86C5;">2.3. Balance of Payments in Theory </span>
---

In theory, the sum of all transactions recorded in CA, KA, and FA should **balance to zero** when accounting and statistical adjustments are included.

$$
\boxed{
\text{Current Account}
+ \text{Capital Account}
+ \text{Financial Account}
= 0
}
$$


If a country has a CA deficit (e.g., when M > X), it must finance that through:
- KA/FA surplus (e.g., foreign portfolio and direct investment), or
- Drawing down reserves.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> In BOP accounting, **CA** and **(FA + KA)** move in opposite directions.
> However, balancing does **not** imply economic health.

---

### <span style="color:#5b9bd5;"> Example: Interpreting a Current Account Surplus in the Balance of Payments

In the BOP, the CA measures a country’s net earnings from trade, income, and transfers with the rest of the world. It is defined as:

$$
\boxed{
\text{CA} = (X_G - M_G) + (X_S - M_S) + \text{Net Primary Income} + \text{Net Transfers}
}
$$

### <span style="color:#5b9bd5;"> Application: India’s Current Account Surplus

According to [Reuters](https://www.reuters.com/world/india/india-posts-first-current-account-surplus-four-quarters-central-bank-says-2025-06-27/?utm_source=chatgpt.com), India recorded a CA surplus of US$13.5 bn, supported by strong services exports and large remittance inflows.

This outcome can be expressed as:

$$
\begin{aligned}
\text{CA}_{\text{India}} 
&= \underbrace{(X_G - M_G)}_{\text{Goods balance (negative)}} \\
&\quad + \underbrace{(X_S - M_S)}_{+\;53.3\text{ bn (services surplus)}} \\
&\quad + \underbrace{\text{Net Transfers}}_{+\;123\text{ bn (remittances)}} \\
&= +13.5 \text{ bn}
\end{aligned}
$$

Although India runs a goods trade deficit, the services surplus (e.g. IT and business services) and remittance inflows more than offset this deficit, resulting in a CA surplus.

The BOP must satisfy:

$$
\text{Current Account} + \text{Capital Account} + \text{Financial Account} = 0
$$

Therefore, India’s CA surplus implies:

$$
\text{Financial Account} = -13.5 \text{ bn}
$$

This means India is a **net lender to the rest of the world**, or equivalently, that foreign exchange reserves increase or net capital outflows occur.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> A country does not need a goods trade surplus to run a CA surplus; strong services exports and remittances can dominate the external balance.

---

## <span style="color:#4A86C5;">2.4. Balance of Payments in Practice</span>
---

In practice, BOP data rarely sum to zero exactly. Therefore, two additional components are included to ensure  the BOP identity holds in accounting terms.

- **Errors and Omissions**
  - Capture statistical discrepancies and measurement errors
  - Reflect timing differences, under-reporting, or misclassification

- **Official Reserves**
  - Central bank interventions (e.g., buying/selling foreign currency)
  - Used to help reconcile observed imbalances

$$
\boxed{\text{Current Account} + \text{Capital Account} + \text{Financial Account} + \text{Errors \& Omissions} + \text{Official Reserves}= 0}
$$

### <span style="color:#5b9bd5;"> Example: ***Negative*** Net FDI

Consider a country that engages heavily in outward FDI:

- If the country’s FDI in foreign countries exceeds FDI into the country, then:
  - More funds are leaving the country than entering.
  - This produces a negative entry (debit)  in the FA.

$$
\text{Net FDI} = \text{Outward FDI} - \text{Inward FDI} < 0
$$

Although outward FDI worsens the FA, it generates primary income receipts (profits, dividends, interest) in the future:

- These income inflows are recorded in the CA.
- As a result, higher outward FDI tends to increase the CA balance over time.

$$
\text{Primary Income}_{\text{received}} > \text{Primary Income}_{\text{paid}} 
\;\Rightarrow\; \text{CA improves}
$$

Thus:
- FA: short-run capital outflow.
- CA: long-run income inflow.

## <span style="color:#4A86C5;">3.1. The Expansion of Global Trade</span>
---

- Trade openness (total trade as % of GDP) has increased significantly for most countries over the past several decades.

```python
# Compare Trade (% of GDP) across up to 4 countries/regions (incl. aggregates) via World Bank WDI API

import requests
import pandas as pd
import matplotlib.pyplot as plt

import ipywidgets as widgets
from IPython.display import display, clear_output

WB_API = "https://api.worldbank.org/v2"
INDICATOR = "NE.TRD.GNFS.ZS"  # Trade (% of GDP)

# -----------------------------
# Helpers: World Bank API calls
# -----------------------------
def _get_json(url: str, params: dict | None = None, timeout: int = 30):
    params = {} if params is None else dict(params)
    params.setdefault("format", "json")
    r = requests.get(url, params=params, timeout=timeout)
    r.raise_for_status()
    return r.json()

def fetch_all_countries_and_aggregates() -> pd.DataFrame:
    """
    Returns a DataFrame of available 'countries' plus WDI 'aggregates' from /country endpoint.
    Includes: id (3-letter), iso2Code (2-letter), name, region, incomeLevel, lendingType
    """
    # The API is paginated; we walk all pages. (Typical total ~296 entries.)
    # per_page can be tuned; keep moderate to avoid server issues.
    per_page = 200
    page = 1
    rows = []

    while True:
        data = _get_json(f"{WB_API}/country", params={"per_page": per_page, "page": page})
        meta, items = data[0], data[1]
        rows.extend(items)
        if int(meta["page"]) >= int(meta["pages"]):
            break
        page += 1

    df = pd.json_normalize(rows)
    # Normalize column names we care about
    keep = {
        "id": "id",
        "iso2Code": "iso2",
        "name": "name",
        "region.value": "region",
        "incomeLevel.value": "income_level",
        "lendingType.value": "lending_type",
    }
    df = df[list(keep.keys())].rename(columns=keep)

    # A few "countries" returned can be blanks/odd; filter defensively.
    df = df[df["id"].notna() & (df["id"].str.len() == 3)]
    df = df.drop_duplicates(subset=["id"]).reset_index(drop=True)

    # Build a friendly label: "Name — Region / Income"
    df["label"] = df["name"] + " — " + df["region"].fillna("Unknown") + " / " + df["income_level"].fillna("")

    return df

def fetch_indicator_series(entity_id_3: str, indicator: str = INDICATOR) -> pd.DataFrame:
    """
    Fetch time series for a given entity (3-letter ID used by WB API, e.g., USA, VNM, WLD, HIC, etc.)
    Returns DataFrame with columns: year (int), value (float)
    """
    url = f"{WB_API}/country/{entity_id_3}/indicator/{indicator}"
    data = _get_json(url, params={"per_page": 20000})  # enough for long series

    if not isinstance(data, list) or len(data) < 2 or data[1] is None:
        return pd.DataFrame(columns=["year", "value"])

    items = data[1]
    out = []
    for it in items:
        date = it.get("date")
        val = it.get("value")
        if date is None:
            continue
        try:
            year = int(date)
        except ValueError:
            continue
        out.append((year, val))

    df = pd.DataFrame(out, columns=["year", "value"])
    df = df.dropna(subset=["value"])
    df["value"] = pd.to_numeric(df["value"], errors="coerce")
    df = df.dropna(subset=["value"]).sort_values("year").reset_index(drop=True)
    return df

# -----------------------------
# UI: 4 dropdown selectors
# -----------------------------
countries_df = fetch_all_countries_and_aggregates()

# Map label -> id
label_to_id = dict(zip(countries_df["label"], countries_df["id"]))

# Provide a searchable-ish experience by sorting labels
labels_sorted = sorted(label_to_id.keys())

def make_selector(title: str):
    return widgets.Dropdown(
        options=[("— None —", "")] + [(lbl, lbl) for lbl in labels_sorted],
        value="",
        description=title,
        layout=widgets.Layout(width="98%"),
        style={"description_width": "90px"},
    )

sel1 = make_selector("Entity 1")
sel2 = make_selector("Entity 2")
sel3 = make_selector("Entity 3")
sel4 = make_selector("Entity 4")

# Year range (auto-adjust after fetch; initialize broadly)
year_range = widgets.IntRangeSlider(
    value=[1990, 2024],
    min=1960,
    max=2024,
    step=1,
    description="Years",
    continuous_update=False,
    layout=widgets.Layout(width="98%"),
    style={"description_width": "90px"},
)

btn = widgets.Button(description="Load & Compare", button_style="primary")
out = widgets.Output()

def _selected_ids():
    chosen = []
    for s in (sel1, sel2, sel3, sel4):
        lbl = s.value
        if lbl:
            chosen.append(label_to_id[lbl])
    # de-duplicate while keeping order
    seen = set()
    uniq = []
    for x in chosen:
        if x not in seen:
            uniq.append(x); seen.add(x)
    return uniq

def _selected_names(ids):
    id_to_name = dict(zip(countries_df["id"], countries_df["name"]))
    return [id_to_name.get(i, i) for i in ids]

def _pull_compare(_):
    with out:
        clear_output(wait=True)

        ids = _selected_ids()
        if len(ids) == 0:
            print("Select at least 1 entity.")
            return

        names = _selected_names(ids)

        # Fetch series for each entity
        series = {}
        all_years = set()

        for entity_id, nm in zip(ids, names):
            df = fetch_indicator_series(entity_id)
            if df.empty:
                print(f"Warning: no data for {nm} ({entity_id}).")
                continue
            series[nm] = df.set_index("year")["value"]
            all_years.update(df["year"].tolist())

        if not series:
            print("No data returned for the selected entities.")
            return

        # Build comparison frame
        comp = pd.DataFrame(series).sort_index()

        # Adjust year slider to available years (optional)
        min_y, max_y = int(min(all_years)), int(max(all_years))
        year_range.min = min_y
        year_range.max = max_y
        if year_range.value[0] < min_y or year_range.value[1] > max_y:
            year_range.value = [max(min_y, year_range.value[0]), min(max_y, year_range.value[1])]

        y0, y1 = year_range.value
        comp = comp.loc[(comp.index >= y0) & (comp.index <= y1)]

        if comp.empty:
            print(f"No data in the selected year range: {y0}-{y1}.")
            return

        # Plot
        ax = comp.plot(figsize=(10, 5))
        ax.set_title("Trade (% of GDP) — World Bank WDI (NE.TRD.GNFS.ZS)")
        ax.set_xlabel("Year")
        ax.set_ylabel("Trade (% of GDP)")
        ax.grid(True, which="major", axis="both")
        plt.show()

        # Table (latest year in range)
        latest_year = int(comp.dropna(how="all").index.max())
        latest = comp.loc[latest_year].sort_values(ascending=False)
        display(pd.DataFrame({"latest_year": latest_year, "trade_%_gdp": latest}).reset_index().rename(columns={"index": "entity"}))

btn.on_click(_pull_compare)

ui = widgets.VBox([
    widgets.HTML("<b>Compare up to 4 entities (countries, regions, income groups, world aggregates)</b>"),
    sel1, sel2, sel3, sel4,
    year_range,
    btn,
    out
])

display(ui)
```

## <span style="color:#4A86C5;">3.2. What International Trade Means to the U.S.</span>

---

### <span style="color:#5b9bd5;"> How the U.S. Benefited from International Trade</span>
International trade raises economy-wide welfare through multiple channels, but also contribute to persistent CA deficits—especially in large, open economies like the United States.

- **Job creation in high-value industries**  
  Trade supports U.S. jobs in sectors where domestic firms have a technology or productivity advantage.

- **Efficiency gains through specialization**  
  Production shifts toward countries that can produce goods more efficiently, lowering overall global costs.

- **Greater competition and lower prices**  
  Increased foreign competition forces firms to innovate and keep prices low.

- **More choice for consumers**  
  U.S. consumers enjoy greater product variety at lower prices as a result of trade openness.

> <span style="color:#5b9bd5;"> **Key Intuition:** </span><br>
> Trade improves consumer welfare and firm efficiency, even though it redistributes production across countries.


---

### <span style="color:#5b9bd5;"> Why International Trade Generates a Current Account Deficit

The CA is defined as:

$$
\boxed{\text{CA} = (X - M) + \text{Net Income} + \text{Net Transfers}}
$$

[For the United States:](https://www.statista.com/statistics/218255/total-value-of-us-trade-in-goods-worldwide-since-2004/?srsltid=AfmBOoreVC4sJaUJPjL2d6e3MRygLU4OXY2_ju-qhPi1lmwxMez66Xul)
- Strong consumption demand + global supply chains → imports grow faster than exports
- Even if exports rise, imports often rise more

</center>


**Result**: The U.S. runs a [**persistent CA deficit**](https://www.statista.com/statistics/215552/balance-of-trade-of-goods-and-services-of-the-us-by-quarter/) because:
$$
X - M < 0 \;\Rightarrow\; \text{CA Deficit}
$$

This is not policy failure — it reflects how the U.S. participates in global trade.

---

### <span style="color:#5b9bd5;"> Financing the Deficit: Capital Inflows

By balance-of-payments identity:

$$
\text{Current Account} + \text{Financial Account} = 0
$$

A U.S. current account deficit therefore implies:

$$
\text{Financial Account Surplus} > 0
$$

Meaning:
- Foreign capital flows into U.S. assets (Treasuries, equities, FDI)
- The U.S. effectively **imports goods and exports financial claims**

> **Interpretation:** The U.S. trades goods for capital.

---

### <span style="color:#5b9bd5;"> Current Account Deficit: Under Trump's View
Under Donald Trump, trade deficit is treated as:
- A bilateral problem (e.g., China vs U.S.)
- A policy failure, not a macroeconomic outcome

Tariffs were justified using a partial-equilibrium Key Intuition: **“If imports are too high, tax them → imports fall → trade deficit shrinks.”**

In April 2025, the U.S. imposed tariffs to reduce imports and shrink the trade deficit.

**What happened in practice:**
- Imports shifted across countries, not eliminated
- Consumer prices increased
- The overall U.S. CA deficit did not disappear

**Why?**
- The deficit is driven by macroeconomic fundamentals:
  - High domestic consumption
  - Low national saving
  - Strong demand for U.S. financial assets

> <span style="color:#5b9bd5;"> **Key Intuition:** </span><br>
> A persistent U.S. CA deficit is the logical outcome of trade growth, consumption patterns, and capital inflows — not evidence that trade policy has “failed.”

This insight is essential for understanding:
- Outsourcing debates  
- Trade wars  
- Capital mobility  
- Financial fragility and sudden stops

## <span style="color:#4A86C5;">3.3. Events That Increased Global Trade Volume</span>
---

The long-run growth in international trade reflects policy decisions and institutional changes that reduced cross-border barriers.

The most important events that accelerated global trade are:

---

### <span style="color:#5b9bd5;"> Fall of the Berlin Wall (1989)

The collapse of the Berlin Wall marked the transition of Eastern Europe toward market-oriented economies.

**Economic Implications:**
- Reduced trade barriers between Eastern and Western Europe  
- Privatization of state-owned firms
- Entry of multinational corporations into Eastern Europe
- Access to low-cost labor and inputs for global firms

> This event dramatically expanded the set of countries participating in global trade.
---

### <span style="color:#5b9bd5;"> Single European Act (1987)

The Single European Act aimed to create a unified European market by:
- harmonizing regulations and 
- removing trade barriers among the European countries.

**Economic implications:**
- Easier access to suppliers across European countries
- Greater access to markets for firms
- Ability to exploit economies of scale through standardized production

Many firms (including U.S. MNCs operating in Europe, such as *Best Foods (now Unilever)*) streamlined manufacturing and distribution to serve multiple countries more efficiently.

### <span style="color:#5b9bd5;"> NAFTA (1993)

The North American Free Trade Agreement (NAFTA) eliminated most trade barriers between the United States, Mexico, and Canada.

**Economic Implications:**
- U.S. firms gained access to Mexican product and labor markets (e.g., U.S. production relocated to Mexico)
- Expansion of cross-border supply chains
- Increased competition for U.S. producers in labor-intensive industries (e.g., U.S. firms lost their market shares to Mexico apparel firms).

> Trade liberalization benefits consumers and firms overall, but can create winners and losers across industries.


### <span style="color:#5b9bd5;"> GATT and the Uruguay Round

The General Agreement on Tariffs and Trade (GATT) culminated in the Uruguay Round within the same month of the NAFTA accord, covering 117 countries.

**Economic Implications::**
- Reduction of tariffs on specified imported goods over 10 years
- Greater access to previously protected foreign markets
- Strong expansion in global trade volumes over the following decade

### <span style="color:#5b9bd5;"> The EU

The European Union (EU) is a supranational economic and political unio* consisting of [27 member countries](https://european-union.europa.eu/principles-countries-history/eu-countries_en?page=1), with a central economic objective is to create a **single integrated market** that allows resources to move freely across national borders.

The EU is built on the four fundamental freedoms, which distinguish it from standard free-trade agreements:
- Free movement of goods – elimination of tariffs, quotas, and most customs controls  
- Free movement of services – firms can provide services across member states  
- Free movement of capital – unrestricted cross-border investment and financial flows  
- Free movement of labor – workers can live and work in any member country  

With a population of ~500 million people and ~one-quarter of global GDP, the EU represents one of the largest the world’s trading blocs and investment destinations.

### <span style="color:#5b9bd5;"> Inception of the Euro (2002)

The euro was introduced as the official currency of the Eurozone, which initially comprised 12 of the then 15 EU member states. Since then, several more countries have adopted the euro, bringing the total to 20 as of 2024.

The adoption of the euro aims to:
- Eliminate ER risk among member countries  
- Reduce transaction costs in cross-border trade and investment  
- Increase price transparency, intensifying competition  
- Deepen financial market integration

This encourages:
- Higher intra-Eurozone trade volumes
- Longer and more complex cross-border supply chains
- Empirical evidence shows that a common currency tends to increase bilateral trade among members, beyond what free trade alone would achieve.

However, not all EU members use the euro. Countries that adopt it **give up** their **national currencies and independent monetary policy** in exchange for deeper financial and trade integration.

While the euro delivers efficiency gains, it imposes important trade-offs:
- Loss of monetary sovereignty
- Limited ability to respond to asymmetric shocks
- Fiscal policy is restricted by EU rules, yet fiscal transfers across countries remain limited

## <span style="color:#4A86C5;">3.3. Impact of Outsourcing on Trade</span>
---
### <span style="color:#5b9bd5;"> Definition

Outsourcing refers to subcontracting production or services to a third party in another country.

- Firms import intermediate goods or services instead of producing them internally.
- Cross-border transactions increase by construction.

Firms outsource for several reasons:
- Lower labor and operating costs
- Ability to remain competitive globally
- Access to specialized skills or technologies

However, outsourcing is controversial:
- Creates jobs in low-wage countries
- May reduce domestic employment in labor-intensive sectors
- Raises ethical, political, and social concerns

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Outsourcing improves efficiency but redistributes income and employment.

---

### <span style="color:#5b9bd5;"> Managerial Decisions about Outsourcing</span>

Managers face a trade-off between:

- Cost minimization and competitiveness
- Domestic employment, reputation, and worker morale

Shareholders may pressure managers to outsource if:
- Identical products can be produced abroad at a fraction of the cost
- Outsourcing increases firm value

A common compromise:
- Keep existing domestic jobs
- Use outsourcing only to accommodate future growth

**Key linkage:**

## <span style="color:#4A86C5;">3.4. Factors Driving International Trade

---

Because international trade strongly affects output, employment, and the BOP, managers and policymakers closely monitor the main drivers of trade flows:
- Cost of labor
- Inflation
- National income
- Credit conditions
- Government policies
- ERs

These factors shift relative prices, purchasing power, and financing capacity, which changes X and M.

---

### <span style="color:#5b9bd5;">Cost of Labor</span>

Differences in labor costs across countries play a major role in shaping international trade patterns.

When a country has lower labor costs, firms operating in that country can produce goods at a lower unit cost, particularly in industries that rely heavily on manual labor. This increases foreign demand and leads to higher exports.

As a result, labor-intensive industries such as apparel, footwear, basic assembly, furniture, toys, and low-end electronics assembly are typically concentrated in low-income or emerging economies.

The advantage of low labor costs is strongest when:
- production is labor-intensive, and  
- technology or capital requirements are not the primary drivers of competitiveness.

---

### <span style="color:#5b9bd5;">Inflation</span>
If a country’s inflation rises faster than its trading partners:
- Domestic goods become relatively more expensive
- Foreign buyers switch away → exports fall
- Domestic buyers switch to cheaper imports → imports rise
- Net exports fall → CA tends to deteriorate

Example:
- If Malaysia inflation rises while competitors’ inflation is stable, Malaysian export prices may increase in USD terms (unless offset by FX moves), reducing competitiveness.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Higher relative inflation usually creates competitiveness loss.

---

### <span style="color:#5b9bd5;">National Income</span>
When a country’s real income grows faster than others:
- Consumption increases
- Part of extra demand falls on foreign goods → **imports rise**
- Holding other factors constant → **CA tends to decrease**

Example
- During expansions, consumers buy more imported items (electronics, cars, luxury goods).
- Firms expand production and may import more intermediate inputs.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Strong domestic income growth often raises imports faster than exports in the short run.

---

### <span style="color:#5b9bd5;">Credit Conditions</span>
Trade often requires credit:
- Importers may use letters of credit (bank guarantees to pay upon delivery).
- If banks tighten credit (recession / higher default risk):
  - Firms struggle to finance imports
  - Corporate spending falls
  - Demand for imported inputs declines
  - Trade flows contract

Example (realistic business constraint)
A manufacturer that normally imports components using a letter of credit may:
- delay orders, reduce production, or find domestic substitutes
when the bank refuses or raises fees.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Tighter credit reduces both trade finance and real trade volumes.

---

### <span style="color:#5b9bd5;">Government Policies</span>

Governments intervene when trade affects jobs, growth, or national priorities.

Restrictions on imports
- Tariffs: tax on imports → higher consumer prices → imports fall
- Quotas: quantity limit → imports fall

**Example:** Tariffs on imported cars can encourage foreign producers to build local plants (FDI) to avoid tariffs.

#### Export promotion
- Subsidies / tax breaks lower firms’ costs → exports become more competitive
- If goods are sold abroad at artificially low prices due to support, it may be viewed as **dumping**

#### Other trade-affecting regulations
- Anti-piracy enforcement: weak enforcement reduces demand for legitimate imports
- Environmental and labor laws: stricter rules raise costs → may reduce competitiveness
- Administrative requirements / bureaucracy: licensing delays can act like hidden trade barriers
- Sanctions / security laws: restrict trade for geopolitical or national security reasons

#### Strategic insight
Trade policy often triggers **retaliation**, so “protecting jobs” in one sector may harm others.

> <span style="color:#4A86C5;">**Key Intuition:**</span><br>
> Government policy reshapes trade by changing prices, access, and competitiveness.

---

### <span style="color:#5b9bd5;">Exchange Rates</span>
If the home currency appreciates:
- exports become more expensive abroad → exports fall
- imports become cheaper domestically → imports rise
- trade balance tends to worsen

If the home currency depreciates:
- exports become cheaper abroad → exports rise
- imports become more expensive → imports fall
- trade balance tends to improve (over time)

**Example:**
If a product is priced at 100 EUR:
- At $1.60/EUR, price to US buyers = $160 → demand lower
- At $1.20/EUR, price to US buyers = $120 → demand higher

#### Why ERs may not “fix” trade deficits
- Offsetting capital inflows can keep the currency strong despite trade deficits
- Time lags (the J-curve effect):
    - After a currency depreciates, imports become more expensive in domestic currency.
    - In the short run, import quantities do not fall immediately, so the value of imports rises.
    - Export volumes take time to increase (contracts, habits and production plans are fixed).
    - As a result, imports exceed exports, so the trade balance deteriorates (moves further into deficit).

# 4. Capital Flows, Fragility, and Sudden Stops</span>

---
## <span style="color:#4A86C5;">4.1. Capital Flows vs. Trade Flows

Although both capital flows and trade flows involve cross-border transactions, they differ fundamentally in their nature, speed, and economic impact.

**Trade flows** arise from the exchange of goods and services across countries. They are driven primarily by:
- comparative advantage, 
- production costs, consumer demand, and 
- trade policies.
Because trade relationships depend on supply chains, contracts, and consumer habits, trade flows tend to adjust relatively slowly over time.

**Capital flows**, by contrast, reflect cross-border movements of financial capital, including:
- foreign direct investment (FDI), 
- portfolio investment, and 
- bank lending. 

These flows are driven by:
- expected returns, 
- IR differentials,
- ER expectations, and 
- global risk appetite. 

---


## <span style="color:#4A86C5;">4.2. Capital Flows Dominate Trade Flows</span>

In modern financial markets, capital flows far exceed international trade flows.

Unlike trade relationships, portfolio flows and FDI can move quickly and reverse fast, making capital-reliant countries more vulnerable to external shocks. According to [Statista (2025)](https://www.statista.com/topics/11902/foreign-direct-investment-fdi-worldwide/#topicOverview):
- 2025 Value of global FDI inflows = 1.3tn USD 
- Global rate of return on inward FDI = 6%

Because capital flows are highly mobile and sensitive to global financial conditions, countries that rely heavily on foreign capital become more exposed to external shocks that triggers large inflows and outflows, such as:
- Changes in interest rates,
- risk appetite, and
- global uncertainty.

---

## <span style="color:#4A86C5;">4.3. Sudden Stops and Financial Fragility</span>

When capital is highly mobile, foreign investors can shift funds in and out quickly and cheaply. 
    - In favorable conditions, this can fuel credit booms and rapid growth 
    - In unfavorable conditions (such as when risk sentiment turns or financing tightens, draining liquidity, straining banks, and triggering sharp downturns), this can reverse abruptly.

A **sudden stop** is an abrupt collapse or reversal of foreign inflows, typically accompanied by:
* sharp currency depreciation,
* falling asset prices, and
* stress in refinancing for firms and banks.

Sudden stops are more likely when capital mobility is high, ERs are tightly managed, and domestic financial markets are shallow—so foreign funding cannot be easily replaced. These dynamics can amplify **financial fragility** and worsen crisis risk.

## <span style="color:#4A86C5;">4.4. Implications for MNCs and Investors</span>

For MNCs:
- Volatile capital flows translate into financial risk. 
- ER movements affect the value of revenues, costs, and cash flows across countries. 

In financial stress periods, governments may impose **capital controls**, which can restrict the repatriation of profits or the movement of funds across borders.

Investment decisions by MNCs are therefore shaped not only by expected returns, but also **by policy regimes, ER stability, and access to financing.**

Similarly, international investors must consider how changes in interest rates, taxes, and currency expectations affect the risk-adjusted returns on foreign assets.

### <span style="color:#4A86C5;">Case 1: U.S. vs China: ERs, Trade Balances, and Policy Disputes
For years, U.S. exporters and policymakers have argued that China keeps the yuan artificially low against the dollar, making Chinese exports cheaper and U.S. goods less competitive—contributing to the U.S. trade deficit with China. The issue drew especially strong political attention in 2008–2012, as the U.S. sought growth and jobs after the global financial crisis.

Supporters claim that a stronger yuan would boost U.S. export competitiveness, reduce imports from China, and raise U.S. output and employment—making exchange-rate adjustment a tool to shrink trade imbalances.

China counters that the imbalance is largely structural, driven by major labor-cost differences, so currency revaluation alone may not erase China’s cost advantage. Even if Chinese goods became pricier, U.S. consumers might shift to other low-wage suppliers (e.g., Vietnam, Mexico, Malaysia) rather than buy U.S.-made products.

This suggests exchange-rate pressure may simply reallocate imports across low-cost countries without materially reducing the overall U.S. trade deficit—raising the question of whether the issue reflects deeper forces like global value chains and comparative advantage rather than currency levels alone.

### <span style="color:#4A86C5;">Case 2: Asia 1997 / EM sudden stop
The Asian Financial Crisis provides a clear illustration of how sudden stops can trigger severe financial instability. In the early 1990s, several East Asian economies—including Thailand, Indonesia, and South Korea—experienced large capital inflows driven by high growth prospects, relatively stable ERs, and liberalized capital accounts. These inflows fueled rapid credit expansion, asset price booms, and increased borrowing by banks and corporations, often in foreign currencies.

Most of these economies maintained fixed or tightly managed ER regimes, which encouraged foreign borrowing by creating the perception of low ER risk. At the same time, domestic financial markets were shallow, and banking regulation was weak, leaving financial systems highly exposed to shifts in investor sentiment.

In 1997, confidence in the region began to deteriorate as concerns emerged about rising debt levels, weak bank balance sheets, and declining export competitiveness. Foreign investors responded by withdrawing capital abruptly. As capital inflows reversed, central banks were unable to defend their ER pegs, leading to sharp currency depreciations.

The sudden stop triggered a collapse in asset prices, widespread bank failures, and severe recessions across the region. Firms that had borrowed heavily in foreign currency faced soaring debt burdens, while banks struggled to refinance short-term foreign liabilities. What began as a reversal of capital flows quickly evolved into a full-scale financial and economic crisis.

This episode highlights how high capital mobility, rigid ER regimes, and shallow domestic financial systems can interact to amplify financial fragility, making sudden stops both more likely and more damaging.
