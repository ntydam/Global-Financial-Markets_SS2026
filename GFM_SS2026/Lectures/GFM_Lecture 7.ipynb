# <span style="color:#4A86C5;">1. Revisiting FX Exposures: An Integrated Framework</span>
---

Recall from **Lecture 6** that exchange-rate changes affect MNCs through three distinct channels. Lecture 7 moves beyond definitions to **measurement and management** of each.

| Exposure | Nature | Horizon | Cash-flow impact? | Covered in Lecture 6? |
|---|---|---|---|---|
| **Transaction** | Contracted FC cash flows | Short-term (days–months) | Yes — direct | Foundations ✓ |
| **Economic** | Future operating CFs & competitiveness | Medium to long-term | Yes — indirect | Introduced only |
| **Translation** | Consolidation of foreign subsidiary FS | Reporting period | No direct CF | Introduced only |

### <span style="color:#4A86C5;">Why do all three matter together?</span>

A single event — say, a 15% depreciation of the Brazilian real — can simultaneously:
- Reduce the USD value of receivables from Brazilian customers (**transaction**)
- Lower the NPV of expected future cash flows from the Brazilian market (**economic**)
- Shrink the reported equity of the Brazilian subsidiary (**translation**)

> <span style="color:#4A86C5;">**Key insight:**</span> Effective FX risk management requires an **integrated approach** — not just hedging individual exposures in isolation.

# <span style="color:#4A86C5;">2. Economic (Operating) Exposure</span>
---

## <span style="color:#4A86C5;">2.1. Definition and Significance</span>
---
- Economic exposure measures how **unexpected** exchange-rate changes affect the **present value of future operating cash flows**.
- Unlike transaction exposure, it matters even when a firm has **no foreign-currency receivables or payables**.
- It is usually the **most important** exposure for long-run firm value.

### <span style="color:#4A86C5;">The valuation channel</span>

$$\boxed{V = \sum_{t=1}^{n}\dfrac{E(CF_t)}{(1+k)^t}}$$

Exchange rates affect $E(CF_t)$ through: revenue, costs, demand elasticity, competitive pricing, and input costs. They also affect the discount rate $k$ through country risk premia.

### <span style="color:#4A86C5;">Why "unexpected" matters</span>

- **Expected** ER changes should already be priced into budgets, forward rates, and interest-rate differentials (recall IRP from Lecture 4).
- Only the **unexpected** component creates economic exposure — the deviation of the realized rate from what was anticipated.

$$\text{Unexpected change} = S_T - E[S_T]$$

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> Economic exposure is a **firm valuation problem**, not just an accounting problem. It determines whether an MNC can compete globally over the next 5–10 years.

---

## <span style="color:#4A86C5;">2.2. Channels of Economic Exposure</span>
---

Exchange-rate shocks propagate through multiple business channels simultaneously:

| Channel | Home currency **appreciates** | Home currency **depreciates** |
|---|---|---|
| **Export revenue** | Foreign buyers find products more expensive → demand falls | Products become cheaper abroad → demand rises |
| **Import costs** | Imported inputs become cheaper → margins improve | Imported inputs become more expensive |
| **Competitive position** | Foreign competitors gain cost advantage | Domestic firms gain cost advantage |
| **Pricing power** | May need to cut prices to maintain market share | Can raise prices or gain market share |
| **Asset values** | Foreign assets worth less in home currency | Foreign assets worth more |

### <span style="color:#4A86C5;">Time-horizon view</span>

| Horizon | What's sticky? | What can the firm change? | Example |
|---|---|---|---|
| **Short run** (0–6 months) | Contracts and prices | Very little; exposure flows directly to margins | Menu prices at restaurants |
| **Medium run** (6–24 months) | Customer/competitor reactions | Reprice, switch suppliers, renegotiate | Car manufacturers adjusting MSRP |
| **Long run** (2+ years) | Strategic position | Relocate production, redesign sourcing, exit markets | Toyota building plants in the U.S. |

---

## <span style="color:#4A86C5;">2.3. Measuring Economic Exposure</span>
---

### <span style="color:#4A86C5;">2.3.1. Regression-based measurement</span>

The standard approach regresses home-currency operating cash flows on the exchange rate:

$$\boxed{\Delta CF_t = \alpha + \beta\,\Delta S_t + \varepsilon_t}$$

- $\beta$ = the firm's **economic FX beta** — measures sensitivity of cash flows to ER movements
- If $\beta > 0$: cash flows increase when FC appreciates (e.g., exporter)
- If $\beta < 0$: cash flows decrease when FC appreciates (e.g., firm competing with imports)
- If $\beta \approx 0$: firm is naturally hedged

### <span style="color:#4A86C5;">2.3.2. Example: Estimating FX beta</span>

A U.S. electronics manufacturer exports 60% of output to the Eurozone. Using quarterly data over 5 years, the regression yields:

$$\Delta CF_t = 2.1 + 18.5\,\Delta(\text{EUR/USD})_t$$

**Interpretation:**
- For every 1-cent increase in EUR/USD (euro appreciation), quarterly USD cash flows rise by \$18.5 million
- The firm has a **positive and significant** FX beta to the euro
- A 10% euro depreciation (from 1.10 to 0.99) would reduce quarterly CFs by approximately \$18.5 × 11 = \$203.5 million

### <span style="color:#4A86C5;">2.3.3. Scenario-based approach</span>

When historical data is insufficient, firms can use scenario analysis:

| Scenario | EUR/USD rate | Revenue (\$m) | COGS (\$m) | Operating CF (\$m) |
|---|---|---|---|---|
| Base case | 1.10 | 500 | 350 | 150 |
| Euro depreciates 10% | 0.99 | 430 | 340 | 90 |
| Euro appreciates 10% | 1.21 | 570 | 360 | 210 |

$$\text{Sensitivity} = \frac{\Delta CF}{\Delta S} = \frac{210 - 90}{1.21 - 0.99} = \frac{120}{0.22} \approx 545 \text{ (\$m per unit of EUR/USD)}$$

## <span style="color:#4A86C5;">2.4. MNC Insight: Airbus — The "Dollar Revenue, Euro Cost" Problem</span>
---

<center>
<img src="../Images/Lecture_5/Pic9.png" alt="Airbus Currency Mismatch" width="850">
</center>

Airbus is one of the world's best-known examples of structural economic exposure:

| Factor | Detail |
|---|---|
| **Revenue currency** | Nearly 100% in USD (aircraft priced in dollars globally) |
| **Cost currency** | >50% in EUR (production in France, Germany, Spain, UK) |
| **Order backlog** | ~8× annual revenue → locked-in USD prices for years |
| **Hedging portfolio** | ~\$100 billion in currency hedges at any given time |

### <span style="color:#4A86C5;">The problem</span>
When the euro **appreciates** against the dollar:
- Revenue in EUR terms **falls** (same USD price → fewer euros)
- Costs remain largely in EUR → **margins squeezed**
- Between 2002–2009, the dollar fell ~50% against the euro, severely pressuring Airbus profitability

### <span style="color:#4A86C5;">Airbus's strategic response</span>
1. **Supply chain relocation**: Shifted work to U.S.-based suppliers to increase dollar-denominated costs
2. **Massive hedging program**: Maintains forward and option hedges covering 3–5 years of projected revenue
3. **Dollar-denominated procurement**: Negotiated contracts with European suppliers in USD where possible

> <span style="color:#4A86C5;">**Lesson for MNCs:**</span> When revenue and costs are in different currencies, financial hedges buy time, but **operational restructuring** is the long-term solution. Airbus demonstrates that even a \$100 billion hedging book cannot fully eliminate economic exposure.

## <span style="color:#4A86C5;">2.5. MNC Insight: Toyota — Production Localization as an Operational Hedge</span>
---

Toyota provides a textbook example of managing economic exposure through **production diversification**:

<center>
<img src="../Images/Lecture_5/Pic11.png" alt="Toyota Production Localization" width="850">
</center>

| Metric | Data (FY2025) |
|---|---|
| **FX sensitivity** | Each ¥1 appreciation vs. USD costs ~¥50 billion in annual operating profit |
| **FY2026 expected FX impact** | ¥745 billion negative impact from yen strengthening |
| **Local production in Europe** | ~70% of vehicles sold in Europe are produced locally |
| **Local production in Asia** | ~95% of vehicles sold in Asia are produced locally |
| **U.S. plants** | Kentucky, Texas, Mississippi, Indiana, Alabama |

### <span style="color:#4A86C5;">Toyota's multi-layered hedging approach</span>

```
Layer 1: Operational hedge (long-term)
├── Build plants where you sell → match revenue and cost currencies
├── Source locally → reduce imported-input exposure
└── Diversify across regions → no single currency dominates

Layer 2: Financial hedge (short/medium-term)
├── Rolling forward contracts → smooth quarterly earnings
├── Currency options → protect against extreme moves
└── Cross-currency swaps → manage long-term debt exposure

Layer 3: Pricing strategy
├── Adjust MSRP in response to sustained ER shifts
└── Absorb short-term fluctuations to maintain market share
```

### <span style="color:#4A86C5;">Recent challenge: Tariffs compound FX risk</span>
In FY2025, Toyota faced a double hit: yen strengthening **plus** U.S. tariffs on imported vehicles. Tariff costs alone reached ¥450 billion in a single quarter — more than double the FX impact. This illustrates that economic exposure extends beyond pure currency risk to broader trade policy.

> <span style="color:#4A86C5;">**Key takeaway:**</span> Toyota's strategy shows that economic exposure is best managed by **moving production to where the revenue is**. Financial hedges alone only buy time.

---

## <span style="color:#4A86C5;">2.6. Managing Economic Exposure: Strategic Toolkit</span>
---

| Strategy | Mechanism | Effectiveness | Time to implement |
|---|---|---|---|
| **Diversify production** | Build plants in multiple countries | High — directly matches costs to revenue | Years |
| **Diversify sourcing** | Source inputs from multiple currency zones | Medium — reduces input-cost exposure | Months |
| **Revenue diversification** | Sell in multiple markets | Medium — reduces dependence on one ER | Months–Years |
| **Natural hedging** | Match FC revenue with FC costs/debt | High — automatic offset | Months |
| **Risk-sharing agreements** | Share ER risk with customers/suppliers | Medium — contractual | Negotiation |
| **Pricing flexibility** | Pass-through ER changes to prices | Variable — depends on elasticity | Immediate |
| **Long-dated swaps** | Lock in ER for multi-year CFs | Medium — costly for long tenors | Days |

### <span style="color:#4A86C5;">Natural hedge: detailed example</span>

A German auto parts manufacturer earns 40% of revenue in USD. To create a natural hedge:

| Action | Currency effect |
|---|---|
| Borrow USD 200m at 5.5% (instead of EUR debt) | Creates USD outflow (interest + principal) |
| Lease a U.S. warehouse | Creates USD operating costs |
| Pay U.S. suppliers in USD | Converts EUR costs into USD costs |
| **Net effect** | USD inflows now partially offset by USD outflows → reduced economic exposure |

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> Financial hedging alone cannot fully offset economic exposure. Long-run FX risk is fundamentally a **business design** problem.

# <span style="color:#4A86C5;">3. Transaction Exposure — Advanced Hedging Strategies</span>
---

*Lecture 6 covered the foundations of transaction exposure: identification, direction of risk, and the basic hedging tools (forward, money market, option). This section focuses on **advanced applications and integration**.*

---

## <span style="color:#4A86C5;">3.1. Quick Review: The Hedging Toolkit</span>
---

| Tool | Locks in rate? | Upside preserved? | Cost | Best for |
|---|---|---|---|---|
| **Forward/Futures** | Yes | No | Minimal (bid-ask) | Known, certain CFs |
| **Money-market hedge** | Yes | No | Interest differential | When forwards unavailable |
| **Put option** (receivable) | Floor only | Yes | Premium | Uncertain CFs; want upside |
| **Call option** (payable) | Cap only | Yes | Premium | Uncertain CFs; want upside |
| **Leading/Lagging** | No | Depends | Opportunity cost | Intra-company; bilateral netting |
| **Cross-currency swap** | Yes (long-term) | No | Swap spread | Long-term debt/investment |
| **Netting** | N/A (reduces gross) | N/A | Internal | Multi-currency MNC operations |

---

## <span style="color:#4A86C5;">3.2. Worked Example: Comparing Hedging Alternatives</span>
---

**Setup:** A U.S. importer must pay **€2,000,000** in 180 days.

| Market data | Value |
|---|---|
| Spot rate (USD/EUR) | 1.0800 |
| 180-day forward rate | 1.0850 |
| EUR 180-day interest rate | 3.0% (annualized) |
| USD 180-day interest rate | 5.2% (annualized) |
| Call option: $K$ = 1.0850, premium = \$0.025/EUR | |

### <span style="color:#4A86C5;">Strategy 1: Forward hedge</span>

$$\text{Cost} = €2{,}000{,}000 \times 1.0850 = \$2{,}170{,}000$$

Certain outcome. No upside if euro depreciates.

### <span style="color:#4A86C5;">Strategy 2: Money-market hedge</span>

**Step 1:** PV of EUR payable:
$$\text{EUR to invest today} = \frac{2{,}000{,}000}{1 + 0.03/2} = \frac{2{,}000{,}000}{1.015} = €1{,}970{,}443.35$$

**Step 2:** Convert to USD at spot:
$$\text{USD needed today} = 1{,}970{,}443.35 \times 1.0800 = \$2{,}128{,}078.82$$

**Step 3:** Borrow USD at 5.2% annualized (2.6% for 180 days):
$$\text{Total USD repayment} = 2{,}128{,}078.82 \times 1.026 = \$2{,}183{,}408.87$$

### <span style="color:#4A86C5;">Strategy 3: Call option hedge</span>

Premium cost: $2{,}000{,}000 \times 0.025 = \$50{,}000$

Future value of premium (at USD rate): $50{,}000 \times 1.026 = \$51{,}300$

| If $S_T$ | Exercise? | Spot cost | + FV Premium | Total cost |
|---|---|---|---|---|
| 1.0500 | No | \$2,100,000 | \$51,300 | **\$2,151,300** |
| 1.0850 | Indifferent | \$2,170,000 | \$51,300 | **\$2,221,300** |
| 1.1200 | Yes | \$2,170,000 | \$51,300 | **\$2,221,300** |
| 1.1500 | Yes | \$2,170,000 | \$51,300 | **\$2,221,300** |

### <span style="color:#4A86C5;">Comparison summary</span>

| Strategy | Cost if EUR at 1.05 | Cost if EUR at 1.085 | Cost if EUR at 1.15 | Flexibility |
|---|---|---|---|---|
| Forward | \$2,170,000 | \$2,170,000 | \$2,170,000 | None |
| Money market | \$2,183,409 | \$2,183,409 | \$2,183,409 | None |
| Call option | **\$2,151,300** | \$2,221,300 | \$2,221,300 | High |
| Unhedged | **\$2,100,000** | \$2,170,000 | **\$2,300,000** | Full (risky) |

> <span style="color:#4A86C5;">**Decision rule:**</span> If the firm expects the euro to **depreciate**, the option preserves upside while capping downside. If the firm wants **certainty**, the forward is cheapest.

## <span style="color:#4A86C5;">3.3. Cross-Currency Swaps for Long-Term Transaction Exposure</span>
---

When transaction exposure extends over **multiple years** (e.g., long-term supply contracts, project finance), forwards become expensive or unavailable. Cross-currency swaps fill this gap.

### <span style="color:#4A86C5;">How it works</span>

A cross-currency swap involves:
1. **Initial exchange** of principals in two currencies at today's spot rate
2. **Periodic interest payments** — each party pays in the currency it received
3. **Final re-exchange** of principals at the original exchange rate

### <span style="color:#4A86C5;">Example</span>

A German firm borrows \$100m for 5 years to fund U.S. operations but earns revenue in EUR:

| Without swap | With EUR/USD cross-currency swap |
|---|---|
| Pays USD interest (exposed to EUR/USD) | Pays EUR interest (matched to revenue) |
| Must convert EUR → USD each period | Counterparty delivers USD interest |
| Repays USD principal at maturity | Re-exchanges at original rate |

The swap effectively converts a USD obligation into a EUR obligation, matching the firm's revenue currency.

## <span style="color:#4A86C5;">3.4. Leading, Lagging, and Reinvoicing Centers</span>
---

### <span style="color:#4A86C5;">Leading and lagging</span>

| Strategy | Mechanism | When to use |
|---|---|---|
| **Lead** (pay early) | Accelerate payment of a FC payable before FC appreciates | Expect FC appreciation |
| **Lag** (pay late) | Delay payment of a FC payable, hoping FC will depreciate | Expect FC depreciation |
| **Lead collection** | Speed up collection of a FC receivable before FC depreciates | Expect FC depreciation |
| **Lag collection** | Delay collection of FC receivable if FC is expected to appreciate | Expect FC appreciation |

### <span style="color:#4A86C5;">Reinvoicing centers</span>

Large MNCs (e.g., Siemens, Nestlé) use **reinvoicing centers** — centralized subsidiaries that:
1. Purchase goods from manufacturing subsidiaries at the transfer price
2. Re-sell to distribution subsidiaries at the market price
3. **All invoicing in the center's currency** → subsidiaries face no FX risk
4. FX exposure is **concentrated** at the reinvoicing center, where specialized treasury teams manage it

### <span style="color:#4A86C5;">MNC fact: Siemens Treasury</span>

Siemens manages over €100 billion in annual internal cash flows through its treasury center in Munich. By centralizing FX management, the company estimates it saves ~€200 million per year in hedging costs compared to decentralized subsidiary-level hedging.

# <span style="color:#4A86C5;">4. Translation (Accounting) Exposure</span>
---

## <span style="color:#4A86C5;">4.1. Definition and Context</span>
---
- Translation exposure arises because foreign subsidiaries report in their **functional currency**, but the parent must consolidate into a **reporting currency**.
- ER changes alter reported net worth and sometimes reported income — **without any cash flow occurring**.

### <span style="color:#4A86C5;">Key terms</span>
| Term | Definition |
|---|---|
| **Reporting currency** | The currency of the parent's consolidated financial statements |
| **Functional currency** | The currency of the subsidiary's primary economic environment |
| **CTA** | Cumulative Translation Adjustment — unrealized gains/losses in equity |

### <span style="color:#4A86C5;">Why it matters even without cash flows</span>

| Impact area | How translation affects it |
|---|---|
| **Debt covenants** | Leverage ratios change → potential covenant violations |
| **Analyst valuations** | Volatile reported equity → lower price multiples |
| **Management KPIs** | ROE, ROA distorted by translation → misleading performance evaluation |
| **Dividend capacity** | Some jurisdictions limit dividends to reported (translated) earnings |
| **Acquisition currency** | If the firm uses stock, volatile book value affects deal terms |

---

## <span style="color:#4A86C5;">4.2. Translation Methods</span>
---
Two key methods (US GAAP ASC 830 / IFRS IAS 21 share similar logic):

### <span style="color:#4A86C5;">4.2.1. Current-rate method</span>
Used when the subsidiary's functional currency is the **local currency** (i.e., it operates relatively independently).

| Item | Translation rate |
|---|---|
| All assets | Current spot rate |
| All liabilities | Current spot rate |
| Common stock | Historical rate (date of issuance) |
| Retained earnings | Cumulative translated amount |
| Revenue & expenses | Average rate for the period |
| **Translation adjustment** | → **CTA in OCI (equity)** |

### <span style="color:#4A86C5;">4.2.2. Temporal method (remeasurement)</span>
Used when the functional currency is the **parent's currency** (e.g., subsidiary is an extension of the parent, or operates in a hyperinflationary economy).

| Item | Translation rate |
|---|---|
| Cash, receivables, payables | Current rate |
| Inventory (at cost) | Historical rate |
| Fixed assets | Historical rate |
| Revenue & expenses | Average rate (except COGS, depreciation → historical) |
| **Translation adjustment** | → **Net income** |

### <span style="color:#4A86C5;">4.2.3. Side-by-side comparison</span>

| Feature | Current-rate | Temporal |
|---|---|---|
| Which items at current rate? | All A & L | Monetary items only |
| Non-monetary assets | Current rate | Historical rate |
| Income statement | Average rate | Mixed |
| Gain/loss goes to | CTA (equity) | Net income |
| Income volatility | Low | High |
| Balance-sheet volatility | High | Low |
| **Net exposed position** | Net assets = Total assets − Total liabilities | Net monetary assets |

---

## <span style="color:#4A86C5;">4.3. Worked Example: Translation Under Both Methods</span>
---

A U.S. parent has a **Mexican subsidiary** with the following balance sheet (in MXN millions):

| Item | MXN (millions) |
|---|---|
| Cash | 50 |
| Accounts receivable | 150 |
| Inventory (at cost) | 200 |
| Fixed assets (net) | 600 |
| **Total assets** | **1,000** |
| | |
| Accounts payable | 100 |
| Long-term debt | 400 |
| **Total liabilities** | **500** |
| | |

| Retained earnings | 200 |

| **Total L + E** | **1,000** |

**Exchange rates:**
- Historical rate (when stock issued / assets acquired): 0.060 USD/MXN
- Beginning-of-year rate: 0.055 USD/MXN
- End-of-year (current) rate: **0.048 USD/MXN** (peso depreciated)
- Average rate for the year: 0.052 USD/MXN

| Item | MXN (m) | Rate | USD (m) |
|---|---|---|---|
| Cash | 50 | 0.048 (C) | 2.40 |
| Accounts receivable | 150 | 0.048 (C) | 7.20 |
| Inventory | 200 | 0.048 (C) | 9.60 |
| Fixed assets | 600 | 0.048 (C) | 28.80 |
| **Total assets** | **1,000** | | **48.00** |
| | | | |
| Accounts payable | 100 | 0.048 (C) | 4.80 |
| Long-term debt | 400 | 0.048 (C) | 19.20 |

**Net exposed assets** = Total assets − Total liabilities = MXN 500m

**Translation loss** from peso depreciation: The CTA of −\$5.00m reflects the cumulative impact of translating MXN 500m of net assets at a declining exchange rate.

### <span style="color:#4A86C5;">4.3.2. Temporal method</span>

| Item | MXN (m) | Rate | USD (m) |
|---|---|---|---|
| Cash | 50 | 0.048 (C) | 2.40 |
| Accounts receivable | 150 | 0.048 (C) | 7.20 |
| Inventory | 200 | 0.060 (H) | 12.00 |

| **Total assets** | **1,000** | | **57.60** |
| | | | |
| Accounts payable | 100 | 0.048 (C) | 4.80 |
| Long-term debt | 400 | 0.048 (C) | 19.20 |
| **Total liabilities** | **500** | | **24.00** |
| | | | |
| Common stock | 300 | 0.060 (H) | 18.00 |
| Retained earnings | 200 | (cumulative) | 11.00* |
| **Remeasurement gain (plug)** | — | — | **+4.60** |

When the peso depreciates and the firm has net monetary **liabilities** in pesos, it has a **remeasurement gain** (owes less in USD terms). This gain goes directly to **net income**.

## <span style="color:#4A86C5;">4.4. Managing Translation Exposure</span>
---

### <span style="color:#4A86C5;">4.4.1. Balance-sheet hedge</span>

The most common approach: **match exposed assets with exposed liabilities** in the same currency.

$$\boxed{\text{Net Exposed Assets} = \text{Exposed Assets} - \text{Exposed Liabilities}}$$

**Goal:** Drive net exposed assets toward zero.

### <span style="color:#4A86C5;">4.4.2. Worked example: designing a balance-sheet hedge</span>

From the Mexican subsidiary example (current-rate method):
- Net exposed assets = MXN 500m
- To eliminate translation exposure: increase MXN liabilities by MXN 500m, or reduce MXN assets

**Option A:** Parent borrows MXN 500m (or equivalent via swap)
- Now: Exposed assets = MXN 1,000m, Exposed liabilities = MXN 1,000m
- Net exposed = **zero** → Translation gains/losses cancel

**Option B:** Subsidiary borrows MXN 300m locally and remits dividends to parent
- Reduces net equity in subsidiary → reduces net exposed assets

### <span style="color:#4A86C5;">4.4.3. Why hedge (or not)?</span>

| Argument **for** hedging | Argument **against** hedging |
|---|---|
| Reduces CTA volatility | CTA doesn't affect cash flows |
| Protects debt covenants | Hedging has real costs (interest, fees) |
| Stabilizes reported metrics | Creates unnecessary transaction exposure |
| Aligns with analyst expectations | Reverses if ER bounces back |

## <span style="color:#4A86C5;">4.5. MNC Insight: Procter & Gamble — Translation Exposure at Scale</span>
---

P&G operates in over 70 countries with significant non-USD revenue and assets.

| Metric | Detail |
|---|---|
| **FY2024 FX headwind** | ~\$300 million after tax (translation + transaction combined) |
| **Argentina impact** | \$0.8 billion charge from liquidating Argentine operations (accumulated CTA losses) |
| **FY2026 outlook** | FX expected to be a **tailwind** of ~\$200 million (dollar weakening) |
| **Currency-neutral reporting** | P&G reports both GAAP and currency-neutral EPS to isolate operational performance |

### <span style="color:#4A86C5;">The Argentina case</span>

Argentina's persistent hyperinflation required P&G to use the **temporal method** (remeasurement) for its Argentine subsidiary. When P&G liquidated operations in 2024:
- Accumulated foreign currency translation losses of \$0.8 billion — previously sitting in equity (OCI) — were **reclassified to net income**
- This is a real-world example of CTA crystallizing into an income statement hit

### <span style="color:#4A86C5;">P&G's currency-neutral reporting</span>

Many MNCs now report **currency-neutral** metrics alongside GAAP figures:

$$\text{Currency-neutral growth} = \text{Reported growth} + \text{FX impact}$$

In FY2025, P&G's currency-neutral core EPS grew 4%, while GAAP EPS growth was lower due to FX headwinds. This practice helps analysts separate **operational performance** from **translation noise**.

> <span style="color:#4A86C5;">**Lesson:**</span> Translation exposure may not affect cash flows directly, but it can have massive effects on reported financials, analyst perceptions, and strategic decisions (like P&G's exit from Argentina).

## <span style="color:#4A86C5;">Case 1: Nestlé — Managing All Three Exposures Simultaneously</span>
---

Nestlé, headquartered in Switzerland (CHF), operates in 186 countries with revenues in dozens of currencies.

### <span style="color:#4A86C5;">Exposure profile</span>

| Exposure type | Nestlé's situation |
|---|---|
| **Transaction** | Purchases coffee, cocoa, and dairy in USD, BRL, GHS; sells in EUR, GBP, JPY, etc. |
| **Economic** | Swiss franc appreciation makes Nestlé products more expensive vs. local competitors |
| **Translation** | Consolidating 400+ subsidiaries into CHF creates massive CTA volatility |

### <span style="color:#4A86C5;">Nestlé's integrated approach</span>

1. **Centralized treasury in Vevey** — all FX positions netted before hedging externally
2. **Natural hedging** — local production in major markets (e.g., coffee processed in Brazil, products manufactured in the U.S.)
3. **Financial hedges** — forwards and options for 12–18 month horizon on net transaction exposures
4. **Selective translation hedging** — only for currencies with high volatility or covenant implications
5. **Pricing power** — strong brands allow partial pass-through of FX costs to consumers

### <span style="color:#4A86C5;">Key numbers</span>
- Nestlé's FX loss in 2023: CHF 3.2 billion in translation effects
- Organic (currency-neutral) growth: 7.2% vs. reported growth of 1.5% — the difference is almost entirely FX translation

> <span style="color:#4A86C5;">**Lesson:**</span> Even the world's largest food company cannot eliminate FX exposure. The goal is to **manage** it cost-effectively, not to eliminate it entirely.

## <span style="color:#4A86C5;">Case 2: Volkswagen Group — When Hedging Backfires</span>
---

In 2014–2015, the EUR/USD rate fell sharply from ~1.35 to ~1.05. For most European exporters, this was excellent news — their products became cheaper in dollar terms.

### <span style="color:#4A86C5;">The problem</span>
- Volkswagen had **heavily hedged** its USD revenue at rates around 1.30–1.35
- When the euro weakened, VW was **locked into unfavorable forward rates** and missed the windfall
- Competitors like BMW, which hedged less aggressively, captured the full benefit

### <span style="color:#4A86C5;">Outcome</span>

| Company | Hedging strategy | EUR/USD benefit captured |
|---|---|---|
| Volkswagen | Aggressive (80%+ hedged) | Minimal — locked in at old rates |
| BMW | Moderate (50% hedged) | Significant — captured ~50% of move |
| Unhedged competitor | No hedge | Full — but took full risk |

### <span style="color:#4A86C5;">Lessons</span>
1. Hedging **eliminates both downside and upside** — it is not free
2. The **optimal hedge ratio** depends on the firm's risk appetite and competitive position
3. Over-hedging can be as costly as under-hedging in a trending market

A U.S. multinational sells industrial equipment globally and owns a Brazilian subsidiary. In Q3, the BRL devalues 30% against USD.

### <span style="color:#5b9bd5;">Solution 1: Multi-Exposure Identification and Quantification

**1.** 
- Japanese subsidiary net assets → **Translation exposure**
- Euro receivable → **Transaction exposure**  
- Tourist revenue (no contract) → **Economic exposure**

**2. Translation loss:**
- Pre-depreciation: ¥20 bn × 0.0067 = \$134.0m
- Post-depreciation (8% yen weakening): new rate = 0.0067 × (1 − 0.08) = 0.006164
- Post: ¥20 bn × 0.006164 = \$123.28m
- **Translation loss = \$134.0m − \$123.28m = \$10.72m** (goes to CTA)

**3. Euro receivable:**
- Current value: €5m × 1.0800 = \$5,400,000
- After 3% euro depreciation: €5m × (1.0800 × 0.97) = €5m × 1.0476 = \$5,238,000
- **Loss = \$162,000** (but this is a spot change; the firm can still hedge with the forward at 1.0830 if it acts now)

**4.** Tourist revenue falls because a stronger dollar makes U.S. travel more expensive for Japanese tourists. This is economic exposure because:
- There is no FC contract to hedge
- The revenue shortfall depends on tourist demand elasticity, which is uncertain
- Forwards/options cannot hedge **volume** risk — only **price** risk on known CF amounts

### <span style="color:#5b9bd5;">Solution 2: Economic Exposure Regression

**1.** The coefficient 32.5 means: for every 1-unit increase in GBP/USD (i.e., £1 per dollar — which would be a massive move), quarterly cash flows increase by £32.5m. More practically, for every **0.01** increase in GBP/USD, cash flows rise by approximately £325,000. The firm benefits from GBP appreciation (consistent with being a U.K. exporter receiving USD).

**2.** $\Delta S = 1.18 - 1.27 = -0.09$

$\Delta CF = 1.2 + 32.5 \times (-0.09) = 1.2 - 2.925 = -£1.725\text{m}$

Quarterly cash flow **falls by approximately £1.7 million**.

**3.** The remaining 59% represents cash-flow variation driven by factors other than GBP/USD, such as:
- Changes in demand/volume unrelated to ER
- Input cost fluctuations (e.g., commodity prices)
- Competitive actions by rivals
- Domestic economic conditions
- Seasonal factors

**4.**
- **Operational 1:** Open a manufacturing facility in the U.S. to convert some USD revenue into USD costs (natural hedge)
- **Operational 2:** Negotiate GBP-denominated contracts with key U.S. customers (shift FX risk)
- **Financial:** Enter into a rolling program of GBP/USD forward contracts covering 50–70% of expected quarterly USD receipts over a 12-month horizon

**5.** A 5-year forward:
- Only fixes the **ER** on a known amount — but the firm's **volume** of exports in 5 years is uncertain
- If sales decline, the firm is over-hedged and faces a loss on the forward
- Forward rates at long tenors embed large interest-rate differentials and liquidity premiums
- Cannot address **competitive position** effects (lost market share if GBP strengthens)
- Economic exposure requires operational, not just financial, solutions

### <span style="color:#5b9bd5;">Solution 3: Translation Methods

**1. Total assets in USD:**

**Current-rate method:** All assets at current rate
$$1{,}000 \times 0.00074 = \$0.740\text{ bn} = \$740\text{m}$$

**Temporal method:** Monetary at current; non-monetary at historical
$$(20 + 80) \times 0.00074 + (150 + 750) \times 0.00082$$
$$= 100 \times 0.00074 + 900 \times 0.00082$$
$$= 0.074 + 0.738 = \$0.812\text{ bn} = \$812\text{m}$$

**2. Net exposed position:**

**Current-rate:** Net assets = Total assets − Total liabilities = 1,000 − 400 = **KRW +600 bn** (positive, exposed to depreciation loss)

**Temporal:** Net monetary assets = (Cash 20 + AR 80) − (AP 60 + LTD 340) = 100 − 400 = **KRW −300 bn** (negative = net monetary liability)

**3.** Under the **temporal method**, the KRW depreciation creates a **gain**. The firm has net monetary liabilities of KRW 300 bn. When KRW depreciates, these liabilities are worth fewer USD → remeasurement **gain**, which appears in **current net income**.

Under the current-rate method, net exposed assets are positive (KRW 600 bn), so depreciation creates a **loss** → goes to **CTA in equity**.

**4.** To hedge under current-rate method:
- Net exposed assets = KRW 600 bn
- **Instrument:** Borrow KRW 600 billion (or use a cross-currency swap to create synthetic KRW liability)
- This brings net exposed assets to zero
- Alternative: sell KRW 600 bn forward against USD for the reporting period
- Approximate USD value: KRW 600 bn × 0.00074 = **\$444 million** forward contract

### <span style="color:#5b9bd5;">Solution 4: Hedging Strategy Comparison

**1. Forward hedge:**
$$\text{EUR proceeds} = \$10{,}000{,}000 \times 0.9310 = €9{,}310{,}000$$

**2. Money-market hedge:**

**Step 1:** Borrow USD today (PV of receivable):
$$\text{USD borrowed} = \frac{10{,}000{,}000}{1 + 0.054/4} = \frac{10{,}000{,}000}{1.0135} = \$9{,}866{,}798.72$$

**Step 2:** Convert to EUR at spot:
$$\text{EUR today} = 9{,}866{,}798.72 \times 0.9250 = €9{,}126{,}788.82$$

**Step 3:** Invest EUR at 3.6% annualized (0.9% for 90 days):
$$\text{EUR in 90 days} = 9{,}126{,}788.82 \times 1.009 = €9{,}208{,}929.92$$

**Step 4:** When USD 10m arrives, repay USD loan. EUR proceeds = **€9,208,930**.

**3. Put option hedge:**

Premium: \$10m × €0.012 = €120,000. FV of premium: €120,000 × 1.009 = €120,893.

| Scenario | $S_T$ | Exercise? | Gross EUR | − FV Premium | Net EUR |
|---|---|---|---|---|---|
| USD strengthens | 0.9500 | No (spot > K) | €9,500,000 | €120,893 | **€9,379,107** |
| Stable | 0.9300 | No (spot > K) | €9,300,000 | €120,893 | **€9,179,107** |
| USD weakens | 0.9000 | Yes | €9,250,000 | €120,893 | **€9,129,107** |
| USD crashes | 0.8600 | Yes | €9,250,000 | €120,893 | **€9,129,107** |

**4. Expected value of option hedge:**

$$E(V) = 0.25(9{,}379{,}107) + 0.40(9{,}179{,}107) + 0.25(9{,}129{,}107) + 0.10(9{,}129{,}107)$$
$$= 2{,}344{,}777 + 3{,}671{,}643 + 2{,}282{,}277 + 912{,}911 = €9{,}211{,}607$$

**5. Comparison:**

| Strategy | Proceeds | Risk |
|---|---|---|
| Forward | €9,310,000 | Zero — locked in |
| Money market | €9,208,930 | Zero — locked in |
| Option (expected) | €9,211,607 | Variable (€9,129,107 – €9,379,107) |

- **Maximizes expected value:** Forward hedge (€9,310,000)
- **Minimizes risk:** Forward hedge (zero variance)
- **Recommendation:** The forward hedge dominates — it offers both the highest expected value **and** zero risk. The money-market hedge is slightly inferior due to interest-rate differentials. The option would be preferred only if there were a high probability of significant USD appreciation (upside scenario), which is not the case here (only 25% chance of 0.9500). Given the heavy downside tail (10% chance of 0.86), the forward's certainty is especially valuable.

### <span style="color:#5b9bd5;">Solution 5: Integrated MNC Case — EuroJet

**1. Impact on annual operating profit:**

At EUR/USD = 1.10:
- Revenue in EUR: \$15 bn ÷ 1.10 = **€13.636 bn**
- EUR costs: €8.0 bn
- USD costs in EUR: \$4.5 bn ÷ 1.10 = €4.091 bn
- GBP costs in EUR: Assume GBP/EUR = 0.87 → £2.3 bn ÷ 0.87 = €2.644 bn
- **Operating profit = 13.636 − 8.0 − 4.091 − 2.644 = −€1.099 bn** 

Wait — let's recalculate. The total revenue is USD 15 bn. Total costs: €8 bn + \$4.5 bn + £2.3 bn. Let's convert everything:

At EUR/USD = 1.10 (i.e., \$1.10 per €1, or €0.909 per \$1):
- Revenue: \$15 bn × (1/1.10) = **€13.636 bn**
- Costs: €8.0 + \$4.5 × (1/1.10) + £2.3 × (1/0.87*) = 8.0 + 4.091 + 2.644 = **€14.735 bn**

*Assume GBP/EUR unchanged for simplicity.

This gives a loss, which suggests the numbers need adjustment. Let's assume the costs are: 55% of revenue in EUR, 30% in USD, 15% in GBP. So total costs < revenue:
- Total cost ≈ \$13.5 bn equivalent → profit margin ~10%
- EUR costs: 0.55 × 13.5 = \$7.425 bn equivalent = €6.75 bn
- USD costs: 0.30 × 13.5 = \$4.05 bn
- GBP costs: 0.15 × 13.5 = \$2.025 bn equivalent

At EUR/USD = 1.10:
- Revenue: €13.636 bn
- EUR costs: €6.75 bn
- USD costs: \$4.05 ÷ 1.10 = €3.682 bn
- GBP costs: \$2.025 ÷ 1.10 = €1.841 bn (assuming GBP moves proportionally)
- **Operating profit = 13.636 − 6.75 − 3.682 − 1.841 = €1.363 bn**

At EUR/USD = 1.21 (10% EUR appreciation):
- Revenue: \$15 ÷ 1.21 = **€12.397 bn**
- EUR costs: €6.75 bn (unchanged)
- USD costs: \$4.05 ÷ 1.21 = €3.347 bn
- GBP costs: ≈ €1.841 bn (assume stable vs EUR for simplicity)
- **Operating profit = 12.397 − 6.75 − 3.347 − 1.841 = €0.459 bn**

**Impact: operating profit falls from €1.363 bn to €0.459 bn — a decline of €904 million (−66%)!**

This dramatic impact occurs because: (a) revenue is 100% in USD, (b) only 30% of costs are in USD, and (c) 55% of costs are in EUR and do not decline.

**2. Transaction exposure:**
- USD receivable: \$2 bn × (1/1.10) = €1.818 bn at current rate → if EUR/USD moves to 1.21: \$2 bn × (1/1.21) = €1.653 bn → **loss: €165m**
- GBP payable: £500m. If GBP/EUR remains constant: no additional FX impact. If GBP appreciates vs EUR, the payable increases.

**3. Hedged revenue (60% at 1.08):**
- Hedged portion: 60% × \$15 bn = \$9 bn
- At forward rate 1.08: \$9 bn ÷ 1.08 = **€8.333 bn**
- Unhedged (at 1.21): \$6 bn ÷ 1.21 = **€4.959 bn**
- Total: €13.292 bn

vs. fully unhedged at 1.21: \$15 bn ÷ 1.21 = €12.397 bn

**The hedge saves €895 million** compared to being fully unhedged.

**4. Three-layer strategy:**

**Layer 1 — Operational (long-term, 2–5 years):**
- Shift 20% of EUR-denominated production to U.S.-based facilities → increases USD cost share from 30% to ~45%
- Negotiate USD-denominated supply contracts with European tier-1 suppliers
- Target: reduce the EUR/USD revenue-cost mismatch structurally

**Layer 2 — Financial (short/medium-term, 0–24 months):**
- Maintain 60–70% forward hedge on USD revenue at 12–18 month tenor
- Purchase EUR call / USD put options for additional 15% of revenue at 6-month tenor for tail-risk protection
- Use cross-currency swaps to convert any EUR-denominated long-term debt into USD debt

**Layer 3 — Strategic (ongoing):**
- Introduce EUR-denominated pricing for select European airline customers (reduces USD revenue concentration)
- Build a reinvoicing center to centralize and net all intra-group FX flows
- Establish risk-sharing clauses in long-term customer contracts (e.g., price adjustment if EUR/USD moves >5%)

**5. Trade-offs of 60% → 90% hedge ratio:**

| Factor | 60% hedge | 90% hedge |
|---|---|---|
| **Downside protection** | Moderate | High |
| **Upside participation** | 40% of revenue benefits if EUR weakens | Only 10% benefits |
| **Cost** | Lower (fewer forwards/options) | Higher premium; wider bid-ask on larger notional |
| **Over-hedge risk** | Low — 40% buffer for volume uncertainty | High — if order cancellations occur, firm is over-hedged |
| **Competitive flexibility** | Can cut EUR prices if EUR weakens | Locked in; competitors may undercut |

The Volkswagen experience (2014–15) is cautionary: aggressive hedging at 80%+ caused VW to miss a major EUR depreciation windfall, while BMW (at ~50%) captured significant benefit. The optimal ratio depends on **volume certainty** and **competitive dynamics**.
