# <span style="color:#4A86C5;">1. Currency Derivatives and Financial Risk Management
---


## <span style="color:#4A86C5;">1.1. Motivation for Derivative Use
---


### <span style="color:#5b9bd5;">1.1.1. The Overall Picture
---

The GFC demonstrated that the financial sector and MNCs are exposed not only to firm-specific vulnerabilities but also to systemic risks arising from the strong interconnections across global financial institutions and markets.

In response, firms have increasingly relied on FX derivatives over the past two decades to hedge currency exposures and stabilise cash flows.

**BIS data (April 2025)** show that:
- Derivatives markets have expanded significantly. 
- The notional amount of FX derivatives has risen steadily:
    - exceeding USD 120 trillion in OTC markets, 
    - while daily FX derivatives turnover has increased sharply, reaching around USD 9 trillion. 

- Although most FX derivatives remain traded over-the-counter, the use of exchange-traded instruments has also grown, reflecting stronger risk management and regulatory reforms after the GFC.
    -  In fact, IR futures and options are highly standardised and widely traded on exchanges (over 60% of total IR derivatives turnover in April 2025 is exchange-traded).

### <span style="color:#5b9bd5;">1.1.2. Financial Risks Faced by MNCs
---
**Exchange rate (ER) risk**
- ER risk arises because MNCs receive revenues, incur costs, borrow, invest, and report results in different currencies.

- ER risk matters because ER fluctuations:
    * Create uncertainty in cash flows and profitability.
    * Affect budgeting, pricing, and financing decisions.
    * Can distort reported financial performance.
    * May reduce firm value if exposures are left unmanaged.
- To measure ER risk, banks calculate measures of net exposure by each currency. It will be equal to the difference between the assets and liabilities denominated in the same currency. 

**Interest rate (IR) risk**
- Interest rate movements affect:
    - borrowing costs,
    - minvestment returns, and 
    - the value of interest-sensitive assets and liabilities.

**Market volatility in global capital markets**
- Volatility in ERs, interest rates, and capital markets increases uncertainty for globally active firms and strengthens the need for risk management tools.

**Main ER exposures faced by MNCs**
- Transaction exposure: risk that contracted foreign-currency payables or receivables change in value before settlement.
- Translation exposure: risk that exchange-rate movements affect the reported value of foreign subsidiaries’ assets, liabilities, income, and equity when consolidated.
- Economic exposure: risk that long-term exchange-rate changes alter the firm’s competitive position, expected cash flows, and firm value.

---




### <span style="color:#5b9bd5;">1.1.3. Role of Financial Derivatives
---

- Modern financial management for MNCs heavily relies on financial derivatives. 
- The term "derivative" means that these instruments derive their value from an underlying asset (e.g., currencies, interest rates, securities, commodities, or market indexes).

Currency derivatives serve three primary functions in global finance:
- **Hedging currency exposure**: They allow firms to reduce the uncertainty of foreign-currency cash flows by locking in ERs or creating a safety net against unfavorable market movements.
- **Managing interest rate risk**: They enable MNCs to stabilize financing costs and protect against adverse fluctuations in interest rates.
- **Speculation and arbitrage**: They provide mechanisms to profit from expected market movements or pricing inefficiencies (This differs fundamentally from corporate risk management, which focuses on risk mitigation rather than profit generation).

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - Derivatives are powerful tools in the hands of careful and competent financial managers. 
> - However, when used recklessly, they can magnify losses rather than reduce risk (~gambling). 
> - Their effectiveness depends entirely on clear objectives, sound internal controls, and a precise understanding of the exposure being managed.

# <span style="color:#4A86C5;">2. Foreign Currency Futures</span>

---


## <span style="color:#4A86C5;">2.1. Fundamentals of Currency Futures</span>

---

### <span style="color:#5b9bd5;">2.1.1. Overview

**Forward contract:**
- A forward contract is a customized, private agreement between two parties to buy or sell an asset at a specified price on a specific future date. 
- Unlike futures, forwards are not standardized or traded on a centralized exchange.

**Futures Contract:**
- A foreign currency futures contract serves as an alternative to a forward contract. 
- It mandates the future delivery of a standardized amount of foreign currency at a fixed time, place, and price.
- Conceptually, futures contracts operate like traditional commodity futures (e.g., contracts for cattle, gold, or lumber).
- In the U.S., the premier trading hub for these instruments is the International Monetary Market (IMM) of the Chicago Mercantile Exchange (CME).

**Major contract specifications include**:
* Contract Size: otional principal (e.g., each Mexican peso contract is strictly for 500,000 pesos).
* Maturity Dates: Contracts typically mature on the third Wednesday of specific months (e.g., January, March, April, etc.).
* Method of stating ERs: Usually stated in USD terms (USD cost of a foreign currency unit).

**Example:** Mexican Peso (CME) Quotes seen in an exchange:

| Maturity | Open | High | Low | Settle | Open Interest |
| --- | --- | --- | --- | --- | --- |
| **Mar** | 0.10953 | 0.10988 | 0.10930 | 0.10958 | 34,481.00 |
| **June** | 0.10790 | 0.10795 | 0.10778 | 0.10773 | 3,405.00 |

---

### <span style="color:#5b9bd5;">2.1.2. Trading Mechanisms & Settlement

Futures employ a standardized trading and settlement mechanism designed to mitigate default risk and provide market liquidity:
* **Brokerage Execution and Commissions:** 
    - Unlike the forward market (dealers quote bid/ask spreads and don't charge fees), futures customers pay a flat commission to a broker to execute trades directly on the exchange.
* **The Role of the Clearinghouse:** 
    - Once a trade is executed, the exchange's clearinghouse steps in to act as the counterparty to both sides. 
    - It becomes the buyer to every seller, and the seller to every buyer.
    - This completely eliminates counterparty default risk.
* **Margin Requirements:** 
    - Both buyers and sellers must deposit an Initial Margin to initiate the trade. 
    - Maintenance margins ensure the account does not drop below a critical level.
* **Marked-to-Market:** 
    - Unlike forwards, which settle entirely at maturity, the futures contract's value is revalued daily based on the market's closing settlement price. 
    - Any resulting profits or losses are credited to or deducted from the trader's margin account in cash at the end of each trading day.
* **Closing Positions:**
    * Only about 5% of currency futures result in the actual physical exchange of currency.
    * The vast majority of participants close out their obligation prior to the delivery date by taking an offsetting, opposite position (a.k.a "round turn").

* **Open Interest:**
    * Represents the number of outstanding contracts that have not yet been offset by a round turn or settled by delivery.
    * Is a good measure of liquidity, an indication of the strength of the market.

---

## <span style="color:#4A86C5;">2.2. Pricing of Currency Futures</span>

---



### <span style="color:#5b9bd5;"> Market Mechanics: When to Buy or Sell (1)

Speculators and hedgers use currency futures to take positions based on the expected future direction of an ER.

**Long Position**
- An investor *buys* a futures contract when they expect the underlying currency to appreciate.
- A long position profits if the spot rate at maturity exceeds the futures price agreed today.

$$\boxed{\text{Profit/Loss} = \text{Notional Principal} \times (S_T - F_0)}$$

**Short Position**
- An investor *sells* a futures contract when they expect the underlying currency to depreciate.
- A short position profits if the spot rate at maturity is below the rate agreed today.

$$\boxed{\text{Profit/Loss} = -\text{Notional Principal} \times (S_T - F_0)}$$

Where:

- $S_T$ = spot ER at maturity  
- $F_0$ = futures price at contract initiation  

---


### <span style="color:#5b9bd5;">Market Mechanics: When to Buy or Sell (2)

**Example:**

An investor expects the Mexican peso (MXN) to depreciate against the U.S. dollar (USD).

They take the following position:

- **Action:** Sell one May futures contract  
- **Notional Principal:** 500,000 MXN  
- **Futures Price:** 0.10958 USD/MXN  

At maturity, the spot rate is:

$$S_T = 0.09500 \text{ USD/MXN}$$

Using the short position formula:

$$\text{Profit/Loss} = -500{,}000 \times (0.09500 - 0.10958)$$

$$= -500{,}000 \times (-0.01458)$$

$$= 7{,}290 \text{ USD}$$

**Result:** The investor earns a **profit of USD 7,290** because the peso depreciated below the agreed futures price.

---


## <span style="color:#4A86C5;">2.3. Hedging with Currency Futures</span>
---


While speculators use futures to profit from directional views of ERs, MNCs use the same instruments to hedge underlying business exposures arising from international trade and investment.

Futures contracts allow firms to lock in FX rates today, thereby reducing uncertainty about the domestic currency value of future foreign-currency transactions.

---

### <span style="color:#5b9bd5;">2.3.1. Hedging Payables</span>
**Protecting against currency appreciation:**
- If an MNC has a future foreign currency payable, the firm faces the risk that the foreign currency appreciates before or on the payment date.
- To hedge this exposure, the firm can take a **long position** in currency futures to lock in the ER at which it can purchase the foreign currency in the future.

* If the foreign currency **appreciates**, the firm experiences:
    - A higher cost when paying the supplier in the spot market.
    - An offsetting gain on the long futures position.

* If the foreign currency **does not appreciate** (or depreciates):
    - The cost paid to supplier remains unchanged or lowered.
    - An offsetting loss on the long futures position.

In both scenarios, the futures contract protects the firm against uncertainty, locking in the effective cost at the original futures price.

---


> <span style="color:#9E86C5;">**Question:**</span><br>
>A U.S. company must pay **€1,000,000 in three months** to a European supplier.
> * Current EUR futures price: $F_0 = 1.10$ USD/EUR
> * Three months later, the spot rate becomes: $S_T = 1.16$ USD/EUR
> 1. What is the gain or loss on the futures contract?
> 2. How does the hedge affect the firm's effective ER?

<span style="color:#9e86C5;">**Solution:**</span><br>

**1. Futures Payoff:**
$$\text{Profit} = \text{Notional} \times (S_T - F_0)$$
$$\text{Profit} = 1,000,000 \times (1.16 - 1.10)$$
$$\text{Profit} = 60,000\ \text{USD}$$

**2. Effective ER:**
- The firm pays more euros in the spot market but gains $60,000 from the futures position, effectively stabilizing the cost of the payable. 
    * Cost to buy euros at new spot rate: €1,000,000 × 1.16 = $1,160,000
    * Less gain from futures contract: -$60,000
    * Net Effective Cost: $1,100,000

---

### <span style="color:#5b9bd5;">2.3.2. Hedging Receivables</span>


**Protecting against currency depreciation:**
- If an MNC expects to receive foreign currency in the future, it faces the risk that the currency may depreciate before the funds are converted into the firm's home currency.
- If depreciation occurs, the domestic currency value of the receivable falls.
- To hedge this exposure, the firm takes a short position in currency futures to lock in the ER at which it can sell the foreign currency in the future.

# <span style="color:#4A86C5;">3. Foreign Currency Options</span>

---



## <span style="color:#4A86C5;">3.1. Fundamentals of Currency Options</span>

---



### <span style="color:#5b9bd5;">3.1.1. Definition

- A **foreign currency option** is a contract giving the option buyer the right, but not the obligation, to buy or sell a given amount of foreign exchange at a fixed price per unit for a specified time period (until the maturity date).

- Unlike forward or futures contracts where execution is mandatory, options provide flexibility.
- Two types of option:
    * A call is an option to *buy* foreign currency.
    * A put is an option to *sell* foreign currency.

---



### <span style="color:#5b9bd5;">3.1.2. Option Elements

#### Price Elements
1. The Exercise (or Strike) Price $K$: The fixed ER at which the foreign currency can be purchased (for a call) or sold (for a put).
2. The Spot Rate $S_T$: The actual, real-time spot ER currently trading in the market.
3. The Premium: The cost, price, or actual value paid to purchase the option.

--- 

#### American vs. European Options
* American Option: Buyers have the right to exercise the option at any time between the date of writing and the expiration/maturity date.
* European Option: Can be exercised only on its expiration date.


<center>
    <img src="../Images/Lecture_5/Pic14.png" alt="Pic14" width="800"> </center>
</center>

---



#### Moneyness (a.k.a Intrinsic value)
Moneyness describes the relationship between the option’s strike price (K) and the current spot rate ($S_T$) at time (T) of the underlying currency, ignoring the premium.

- **At-the-money (ATM)**  
    - An option is ATM when the strike price is approximately equal to the current spot ER.
    - At this point, exercising the option immediately would generate **no gain and no loss** as $K \approx S$

- **In-the-money (ITM)**  
    - An option is ITM if exercising it immediately would generate a **positive payoff**.
        - For call options: $S_T > K$
        - For put options: $S_T < K$

- **Out-of-the-money (OTM)**  
    - An option is OTM if exercising it immediately would **not generate a positive payoff**.
        - For call options: $S_T < K$
        - For put options: $S_T > K$

---


#### Example: A simple Profit/Loss Profile for a Call Option Buyer
* Underlying: Swiss franc (CHF)
* Option type: European - Long call
* Strike price: $K = 0.585$ USD/CHF
* Premium: 0.005 USD/CHF

<center>
    <img src="../Images/Lecture_5/Pic3.png" alt="pic3" width="620"> </center>
</center>

Profit Formula
$$\boxed{\text{Net Profit } \Pi_{\text{long call}} = \max(S_T - K, 0) - \text{premium}}$$

The break-even rate is:
$$\boxed{\text{Break-even } S_T = K + \text{premium} = 0.585 + 0.005 = 0.590}$$

Where:
* $S_T$ = spot rate at maturity
* $K$ = strike price



Therefore:
* $S_T<0.590$: The buyer has a net loss (which is capped at the maximum loss of the 0.005 premium if $S<K$.
* $S_T>0.590$: The buyer recovers the premium and begins earning a net profit.

If the CHF spot rate at maturity S_T is **0.595** USD/CHF:
$$\Pi = (0.595 - 0.585) - 0.005$$
$$\Pi = 0.005\ \text{USD/CHF}$$

That means the buyer earns a net profit of **$0.005** per CHF.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> * Long call options have **limited loss** and **unlimited gain**.
> * An option writer (short position) has the exact opposite profit/loss profile to an option buyer.

---

Over the last few decades, foreign currency options have become an important segment of global financial markets, serving both **hedging** and **speculative** purposes.

Firms and investors can obtain currency options through two main types of markets:

* **Over-the-Counter (OTC) Market:** 
    - Currency options are negotiated directly with banks or financial institutions. 
    - These contracts can be customized to match the firm's exact needs in terms of notional amount, maturity, and strike price. 
    - This flexibility is a major advantage, but because the contract is private, the firm is exposed to counterparty risk (the possibility that the other party may fail to honor the agreement).

* **Organized Exchange:** 
    - Currency options are traded in standardized contract sizes and with predetermined maturity dates. 
    - Although this standardization reduces flexibility, it offers important benefits such as greater transparency, liquidity, and significantly lower counterparty risk.

# <span style="color:#4A86C5;">4. Interest Rate Risk and Swap </span>
---



## <span style="color:#4A86C5;">4.1. Interest Rate Foundation</span>

---
### <span style="color:#5b9bd5;">4.1.1. The Reference Rates: The Fall of LIBOR and the New Era

#### The benchmark: Interbank Rate</span>
- Banks constantly lend and borrow money from each other overnight (or for short periods) to meet regulatory reserve requirements or manage daily liquidity. 

- The interest rate at which banks lend to one another is called the interbank rate.

- Because these are loans between massive, highly creditworthy institutions, interbank rates represent the wholesale "cost of money."

- They serve as the baseline reference rate for millions of financial contracts worldwide.

---

#### Historical LIBOR Movements During Crises</span>
The defining trend is that LIBOR (London Interbank Offered Rate) plummets during major financial crises. This happens because central banks (like the US Federal Reserve) aggressively cut base interest rates to make borrowing cheaper, stimulate the economy, and pump liquidity into struggling banks. For example:

- Early 90s Recession (1990–1992): Rates dropped steeply from around 9% down to 3% as the global economy contracted.

- Dot-Com Crash (2000–2003): Following the burst of the tech bubble and the aftermath of 9/11, rates plunged from over 6% to just above 1% to force economic recovery.

- Global Financial Crisis (2007–2009): After peaking near 5.5% during the housing boom, the 2008 crash triggered a massive, unprecedented drop. Rates flatlined near 0% as central banks desperately slashed rates to save the global banking system from collapse.

<center>
    <img src="../Images/Lecture_5/Pic6.png" alt="Pic6" width="900"> </center>
</center>


---



#### The Rise and Fall of LIBOR</span>
- For decades, the undisputed king of reference rates was LIBOR. 

- It was famously dubbed "the world's most important number," underpinning over $300 trillion in financial contracts globally.

**How LIBOR Worked:**
- Every morning, a panel of major global banks would submit an answer to a simple question:
*"At what rate could you borrow funds, were you to do so by asking for and then accepting inter-bank offers in a reasonable market size just prior to 11 am?"*

- The highest and lowest submissions were thrown out, and the remaining numbers were averaged to create the official daily LIBOR rates across multiple currencies and timeframes (e.g., 3-month USD LIBOR).

**Why is LIBOR dead?**
LIBOR was officially phased out (with the final USD rates ceasing in June 2023) due to two massive structural flaws:
- The Manipulation Scandal (2012): LIBOR was based on estimates, not actual transactions. During the 2008 financial crisis, and for years prior, traders at major banks realized they could collude to submit artificially high or low estimates. By moving LIBOR just a fraction of a percent, traders generated millions in illicit profits on their massive interest rate derivative portfolios.

- The Underlying Market Dried Up: Following the 2008 crisis, regulations changed, and banks largely stopped lending to each other on an unsecured, long-term basis. LIBOR was effectively measuring a market that no longer existed.

---


#### The Shift to Risk-Free Rates (RFRs)</span>
- To replace LIBOR, global regulators developed new benchmarks called Risk-Free Rates (RFRs).

- The critical difference between LIBOR and the new RFRs is that RFRs are based strictly on massive volumes of actual, observable, historical transactions, making them nearly impossible to manipulate.

### Primary Risk-Free Rate (RFR) Alternatives to LIBOR

| Alternative Reference Rate | Country / Region | Replaces | Key Characteristics & Description |
| :--- | :--- | :--- | :--- |
| **SOFR** <br>*(Secured Overnight Financing Rate)* | United States | USD LIBOR | • Measures the cost of borrowing cash overnight using U.S. Treasury securities as collateral (repos). <br>• Considered truly "risk-free" because it is secured (unlike LIBOR, which included a premium for bank credit risk). |
| **SONIA** <br>*(Sterling Over Night Index Average)* | United Kingdom | GBP LIBOR | • The primary overnight rate replacement for British pound sterling transactions. |
| **€STR** <br>*(Euro Short-Term Rate)* | Eurozone | EURIBOR / EONIA | • Reflects the wholesale euro unsecured overnight borrowing costs of banks in the euro area. |
| **TONAR** <br>*(Tokyo Overnight Average Rate)* | Japan | JPY LIBOR | • The primary overnight rate replacement for Japanese yen transactions. |

**A Structural Challenge: Forward-Looking vs. Backward-Looking**
LIBOR was forward-looking: You knew your 3-month interest rate at the beginning of the 3-month period.

SOFR is backward-looking: it is an overnight rate. To figure out the interest owed on a 3-month loan, you have to wait until the end of the 3 months, compound the daily overnight rates historically, and then you know what you owe.

(To solve this headache for corporate treasurers who need predictable cash flows, regulators have since developed "Term SOFR"—a forward-looking estimate based on SOFR futures derivatives).

---


### <span style="color:#5b9bd5;">4.1.2. The Cost of Debt</span>

The Cost of Debt is the effective IR that a company or a sovereign government pays on its borrowed funds. 

It is not just a single, arbitrary number; rather, it is priced by the markets based on two components:

$$\boxed{\text{Cost of Debt} = \text{Risk-Free Rate} + \text{Credit Risk Premium}}$$

**The Base Rate (Risk-Free Rate)**
This is the foundational cost of money in the economy, assuming zero risk of default. 
* *For corporations:* This is often a benchmark rate like **SOFR** or **SONIA**.
* *For governments:* This is the yield on the debt of the most stable country in the region (e.g., U.S. Treasury bonds or German Bunds).

**The Credit Risk Premium (The Spread)**
Investors demand extra compensation for taking on the risk that the borrower might default.
- The lower the borrower's credit rating, the higher the risk premium. 

#### Why the Cost of Debt Matters
* **For Corporations:** 
    - A rising cost of debt eats directly into profit margins and raises the "hurdle rate" for new investments. 
    - If borrowing becomes too expensive, companies halt expansion, hiring, and research.
* **For Governments:** 
    - A rising cost of debt can trigger a debt spiral.
        - If a country has to issue new bonds at 10% interest just to pay off old bonds that were at 3% interest, the national budget is quickly consumed by interest payments.

---

### <span style="color:#5b9bd5;">4.1.3. Sources of Interest Rate Risk</span>
IR risk is the exposure of a firm's financial condition to adverse movements in interest rates.

For MNCs, this risk is often intertwined with ER risk, but it requires its own distinct management strategies.

Major sources of IR risk include:
* **Monetary Policy Changes:** 
    * Central banks actively adjust short-term IRs to control economic growth and employment. 
    * Tightening monetary policy (raising rates) increases the cost of capital market-wide, while lowering rates makes borrowing cheaper.
* **Inflation Expectations:** 
    * Investors and lenders demand compensation for the loss of purchasing power over time. 
        * According to the Fisher Effect, if expected inflation rises, nominal IRs must also rise to maintain the real return for lenders.
* **Global Capital Flows:** 
    * Capital flows toward countries offering higher risk-adjusted IRs. 
    * Massive inflows or outflows of foreign investment can alter a country's domestic money supply and liquidity, forcing local IRs to shift regardless of domestic central bank policy.

---



### <span style="color:#5b9bd5;">4.1.4. Impact on Firms</span>

* **Cash Flow Risk:** 
    * Firms holding **floating-rate debt** (e.g., SOFR-linked loans) will see their interest expenses directly increase when market rates rise, straining immediate cash flows.

* **Investment Valuation (Discount Rate Risk):** 
    * Interest rates represent the "hurdle rate" or cost of capital for new projects. 
    * When rates rise, the discount rate applied to future cash flows increases, thereby lowering the **Net Present Value (NPV)** of corporate investments and making fewer projects viable.

* **Financing Decisions (Capital Structure):** 
    * Interest rate volatility forces CFOs to make strategic choices: 
        - Should they lock in fixed rates now, or ride floating rates hoping they go down? 
        - Should they issue short-term commercial paper or long-term bonds?
        * Misjudging the interest rate environment can lock a firm into uncompetitive, expensive debt for years.

---

## <span style="color:#4A86C5;">4.2. Managing IR Risks</span>
---


### <span style="color:#5b9bd5;">4.2.1. Forward Rate Agreements (FRAs)</span>
---

A Forward Rate Agreement (FRA) is a customized, OTC contract that allows a firm to lock in a guaranteed IR for a specific borrowing or lending period in the future. 

* **Long Position:** Locks in a borrowing rate. This protects against *IRs rising* before a planned loan.
* **Short Position:** Locks in an investment rate. This protects against *IRs falling* before depositing excess cash.

FRA is the primary financial derivative MNCs use to hedge IR exposure.

**FRA Terminology & Settlement**
- FRAs are often quoted by their timeline, such as a **"3x6 FRA"** (read as "three by six"). This means the underlying interest period starts in 3 months and ends in 6 months (a 3-month loan). 
- FRAs do not involve exchanging the actual principal. 
- Instead, they are settled in cash at the *beginning* of the borrowing period. Because the interest would normally be paid at the *end* of the loan, the FRA payoff must be discounted to its present value.

**FRA Payoff Formula (Long Position):**
$$\boxed{\text{Payoff} = \frac{\text{Notional} \times (\text{Market Rate} - \text{FRA Rate}) \times (\frac{\text{Days}}{360})}{1 + \left(\text{Market Rate} \times \frac{\text{Days}}{360}\right)}}$$

**Example:**
- A firm plans to borrow $10 million for 3 months, starting in 3 months (3x6 FRA).
- The FRA rate agreed today is 2.5% (annualized).
- In 3 months, the market rate for a 3-month loan is 3.0%.     
- The payoff to the long position is:
$$\text{Payoff} = \frac{10,000,000 \times (0.03 - 0.025) \times \frac{90}{360}}{1 + (0.03 \times \frac{90}{360})} \approx 12,407.41
$$ 

The firm receives approximately $12,407.41 in cash at the start of the loan period, which offsets the higher interest cost of borrowing at 3.0% instead of the locked-in 2.5%.


---


### <span style="color:#5b9bd5;">4.2.2. Interest Rate Futures</span>

- Unlike FRAs, which are customized OTC contracts, IR Futures (such as SOFR or Eurodollar futures) are standardized contracts traded on an exchange.
-  They serve the same fundamental purpose—locking in a future IR, but with different mechanics.

* Standardization: Fixed contract sizes (e.g., $1,000,000 notional) and specific maturity dates.
* Margin & Marking-to-Market: Buyers and sellers must post initial margin, and profits/losses are settled daily.
* Inverse Pricing: 
$$\boxed{\text{Price}_\text{IR Futures} = 100.00 - \text{Implied Interest Rate}}$$

* **Example:** 
     * If the expected interest rate is 4.50%, the futures price is 95.50.
     * If rates rise, the futures price falls. 
     * Therefore, to hedge against rising rates, a borrower must take a short position (sell futures).

---

### <span style="color:#5b9bd5;">4.2.3. Interest Rate Swaps (IRS)</span>


While FRAs and futures are useful for single, short-term future loans, **Interest Rate Swaps** are used to hedge long-term, ongoing IR exposure. A swap is an agreement between two parties to exchange a series of periodic IR cash flows based on a specified notional principal.

#### The Plain Vanilla Swap
The most common IRS involves exchanging a fixed rate for a floating rate (e.g., tied to SOFR or EURIBOR).

* **Paying Fixed / Receiving Floating:**
    - Used by a firm with floating-rate debt that expects IRs to rise. 
    - The floating payments received from the swap cover their floating loan payments, effectively converting the debt into a fixed-rate loan.
* **Paying Floating / Receiving Fixed:** 
    - Used by a firm with fixed-rate debt that expects IRs to fall.
     - This allows firms to benefit from cheaper borrowing costs.

**Example:** Hedging with a Swap</span>

- An MNC has a $10,000,000 bank loan with a floating IR of SOFR + 1%. The firm is worried that inflation will cause rates to rise, increasing their interest expense. 

- The firm enters into a plain vanilla IR swap with a dealer:
    * The firm pays a fixed rate of 4% to the dealer.
    * The dealer pays a floating rate of SOFR to the firm.

### <span style="color:#5b9bd5;">4.2.4. Cross-Currency IR Swaps</span>

---

- For MNCs, interest rate risk is frequently tied to foreign exchange risk. 
- A Cross-Currency Swap allows a firm to transform debt from one currency and interest rate structure into another.

    * Mechanism: Two parties exchange both the principal amount and the interest payments in two different currencies.
    * Use Case: A U.S. firm might issue cheap domestic dollar debt, but use a cross-currency swap to convert it into a euro-denominated liability to perfectly match the revenues generated by its European subsidiary, hedging both FX and interest rate risk simultaneously.

### <span style="color:#5b9bd5;">4.2.5. Interest Rate Options (Caps, Floors, and Collars)</span>
---
Unlike swaps or forwards that lock in a specific rate (creating symmetrical risk), interest rate options provide asymmetric protection, acting like insurance.

* **Interest Rate Cap (For Borrowers):** A series of call options on interest rates. It pays the buyer if the market rate exceeds a specified "strike rate." It limits maximum borrowing costs while allowing the firm to benefit if rates fall.
* **Interest Rate Floor (For Investors):** A series of put options on interest rates. It pays the buyer if the market rate falls below a specified strike rate, guaranteeing a minimum return on investments.
* **Interest Rate Collar:** A strategy where a firm buys a Cap and simultaneously sells a Floor. 
    * *Benefit:* The premium earned from selling the floor offsets the cost of buying the cap (often creating a "zero-cost collar").
    * *Trade-off:* The firm is protected from rising rates but sacrifices the benefit of falling rates below the floor level.

---

### <span style="color:#5b9bd5;">Summary: Comparing Hedging Strategies for a Floating-Rate Borrower</span>

| Strategy | Protection if Rates Rise | Benefit if Rates Fall | Upfront Cost | Best Used When... |
| :--- | :--- | :--- | :--- | :--- |
| **Do Nothing (Unhedged)** | None (Costs increase) | Maximum benefit | Zero | Firm expects rates to drop significantly. |
| **Interest Rate Swap** | Complete (Locks in rate) | None (Locked into higher rate) | Zero (Usually) | Firm wants total certainty of cash flows. |
| **Interest Rate Cap** | Complete (Capped max rate) | Maximum benefit | High (Premium) | Firm wants insurance but expects rates might fall. |
| **Interest Rate Collar** | Complete (Capped max rate) | Limited (Floored min rate) | Zero/Low | Firm wants free/cheap protection and accepts limited downside benefit. |

### <span style="color:#5b9bd5;">Case 2: Managing the Cost of Debt in the Eurozone Crisis</span>

How does a region deal with a skyrocketing cost of debt? The European Sovereign Debt Crisis (2010–2012) provides a masterclass in market panic and central bank intervention.

**The Greek Debt Spiral**
Following the GFC, markets realized Greece had massive, unsustainable debt levels. Investors panicked and demanded a massive **credit risk premium** to hold Greek bonds. 
* Greece's 10-year cost of debt spiked to nearly **30%** (while Germany’s was below 2%). 
* At 30% interest, it became mathematically impossible for Greece to refinance its maturing debt. They were locked out of the financial markets.

**The Solution: Regional Intervention**
Because Greece uses the Euro, it could not simply print its own money to pay off the debt. The region had to step in using three tools:

1. **Bailouts with Austerity:** The EU and the IMF provided emergency loans at below-market rates. In exchange, Greece was forced to implement severe "austerity" measures (slashing government spending, cutting pensions, and raising taxes) to prove they could balance their budget.
2. **"Whatever It Takes":** In 2012, European Central Bank (ECB) President Mario Draghi famously announced the ECB was ready to do "whatever it takes to preserve the euro." This verbal guarantee acted as an ultimate backstop, instantly calming investor panic.
3. **Quantitative Easing (QE):** The ECB became the buyer of last resort. They stepped into the open market and bought billions of euros worth of peripheral government bonds. By creating massive demand for these bonds, their prices went up, which forced their yields (the cost of debt) back down to manageable levels.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> A sovereign cost of debt crisis is rarely solved by the struggling country alone. It requires massive, coordinated intervention from central banks to artificially lower the risk premium and restore market confidence.

### <span style="color:#5b9bd5;"> Question 2:

1. FV of Premium: 
$$FV_\text{Premium} = \text{Premium} \times (1 + (r \times t))$$

$$FV_\text{Premium} = 0.03 \times (1 + (0.04 \times 0.5)) = 0.0306$$

### <span style="color:#5b9bd5;"> Question 3: 
---

**1. Constructing the hedge:**

Borrow the PV of the future receivable in euros, convert it to dollars at the spot rate, invest the dollars, and repay the euro loan when the receivable is received.

- 90-day EUR borrowing rate = 2% p.a. → 0.5% for 90 days
- Amount borrowed today:

$$
\frac{1{,}000{,}000}{1 + 0.02 \times \frac{90}{360}} = 995{,}024.88\ \text{EUR}
$$

- Convert to USD at spot rate $S_0 = 1.10$ USD/EUR:

$$
995{,}024.88 \times 1.10 = 1{,}094{,}527.36\ \text{USD}
$$

- Invest USD for 90 days at 1% p.a. → 0.25% for 90 days:

$$
1{,}094{,}527.36 \times \left(1 + 0.01 \times \frac{90}{360}\right) = 1{,}097{,}263.68\ \text{USD}
$$

When the exporter receives €1,000,000 in 90 days, it uses that amount to repay the EUR loan. Therefore, the hedge locks in about $1,097,263.68.

**2. Compare with the forward hedge:**

Using the forward contract:

$$
1{,}000{,}000 \times 1.0973 = 1{,}097{,}300\ \text{USD}
$$

So the forward hedge yields $1,097,300, while the money market hedge yields about $1,097,263.68. The difference is very small, which is what we expect under covered interest rate parity.

**3. Forward premium / discount:**

Since the forward rate is below the spot rate, the euro is selling at a forward discount.

$$
F = 1.0973 < S_0 = 1.10
$$


Annualized forward premium / discount:

$$
\frac{F - S_0}{S_0} \times \frac{360}{90}
= \frac{1.0973 - 1.10}{1.10} \times 4
\approx -0.98\%
$$

So the euro is at an annualized 0.98% forward discount, which is consistent with IRP because the euro IR is higher than the U.S. IR.

### <span style="color:#5b9bd5;"> Question 4: 
---
A plain-vanilla interest rate swap allows the firm to exchange floating-rate exposure for fixed-rate exposure.

The firm pays on its loan:

$$
\text{SOFR} + 1.5\%
$$

Under the swap, the firm:

- pays fixed 4.0%
- receives floating SOFR

The SOFR received from the swap offsets the SOFR paid on the loan.

So the net effective borrowing cost becomes:

$$
(\text{SOFR} + 1.5\%) + 4.0\% - \text{SOFR} = 5.5\%
$$

Therefore, the firm's effective interest cost is locked in at 5.5%.

**If SOFR rises to 6%:**

- Unhedged borrowing cost:

$$
6.0\% + 1.5\% = 7.5\%
$$

- Hedged borrowing cost:

$$
5.5\%
$$

So the firm saves:

$$
7.5\% - 5.5\% = 2.0\%
$$

On a $50,000,000 notional, that is:

$$
0.02 \times 50{,}000{,}000 = 1{,}000{,}000
$$

So the swap saves $1,000,000 per year.

**If SOFR falls to 2%:**

- Unhedged borrowing cost:

$$
2.0\% + 1.5\% = 3.5\%
$$

- Hedged borrowing cost remains:

$$
5.5\%
$$

So the opportunity cost is:

$$
5.5\% - 3.5\% = 2.0\%
$$

That is also:

$$
0.02 \times 50{,}000{,}000 = 1{,}000{,}000
$$

So the firm gives up $1,000,000 per year in savings when rates fall. The swap protects against rising rates, but sacrifices the benefit of falling rates.

### <span style="color:#5b9bd5;"> Question 5: 
---

The importer buys a European call option on €1,000,000 with:

- Strike price: $K = 1.12$/€
- Premium: $0.03$/€

The call option payoff is:

$$
\max(S_T - K, 0) \times 1{,}000{,}000
$$

**Scenario 1: $S_T = 1.18$/€**

The option is in the money.

$$
(1.18 - 1.12) \times 1{,}000{,}000 = 60{,}000
$$

So the option payoff is $60,000.

The effective cost per euro is:

$$
1.12 + 0.03 = 1.15
$$

So if the option is exercised, the importer effectively caps the cost at $1.15 per euro.

**Scenario 2: $S_T = 1.10$/€**

Since:

$$
1.10 < 1.12
$$

the option is out of the money, so payoff = 0.

The importer should not exercise the option. Instead, it buys euros directly in the spot market at $1.10/€ and loses only the premium paid.

So the total effective cost per euro becomes:

$$
1.10 + 0.03 = 1.13
$$

Thus, the option protects against euro appreciation but still lets the firm benefit if the euro depreciates.

### <span style="color:#5b9bd5;"> Question 6: 
---

The exporter buys a European put option on JPY 100,000,000 with:

- Strike price: $K = 0.0088$/JPY
- Premium: $0.0002$/JPY

The put payoff is:

$$
\max(K - S_T, 0) \times 100{,}000{,}000
$$

**Scenario 1:** $S_T = 0.0085$/JPY

The option is in the money.

$$
(0.0088 - 0.0085) \times 100{,}000{,}000 = 30{,}000
$$

So the payoff is $30,000.

Gross proceeds from selling yen at the strike price:

$$
100{,}000{,}000 \times 0.0088 = 880{,}000
$$

Premium paid:

$$
100{,}000{,}000 \times 0.0002 = 20{,}000
$$

Net USD proceeds:

$$
880{,}000 - 20{,}000 = 860{,}000
$$

So net proceeds are $860,000.

**Scenario 2:** $S_T = 0.0092$/JPY

Since:

$$
0.0092 > 0.0088
$$

the option is out of the money, so payoff = 0.

The exporter should not exercise the option and instead sell yen at the favorable spot rate.

Spot proceeds:

$$
100{,}000{,}000 \times 0.0092 = 920{,}000
$$

After subtracting the premium:

$$
920{,}000 - 20{,}000 = 900{,}000
$$

So net USD proceeds are $900,000.

Thus, the put option provides downside protection while still allowing the exporter to benefit if the yen strengthens.
