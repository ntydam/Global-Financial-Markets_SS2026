# 1. International Capital Markets
---


## <span style="color:#4A86C5;">1.1. Definition of International Capital Markets</span>
---

International capital markets are the markets through which governments, corporations, and financial institutions raise and invest long-term funds across borders.

They connect entities that need long-term financing with investors who are willing to provide capital in exchange for future returns.

Compared with domestic markets, international capital markets expands firms' financing options.

<center>
<img src="../Images/Lecture_8/Pic1.png" alt="International Capital Markets" width="850">
</center>
<br></br>

### <span style="color:#4A86C5;">Why they matter</span>

Firms use international capital markets to:

- access **larger amounts of capital**
- obtain **better financing terms**
- reach a **broader base of investors and lenders**
- support **global expansion**

This is especially important when the home market is:
- too small
- illiquid
- segmented
- unable to support large debt or equity issues

### <span style="color:#4A86C5;">Why they matter for MNCs</span>

For MNCs, access to international capital markets allow firms to:
- finance large-scale projects  
- diversify sources of capital  
- match funding structures with business strategy and risk exposure  

**Example**
- An Indian MNC with revenues in USD and EUR may prefer to raise funds in EUR and USD rather than rely only on domestic financing.

---

## <span style="color:#4A86C5;">1.2. Debt and equity
---
There are 3 sources of capital for a business: Retained Earnings, Debt, and Equity.

Two main sources of long-term finance:
- **Debt**
    - Firms can borrow by issuing bonds. 
    - In return, they commit to:
        - periodic interest payments
        - principal repayment at maturity

- **Equity**
    - Firms can raise capital by issuing shares.
    - Equity gives investors an ownership claim on the firm’s future profits.

---



## <span style="color:#4A86C5;">1.3. Domestic vs international capital markets
---

**Domestic capital markets**
- funding is raised mainly from local investors  
- securities are issued under domestic regulations  
- financing options may be limited in smaller or less liquid markets  

**International capital markets**
- funding can be raised from foreign as well as domestic investors  
- firms can access larger and more diversified capital pools  
- financing decisions must consider exchange-rate, political, and regulatory risk  

---

## <span style="color:#4A86C5;">1.4. Main Participants
---

**Corporate issuers**
- raise long-term funds in international capital markets through debt or equity
- use raised funds to finance:
    - expansion, 
    - acquisitions, 
    - infrastructure (e.g. technology), and 
    - other long-term investment projects

**Institutional investors**
- are large organizations that manage large pools of savings
    - Pension funds
    - Insurance companies
    - Mutual funds and asset managers
- determine the demand for international bonds and equities
    - supply funds in capital markets by purchasing securities issued by firms and governments.


**Financial intermediaries and exchanges**
- make international capital markets function more efficiently.
- connect issuers with investors, such as:
    - Investment banks arrange and underwrite securities
    - Stock exchanges and bond markets provide trading platforms
    - Brokers and dealers facilitate market transactions
    - Valuation provides pricing for securities

# 2. International Bond Markets
---


## <span style="color:#4A86C5;">2.1. Types of International Bonds
---
International bond markets allow firms and governments to raise long-term debt across borders. 

- In most years, Eurobonds account for around 80% of new international bond issues, while foreign bonds make up a smaller share.

<center>
<img src="../Images/Lecture_8/Pic3.png" alt="Bonds" width="850">
</center>
<br></br>

- **Eurobonds**:
    - Euromarkets originally developed in Europe, they now operate all over the world.
    - In FX context, “Euro” means outside the domestic market of the currency involved.
    - Eurobonds are usually identified by the *currency in which they are issued*, such as:
        - U.S. dollar Eurobonds
        - yen Eurobonds
    - **Example:** An Australian company issues bonds in Singapore, but the bonds are denominated in USD.
    

- **Foreign bonds** are often identified by *special names* based on the country where they are sold.
    - Yankee bonds: U.S. dollar-denominated foreign bonds issued in the United States  
    - Samurai bonds: yen-denominated foreign bonds issued in Japan  
    - Bulldog bonds: pound sterling-denominated foreign bonds issued in the United Kingdom  

---

## <span style="color:#4A86C5;">2.2. Bond Market Structures and Bond Features
---

### <span style="color:#5b9bd5;">2.2.1. Bond Market Structures
If a foreign borrower raises funds in an international financial market, it must usually follow the laws and regulations of that market (unless another legal system is applied as the governing law).
- **Euro markets** are more self-regulated.
- **Foreign bond** are subject to the **domestic regulation** of the country where the bonds are issued.

Two major international debt market groups are often examined:

- The Euromarkets, which include:
    - the Eurocurrency market
    - the Euronote market
    - the Eurobond market

- The U.S. debt markets, which include:
    - the commercial paper market
    - the medium-term note market
    - the bond market

### <span style="color:#5b9bd5;">2.2.2. Bond Features
International bond issues differ across markets, but three important features are:

- the **currency of issue**
    - is the currency in which interest and principal are paid
    - affects ER exposure for both the issuer and the investor
    - Firms often issue/borrow in currencies that match their operating CFs.

- the **maturity of the bond**
    - is the time until the principal is repaid:
        - Some firms prefer shorter maturities to reduce long-term commitment
        - Others prefer longer maturities to secure stable funding

- the **relationship between credit rating and yield**
    - A bond’s credit rating reflects the issuer’s ability to meet its debt obligations.
        - Higher-rated bonds usually offer **lower yields**.
    - Major international credit rating agencies include:
        - Fitch
        - Moody’s
        - S&P Global

### <span style="color:#5b9bd5;">2.2.3. More Sophisticated Features
Some bonds also contain:
- **Call protection**
    - prevents early redemption for an initial period
    - helps protect investors from the bond being refinanced too quickly at lower IRs

- **Sinking fund provisions**
    - require the issuer to retire part of the bond issue periodically
    - or set aside funds over time to help repay the principal at maturity

---

## <span style="color:#4A86C5;">2.3. Why Firms Issue Bonds Internationally
---

Firms issue bonds internationally to improve financing flexibility and to obtain funding on terms that may be better than those available in the domestic market.

**Lower borrowing cost**

International markets may allow firms to borrow at a lower cost, especially if the home market is small, illiquid, or expensive, because of:
- better market conditions
- more competitive pricings

**Broader investor access**

Issuing bonds internationally gives firms access to a wider range of investors:
- increase demand for the bond issue
- improve the firm’s ability to raise large amounts of capital.

**Matching debt currency with cash flows**

Firms may issue bonds in the same currency as their expected revenues or costs to reduce ER risk.

# 3. International Equity Markets
---


## <span style="color:#4A86C5;">3.1 Background</span>
---

### <span style="color:#5b9bd5;">3.1.1 Shares, Stocks, and Equity</span>
- A **share (stock)** represents a unit of equity ownership in a company.
- Shareholders may earn returns through **dividends** and **capital gains**.
- A share price reflects investors’ expectations about the **present value (PV) of future cash flows (CFs)**.
- Shares are commonly classified as **common shares** or **preferred shares**.

$$\boxed{\text{Market Capitalisation} = \text{Share Price} \times \text{Number of Shares Outstanding}}$$

- The **equity market** enables firms to raise **long-term capital** from investors.

### <span style="color:#5b9bd5;">3.1.2 International Equity Markets</span>
- International equity markets allow firms to raise funds beyond their home market.
- Key benefits include:
    - access to a larger pool of capital
    - a more diversified investor base
    - stronger international visibility and credibility

---


### <span style="color:#5b9bd5;">3.1.3. Key Trends in Global Equity Markets</span>
- A stock exchange provides the platform and rules for trading listed shares.

- Equity trading has become increasingly globalised, making it important to understand how major stock exchanges operate.

- Global market capitalisation of listed companies reached about **US$114.5 trillion** in 2024 (LSEG, 2024).

- China is now the world’s second-largest equity market, matching its position as the second-largest economy.

- More than 80% of stock trading is now automated.


---

## <span style="color:#4A86C5;">3.2. Foreign Issuance vs Cross Listing
---
### <span style="color:#5b9bd5;">3.2.1. Cross-Listing</span>

Cross-listing:
- occurs when a firm lists its shares on **one or more foreign stock exchanges** in addition to its home market.
- increases market access, visibility, and liquidity
- broadens the investor base

Funding effect:
- Does not necessarily raise new capital
- Can facilitate future equity issuance at better terms

> <span style="color:#4A86C5;">**Key intuition:**</span>
> Cross-listing can improve liquidity and investor perception, especially for firms from countries with weaker governance, although these benefits may diminish over time.

---


### <span style="color:#5b9bd5;">3.2.2. Foreign Equity Issuance / Direct Listing</span>


Foreign equity issuance:
- Issuing and selling new shares in a foreign market
- Objective: raise new equity capital

Direct listing:
- A firm lists existing shares on an exchange without issuing new shares
- No capital raised at listing; 
- Mainly for liquidity and visibility

> <span style="color:#9e86C5;"><strong>**Question**:</strong></span><br>
> **When is cross-listing more strategic than immediate foreign equity issuance?**

> <span style="color:#9e86C5;"><strong>**Answer**:</strong></span><br>
Cross-listing is preferred:
- Reducing information asymmetry - listing in a major market (e.g., U.S.)
- Enhancing firm visibility and reputation - important for firms from EMs
- Achieving better valuation - reduce segmentation effects
- Testing foreign market conditions
- Lowering future cost of capital

---

## <span style="color:#4A86C5;">3.3. American Depository Receipts & Global Registered Shares
---

### <span style="color:#5b9bd5;">3.3.1. American Depository Receipts (ADRs)
- An ADR is a U.S.-traded certificate representing shares of a foreign company held by a custodian in the home market.
- ADRs make foreign stocks easier for U.S. investors to buy and trade.
- First ADRs began trading in 1927.

Types of ADRs
- Sponsored ADRs: created with the foreign firm’s approval; can be listed on U.S. exchanges.
- Unsponsored ADRs: created without direct issuer involvement; disclosure may be more limited.

---


### <span style="color:#5b9bd5;">3.3.2. Global Registered Shares (GRSs)
- A GRS is a single class of share:
    - registered in multiple markets and 
    - traded in multiple currencies
- Unlike ADRs, GRSs are ordinary shares, not receipts representing foreign shares.
- GRSs are fully fungible: shares bought on one exchange can be sold on another.

Pros and Cons:
- Equal status for all shareholders
- Direct voting rights
- Wider global investor access
- More costly and complex to establish global registration and clearing systems

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - ADRs improve foreign access mainly for U.S. investors
> - GRSs create a more integrated global share structure, but at higher cost


> <span style="color:#9e86C5;">**Question:**</span><br>
> Why might a firm prefer foreign-currency debt even when domestic debt is available?

---

> <span style="color:#9e86C5;">**Answer:**</span><br>
A firm may prefer foreign-currency debt because it:
> - **matches the currency of operating cash flows** (natural hedge against FX risk),
> - may be **cheaper** when foreign interest rates are lower or foreign investor demand is stronger,
> - **diversifies** the firm's funding base,
> - opens access to **deeper or more liquid** markets (e.g., USD or EUR Eurobond markets),
> - signals a **commitment to international markets** and improves global brand visibility.

---


## <span style="color:#4A86C5;">3.4. Why Firms Raise Equity Internationally</span>
---

Firms may raise equity internationally or cross-list their shares to:

- increase demand for their shares and improve liquidity.
- strengthen visibility and reputation in global capital markets.
- raise the firm’s profile among foreign investors and consumers.
- signal stronger corporate governance and transparency.
- improve valuation and support future capital raising.
- make hostile takeovers more difficult in some cases.

---

## <span style="color:#4A86C5;">3.5. Factors Affecting International Equity Returns</span>
---

International equity returns are influenced by several key factors:

**Macroeconomic factors**
- Domestic variables (e.g., GDP growth, inflation, interest rates) tend to have stronger influence on equity returns than international monetary variables.  
- Empirical evidence shows that macroeconomic variables explain only a limited portion of equity return variation, while broader market indices often explain more.

**Exchange rates**
- Exchange rate movements can affect returns, especially for international investors.  
- Evidence suggests:
  - Exchange rate changes explain more variation in foreign bond returns than in equity returns.
  - The correlation between exchange rates and stock markets is generally low but positive.
  - Exchange rate risk is often not fully priced in equity markets.

**Industrial structure**
- Differences in industry composition across countries affect returns.  
- Countries with higher exposure to certain sectors (e.g., tech, energy, manufacturing) may experience different performance due to sector-specific shocks.

---

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - International equity returns are driven more by local economic conditions and industry structure, while exchange rates and global macro factors play a more limited or indirect role.

# 4. Market Access, Liquidity, and Cost of Capital
---


## <span style="color:#4A86C5;">4.1. Methods of International Security Issuance</span>
---

Security issuance = when a firm raises money by selling stocks or bonds to investors in different markets.

| Method | Definition | Key idea | Example |
|---|---|---|---|
| **Domestic issue** | Issued in the firm’s home country, in local currency, under local rules | Simple, but limited to domestic investors | Australian firm issuing AUD bonds in Australia |
| **Foreign issue** | Issued in a foreign country, in that country’s currency, under its rules | Access to foreign investors and currency | Japanese firm issuing USD bonds in the U.S. |
| *Offshore issue** | Issued outside the home country, often in international markets | Flexible, global investor access | Brazilian firm issuing USD bonds in London |

<br>

> <span style="color:#4A86C5;">**Key intuition:**</span>
> - Domestic → home market
> - Foreign → specific foreign market
> - Offshore → international market (more flexible)

---

## <span style="color:#4A86C5;">4.2. Liquidity and Market Segmentation
---

- **Liquidity**:
    - refers to how easily securities can be issued or traded without causing a large change in price.
    - helps firms raise funds more easily because investors value:
        - faster trading
        - lower transaction costs
        - less price disruption

- **Market segmentation**
    - occurs when securities are priced mainly by **local conditions** rather than global market forces.
    - caused by barriers that prevent full integration with global markets, such as:
        - capital flow restrictions
        - weak transparency
        - high transaction costs
        - ER risk
        - political or legal barriers
        - governance differences
    - In segmented markets, investors demand a **higher required return**, leading to:
        - a higher cost of capital
        - possible undervaluation of securities
        - more limited access to financing

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> A firm can be strong, but poor market quality makes financing more costly.

---

## <span style="color:#4A86C5;">4.3. Implications of Global Market Access
---
**Lower Financing Cost**
- Global market access reduces financing costs because firms can choose:
    - where to raise capital  
    - which currency to issue in  
    - which investor base to target  

**Greater Financial Resilience**
- Access to multiple markets improves resilience: 
    - If one market becomes temporarily expensive or inaccessible because of political shock, monetary tightening, or a local financial crisis, the firm may still raise funds elsewhere.

**Enhanced Competitiveness**
- Firms with better access to capital can:
    - fund more projects
    - respond more quickly to opportunities
    - survive downturns, and 
    - compete more aggressively

# 5. Risks and the Role of International Capital Markets for MNCs
---



## <span style="color:#4A86C5;">5.1. Major Risks
---

**Interest-rate risk**
- Changes in IRs alter:
    - the cost of new financing, 
    - the value of existing fixed-income securities
    - the cash burden of floating-rate debt

**Exchange-rate risk**
- ER risk arises when the currency of financing differs from the currency of operating CFs.
    - A depreciation of the operating currency relative to the debt currency increases the real burden of debt service. 
- Currency matching and hedging are central to international financing strategy.

**Regulatory and country risk**
- Country risk can affect investor demand and the firm’s ability to access markets and transfer funds.

- International financing also exposes firms to differences in:
    - securities regulation, 
    - accounting rules, 
    - tax treatment, 
    - capital controls, and
    - political conditions. 

---



## <span style="color:#4A86C5;">5.2. Barriers to International Market Access
---
**Listing and compliance costs**
- Accessing foreign markets requires substantial legal, accounting, governance, and disclosure expenditure. 
- These costs may be justified for large issuers but can be burdensome for smaller firms.

**Legal and accounting differences**
- Differences in disclosure standards, governance expectations, accounting frameworks, and investor protection rules can complicate cross-border issuance and affect investor confidence.

**Limited access for smaller firms**
- Smaller firms may struggle to achieve sufficient scale, credit standing, analyst coverage, or governance credibility to attract global investors at reasonable cost.
- In practice, international market access is often easier for larger, more visible firms.

---



## <span style="color:#4A86C5;">5.3. How International Capital Markets Serve MNCs
---

- If a MNC operates globally but uses only domestic financing, it may face unnecessary mismatches in:
    - currency
    - maturity
    - investor base

- International capital markets are closely linked to **investment decisions**:
    - which projects the firm can afford to undertake
    - what discount rate should be used
    - how much debt the firm can safely take on
    - how foreign expansion should be financed

- This is why **financing strategy** and **FDI strategy** should be analyzed together.

# 6. Foreign Direct Investment (FDI)
---

## <span style="color:#4A86C5;"> 6.1 Nature of FDI
---

<div style="display: flex; gap: 16px;">

<div style="flex: 1; border: 2px solid #1E90FF; padding: 12px; border-radius: 8px; background-color: #F0F8FF;">
<b>Portfolio Investment</b><br>
Involves holding foreign financial assets primarily for financial return, without significant influence over business operations.
<br><br>
<b>Ways:</b>
<ul>
  <li>Purchasing foreign stocks</li>
  <li>Buying foreign bonds</li>
  <li>Investing in mutual funds or ETFs with international exposure</li>
  <li>Holding foreign currency deposits</li>
  <li>Investing in derivatives on foreign assets</li>
</ul>
</div>

<div style="flex: 1; border: 2px solid #1E90FF; padding: 12px; border-radius: 8px; background-color: #F0F8FF;">
<b>FDI</b><br>
Occurs when the investor acquires a measure of control of the foreign business.
<br><br>
<b>Ways:</b>
<ul>
  <li>establishing new operations</li>
  <li>acquiring existing firms</li>
  <li>taking a substantial stake in a foreign enterprise</li>
</ul>
</div>

</div>

- FDI:
    - is more than the purchase of foreign securities
    - is a strategic commitment to productive assets abroad.
    - gives the investor greater control and potentially stronger long-term returns
    - also increases exposure to host-country risks

- Control is central to FDI because the investor can shape decisions on production, marketing, staffing, supply chains, pricing, and reinvestment.

- Compared with exporting or licensing, FDI allows firms to enter foreign markets more deeply by:
    - creating a physical presence
    - improving responsiveness to customers
    - reducing logistics costs
    - securing access to resources, and 
    - strengthening long-term competitive position.

- **Example:**
    - *FDI into the United States:* Nestlé, a Swiss multinational company, purchased Carnation, an American firm.
    - *FDI from the United States:*
      - Coca-Cola established bottling plants in many countries worldwide.
      - More recently, many American corporations have relocated production operations to places like Mexico and China to benefit partly from lower manufacturing costs.
    - In general, FDIs occur when businesses seek to exploit market imperfections, such as cheap labor and safe markets.



---

## <span style="color:#4A86C5;"> 6.2 Modes of FDI
---

### <span style="color:#4A86C5;">6.2.1. Greenfield investment
- The firm builds a new business operation from the beginning.
- This gives the firm **high control** over:
  - technology
  - company culture
  - production processes
  - plant and facility design
- Main drawback:
  - takes more time
  - needs high upfront investment

---

### <span style="color:#4A86C5;">6.2.2. Joint ventures and strategic alliances
- The firm works with a local or foreign partner.
- Partners share ownership, cost, and risk.
- This is useful when:
  - the market is uncertain
  - local knowledge is important
  - political connections matter
  - regulations make partnership helpful
- Main drawback:
  - the firm has less than full control

---


### <span style="color:#4A86C5;">6.2.3. Mergers and acquisitions (M&A)
- The firm enters a foreign market by **buying an existing company**.
- Main advantages:
  - faster market entry
  - immediate market share
  - access to local knowledge
  - existing employees
  - established distribution channels
  - existing licenses and permits
- Main drawback:
  - integration can be difficult and risky (negative synergy)

## <span style="color:#4A86C5;"> 6.3 Alternative modes to FDI
---

| Entry Mode | Description | Advantages | Disadvantages / Limits | Most Appropriate When |
| --- | --- | --- | --- | --- |
| **Exporting** | The firm sells to foreign customers without owning production assets abroad. | - lower commitment  <br> - lower risk | - less local control  <br> - weaker presence | - the firm wants foreign sales without major capital commitment  <br> - market uncertainty is still high |
| **Licensing / Franchising** | The firm allows a foreign partner to use its brand, technology, or business model, while the local partner handles operations. | - fast expansion  <br> - limited capital needed | - less control over operations  <br> - possible quality problems  <br> - risk of technology or brand misuse | - control is less important  <br> - the firm wants growth with lower investment |

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> - FDI is not always the best option.
> - Firms choose FDI when:
>   - the benefits of control are high
>   - the long-term gains are worth the extra cost and risk

# 7. Motives, Determinants, Benefits, and Limitations of FDI
---


## <span style="color:#4A86C5;">7.1 Why Firms Undertake FDI</span>
---

### <span style="color:#4A86C5;">7.1.1. Why foreign projects may be attractive</span>
Firms undertake FDI when the expected return on investing abroad exceeds the cost of capital after adjusting for:

- ER risk
- political risk
- operating risk

**Example: Japanese FDI in the United States**
- Japanese FDI in the United States rose sharply in the second half of the 1980s.
- One reason was the strong appreciation of the yen against the U.S. dollar.
- A stronger yen increased the purchasing power of Japanese firms, making U.S. assets relatively cheaper in yen terms.

---


### <span style="color:#4A86C5;">Three main motives for FDI</span>

1. **Market-seeking FDI**</span>
- Firms invest abroad to expand sales and establish a direct presence in foreign markets.
- Producing inside the target country can help firms:
    - serve customers more effectively
    - avoid tariffs, quotas, and other trade barriers
    - reduce transportation and distribution costs
    - adapt products and services to local preferences

2. **Resource- and efficiency-seeking FDI**</span>
- Firms invest abroad to gain access to inputs or production conditions that are more favorable than those available at home.
- These may include:
    - cheaper labor
    - raw materials or natural resources
    - lower-cost intermediate goods
    - more efficient production or logistics networks

3. **Strategic-seeking FDI**</span>
- Some FDI is motivated by long-term strategic goals rather than immediate sales or cost advantages.
- Firms may invest abroad to:
    - acquire technology or managerial expertise
    - secure supply chains
    - establish regional production or distribution hubs
    - respond to competitors’ global expansion
    - strengthen long-term international competitiveness

---

## <span style="color:#4A86C5;">7.2. Determinants of FDI</span>
---

FDI is not determined by host-country attractiveness alone.  
Firms invest abroad when three conditions are present:

This means FDI is a **two-level decision**:

- First, the firm must ask whether it has something valuable to take abroad.

---


### <span style="color:#4A86C5;">7.2.1. Ownership Advantages</span>

A firm will not invest abroad unless it has some **competitive advantage** that can be transferred internationally
- sustainable
- difficult to copy
- transferable across borders

These advantages may include:
- technology
- managerial expertise
- brand reputation
- patents
- production know-how

Without such an advantage, the firm has little reason to undertake the cost and complexity of FDI.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> FDI begins with the **firm**, not the country.  
> Even an attractive host market may not receive investment from a company that has no real competitive edge abroad.

---


### <span style="color:#4A86C5;">7.2.2. Location Advantages</span>

Once the firm has a transferable advantage, it evaluates whether a foreign country offers strong enough **location advantages** to make FDI worthwhile.

These advantages may arise from:
- **market size & growth potential**
    - Large or fast-growing markets are attractive because they allow firms to spread the fixed cost of entry over a larger base of sales.
    - If the local market is large enough, producing inside the country may be more profitable than exporting into it.

- **low-cost but productive labor**
- **access to raw materials**
- **technological clusters**
- **supportive government policy**
    - Government policy can strengthen a country’s location advantage.
    - Important policy-related determinants include:
        - tax policy
        - investment incentives
        - regulatory predictability
        - administrative efficiency
        - openness to foreign ownership

- **efficient infrastructure**
    - transport networks
    - ports
    - electricity supply
    - supplier networks
    - production reliability

---

**Examples:** 

1. **Tesla in China - Market Size & Growth Potential**
    - Tesla invested in **Giga Shanghai** because China offered:
        - very large market size
        - rapid growth in EV demand
        - strong long-term sales potential
    - Instead of relying only on exports, Tesla established local production to compete directly in one of the world’s most important EV markets.

2. **Samsung in VietnamL Low cost labor & adequate infrastructure**
    - Samsung’s expansion in Vietnam reflects efficiency-seeking FDI.
    - Vietnam offered:
        - labor-cost competitiveness
        - export-oriented production capacity
        - improving industrial coordination
        - adequate infrastructure for electronics manufacturing
    - Samsung did not invest only because labor was cheaper, but because Vietnam combined cost advantages with the ability to support large-scale manufacturing.


3. **Intel in Ireland: Supportive government policy**
    - Intel’s investment in Ireland reflects:
        - a pro-business environment
        - tax advantages
        - government support
        - a strong base for high-tech activity
    - Public policy reinforced Ireland’s location advantage.

---


### <span style="color:#4A86C5;">7.2.3. Internalization Advantages</span>

Even if a firm has ownership advantages and finds an attractive location, it still must decide whether to enter through:

- **FDI**
- exporting
- licensing
- franchising
- management contracts

Firms often choose to **internalize** operations in order to:
- protect proprietary knowledge
- maintain product quality
- preserve managerial control
- reduce transaction costs
- avoid weak contract enforcement

This means the real question is not only:

***“Why invest in this country?”***

but also:
***“Why invest directly instead of using another entry mode?”***

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> If the firm fears imitation, leakage of technology, or weak legal enforcement, direct ownership becomes more attractive than licensing.

---


### <span style="color:#4A86C5;">7.2.4. Bringing the Determinants Together</span>
---

The determinants of FDI should not be treated as a simple checklist.

A firm invests abroad only when the following conditions hold together:

- it has a **firm-specific advantage**
- the host country offers strong **location advantages**
- direct ownership creates more value than alternative entry modes
- the expected return is high enough to compensate for **political and financial risk**

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> FDI is ultimately a **risk-return decision**.  
> A firm undertakes FDI only when its own competitive strengths, the host country’s advantages, and the value of direct control together outweigh the risks of investing abroad.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> **So the real FDI questions are:**
> - Does this firm have an advantage worth exploiting abroad?
> - Is this location suitable for that advantage?
> - Is direct ownership better than exporting or licensing?
> - Are the expected returns high enough to justify the risks?

---

## <span style="color:#4A86C5;">7.3 Benefits and Limitations of FDI</span>

---

FDI offers important strategic benefits, but it also involves significant costs and risks.

<div style="display: flex; gap: 16px;">

<div style="flex: 1; border: 2px solid #4A86C5; padding: 14px; border-radius: 8px; background-color: #F4F9FF;">
<b>Benefits of FDI</b>
<ul>
  <li><b>Growth and diversification:</b> FDI can expand the customer base and diversify CFs across countries.</li>
  <li><b>Greater control:</b> The firm has more control over production, quality, technology, and strategy than under exporting or licensing.</li>
  <li><b>Stronger presence:</b> FDI can strengthen long-term competitiveness in foreign markets.</li>
</ul>
</div>

<div style="flex: 1; border: 2px solid #4A86C5; padding: 14px; border-radius: 8px; background-color: #F4F9FF;">
<b>Limitations of FDI</b>
<ul>
  <li><b>High commitment:</b> FDI requires substantial capital, time, and managerial resources.</li>
  <li><b>Lower flexibility:</b> Fixed assets abroad are difficult to reverse or relocate.</li>
  <li><b>Greater exposure to country risk:</b> Firms face political, regulatory, ER, and remittance risk in the host country.</li>
</ul>
</div>

</div>

<br>

> **Key trade-off:** FDI can offer higher control and stronger long-term returns, but it also requires greater commitment and creates greater exposure to host-country risk. As a result, FDI decisions should be evaluated together with country risk analysis.

**Example:**  
Suppose an electronics producer compares two host countries:
- Country A has a larger market but strict profit-remittance controls.
- Country B has a smaller market but more stable regulation and better logistics.

The better investment choice depends on both expected operating return and country risk.

# 8. Risks in FDI and Country Risk Analysis

---


## <span style="color:#4A86C5;">8.1 Main Risks in FDI</span>
---
FDI exposes firms to risks that can reduce project CFs, limit managerial control, delay remittances, or weaken the overall value of the investment.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> A foreign project may appear attractive in terms of market size, costs, or resources, but still fail if the risks are large enough to reduce, delay, or block the expected returns.

<div style="border: 2px solid #4A86C5; border-radius: 10px; padding: 15px; margin-bottom: 15px; background-color: #EEF5FF;">
<b>Political & Regulatory Risk</b><br><br>

- Government actions affect operations or profit repatriation  
- Includes:
  - regulatory changes  
  - blocked funds / transfer restrictions  
  - expropriation / nationalization  
  - legal uncertainty  

</div>

---

<div style="border: 2px solid #2E8B57; border-radius: 10px; padding: 15px; margin-bottom: 15px; background-color: #EAF7F0;">
<b>Economic & Financial Risk</b><br><br>

- Macroeconomic conditions affect project value  
- Includes:
  - inflation  
  - exchange-rate volatility  
  - interest-rate instability  
  - recession / financial crises  

<i>Reduces real returns and affects financing conditions
</div>

---

<div style="border: 2px solid #C58F2E; border-radius: 10px; padding: 15px; margin-bottom: 15px; background-color: #FFF6E6;">
<b>Business & Operational Risk</b><br><br>

- Day-to-day execution risks in foreign markets  
- Includes:
  - demand uncertainty  
  - supply disruptions  
  - labor issues  
  - cultural and governance differences  

<i>Even with stable macro conditions, poor execution can destroy value
</div>

---

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> FDI risk comes from three layers:  
> **country risk (political), macro risk (economic), and firm-level execution (operational).**

---

## <span style="color:#4A86C5;">8.2. Country Risk</span>
---

Country risk refers to the possibility that conditions in the host country will reduce the expected return on a foreign investment.

Even if the project itself looks attractive, the country environment may still prevent the firm from realizing its expected value.


| Country-risk question | Why it matters |
| --- | --- |
| Can profits be repatriated freely? | Parent-company cash flow may be delayed, reduced, or trapped. |
| How stable is the ER? | Depreciation can reduce translated profits and raise debt burden. |
| How predictable is regulation? | Rule changes can affect taxes, pricing, ownership, and project viability. |
| How resilient is the banking system? | Financial stress can disrupt borrowing, working capital, and customer demand. |

<br>

---



### <span style="color:#4A86C5;">8.2.1. Why country risk matters</span>
---

Country risk affects three core parts of the investment decision:

<div style="border: 2px solid #4A86C5; border-radius: 10px; padding: 14px; margin-bottom: 12px; background-color: #EEF5FF;">
<b>1. Expected cash flows</b><br><br>
Country conditions can affect:
<ul>
<li>taxes</li>
<li>inflation</li>
<li>regulation</li>
<li>demand conditions</li>
<li>supply reliability</li>
</ul>
</div>

<div style="border: 2px solid #2E8B57; border-radius: 10px; padding: 14px; margin-bottom: 12px; background-color: #EAF7F0;">
<b>2. Investment feasibility</b><br><br>
Country conditions can affect:
<ul>
<li>ownership rules</li>
<li>licensing conditions</li>
<li>market access</li>
<li>operating freedom</li>
</ul>
</div>

<div style="border: 2px solid #C58F2E; border-radius: 10px; padding: 14px; margin-bottom: 12px; background-color: #FFF6E6;">
<b>3. Financing and profit repatriation</b><br><br>
Country conditions can affect:
<ul>
<li>exchange controls</li>
<li>sovereign risk</li>
<li>banking conditions</li>
<li>the ability to move profits back to the parent</li>
</ul>
</div>

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> A good project in a bad country can still be a bad investment.
---


### <span style="color:#4A86C5;">8.2.2. How Firms Assess Country Risk</span>
---

Firms usually assess country risk using three complementary approaches:

- **Qualitative assessment**
    - political context
    - policy direction
    - institutional quality
    - local operating constraints

- **Quantitative indicators**
    - sovereign spreads
    - inflation
    - debt ratios
    - reserve levels
    - ER volatility
    - country-risk ratings

- **Scenario analysis**
    - compares base, downside, and stress cases
    - tests how country shocks affect project value

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> Country risk should not be judged using only one forecast.  
> Firms usually combine qualitative judgment, quantitative indicators, and scenario analysis.

> <span style="color:#9e86C5;">**Question:**</span><br>
> 1. **How should the firm evaluate the possibility that profits cannot be repatriated immediately?**  
> 2. **How can country risk reduce project NPV through both cash flows and the discount rate?**

---

> <span style="color:#9e86C5;">**Answer:**</span><br>
> 1. The firm can model **scenarios with delayed remittance** (e.g., funds blocked for 1, 2, or 3 years) and discount the cash flows accordingly. The longer the delay, the larger the present-value loss. The firm can also assume some funds become **permanently blocked** (haircut to zero) under stress scenarios. Alternatively, lower the project's expected CFs by a "remittance haircut" reflecting expected blocked-funds probability.
>
> 2. **Country risk reduces project NPV through both channels:**
>    - **Cash-flow channel:** lower expected operating CFs (taxes, controls, regulatory burden, demand shocks), and lower remittable share of those CFs.
>    - **Discount-rate channel:** higher required return ($k$ rises by a country-risk premium), which reduces PV of all future CFs.
>    - These should not be applied **simultaneously to the same risk** — that double-counts the cost of country risk.

# 9. FDI Decisions
---


## <span style="color:#4A86C5;"> 9.1 Integrating risk into FDI decisions
---
Step 1: Comparing expected return with country risk

Managers should not evaluate foreign projects using operating return alone. They must compare expected return with the incremental country risk borne by the firm. Projects in higher-risk countries may require stronger margins, smaller scale, staged commitment, or different financing structures.

Step 2: Entry mode choice under uncertainty

Entry mode itself is a risk-management choice. Greenfield projects offer high control but high commitment. Acquisitions provide speed but integration risk. Joint ventures can share cost and local uncertainty. Exporting and licensing may be better if strategic control is not essential.

Step 3: Timing and scale of FDI

Managers can manage uncertainty by staging investment over time. Rather than committing immediately to full scale, a firm can enter gradually, gather information, and expand only if performance and country conditions justify further commitment.

---


## <span style="color:#4A86C5;"> 9.2 Managing and mitigating FDI risk
---
* **Diversification across countries:** avoid excessive dependence on a single host market.
* **Partnering with local firms:** use local knowledge, relationships, and regulatory familiarity.
* **Operational hedging:** diversify suppliers, production locations, and revenue bases.
* **Financial hedging:** use derivatives or currency-matched borrowing where appropriate.
* **Local financing:** borrowing locally can partially align liabilities with local-currency cash flows and reduce parent-level exposure.
* **Contract design and governance safeguards:** use clear control rights, dispute-resolution clauses, and exit mechanisms.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> - Risk management is most effective when built into project design from the beginning. 
> - The best time to manage country risk is before entry mode, financing mix, ownership structure, and operational footprint are fixed.

---


## <span style="color:#4A86C5;"> 9.3 Strategic role of FDI in multinational finance
---
FDI affects capital budgeting, financing needs, exposure to country and currency risk, internal capital allocation, and the future capacity of the firm to raise funds in global markets.

In practice, the strongest MNC decisions link three questions:

1. **Where should the firm invest?**
2. **How should the investment be financed?**
3. **How should the firm manage the resulting risk exposure?**

When these questions are analyzed jointly rather than separately, the firm makes better decisions about project feasibility, valuation, control, and resilience.

**Example:** 
- Suppose a firm is considering a new production facility abroad. 
- It may:
    - choose a joint venture rather than full ownership, 
    - finance part of the project with local-currency debt, and phase the project in stages. 
- This decision simultaneously addresses:
    - operating uncertainty, 
    - currency risk,
    - financing structure, and 
    - country risk.

### <span style="color:#5b9bd5;">Question 4: Country Risk and Project NPV

A project in Country X has:
- Expected USD CFs (parent perspective): $\$10$ m per year for 5 years
- Initial investment: $\$30$ m
- Required return: 10% (without country risk premium)

**Questions:**
1. Compute the project NPV ignoring country risk.
2. Now apply a country-risk premium of 4% (so $k = 14\%$). Recompute NPV.
3. Alternatively, apply a 20% haircut to the cash flows (CF = \$8 m) and use $k = 10\%$. Recompute NPV.
4. Compare the two adjustments and discuss when each is more appropriate.

---

## <span style="color:#4A86C5;">Solutions

---

### <span style="color:#5b9bd5;">Solution 1: Eurobond vs Foreign Bond

**1.**
- **Eurobond:** USD-denominated bond issued **outside** the U.S. (e.g., London), targeting global investors. Lighter regulation; underwritten by international syndicate.
- **Yankee bond:** USD-denominated bond issued **in the U.S.** by the foreign firm. Subject to SEC registration and full U.S. disclosure.

**2.**

| Feature | Eurobond | Yankee |
|---|---|---|
| Regulation | Light, market-driven | Strict (SEC) |
| Investor base | Global | Mostly U.S. |
| Disclosure cost | Lower | Higher |
| Speed of issuance | Faster | Slower |

**3.** If the firm has substantial USD revenues, either provides a natural hedge. The **Eurobond** is typically faster and cheaper to issue, but the Yankee gives broader access to U.S. investors and may achieve a tighter spread for high-quality issuers known to U.S. analysts.

---

### <span style="color:#5b9bd5;">Solution 2: Cross-Listing

**1. Benefits**
- Wider investor base → higher demand and lower cost of equity
- Improved liquidity in shares
- Stronger governance signal (must meet LSE disclosure standards)

**2. Costs**
- Listing and ongoing compliance costs (sometimes USD millions per year)
- Higher disclosure burden (potentially exposing competitive information)

**3. Cross-listing vs new equity issuance abroad**
- **Cross-listing:** lists existing shares on a foreign exchange — no new capital raised.
- **Foreign equity issuance:** sells *new* shares abroad — actually raises new capital (and dilutes existing shareholders).

---

### <span style="color:#5b9bd5;">Solution 3: FDI Mode Selection

**1.**

| Mode | Control | Capital | Speed | Country-risk exposure |
|---|---|---|---|---|
| Export | Low | Low | Fast | Low |
| License | Low | Very low | Fast | Low |
| Greenfield | High | Very high | Slow | High |

**2.** **Greenfield** is best for long-term presence and brand control: the firm builds its own facility with its own quality standards, technology, and culture.

**3.** **Exporting** or **licensing** is best when long-term demand is uncertain — both can be scaled back or terminated quickly with minimal sunk cost.

---

### <span style="color:#5b9bd5;">Solution 4: Country Risk and Project NPV

**1. NPV without country risk** ($k = 10\%$):

$$NPV = -30 + \sum_{t=1}^{5} \dfrac{10}{1.10^{t}} = -30 + 10 \times 3.7908 = -30 + 37.91 = +\$7.91 \text{ m}$$

**2. With higher discount rate** ($k = 14\%$):

$$NPV = -30 + 10 \times 3.4331 = -30 + 34.33 = +\$4.33 \text{ m}$$

**3. With haircut to cash flows** (CF = 8, $k = 10\%$):

$$NPV = -30 + 8 \times 3.7908 = -30 + 30.33 = +\$0.33 \text{ m}$$

**4. Discussion**
- **Discount-rate adjustment** is appropriate when country risk is **systematic and persistent** (chronic political instability, sovereign default risk).
- **Cash-flow adjustment** is appropriate when risks are **specific and modellable** (e.g., probability of partial expropriation in Year 3).
- Doing **both** to the same risk double-counts the penalty.
