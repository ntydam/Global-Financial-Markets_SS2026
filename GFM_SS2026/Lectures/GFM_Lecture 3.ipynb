# 1. Overview of the Foreign Exchange Market

---



## <span style="color:#4A86C5;">1.1. Overview</span>

---
### <span style="color:#5b9bd5;">Definition

The foreign exchange (FX) market is a global marketplace where participants exchange one currency for another.

It is decentralized (OTC):
- There is no single physical trading floor or central exchange.
- Prices are quoted and trades are executed via banks, dealers, and electronic trading platforms.

FX is one of the largest and most liquid financial markets (as shown in Lecture 1).

---

### <span style="color:#5b9bd5;"> Trading activities and liquidity

FX trading is near 24 hours a day during the workweek (markets open sequentially across regions). 
<center>
<img src="../Images/Lecture_3/pic19.png" alt="pic19" width="800">
</center>

It is primarily a wholesale (interbank) market, where:
- Large banks and institutional players trade with each other
- Retail flows typically reach the market through intermediaries (brokers/banks/platforms)

However, activity is concentrated in a few key financial centers: London, New York, Tokyo, and China.

<center>
<img src="../Images/Lecture_3/pic2.png" alt="pic2" width="600">
</center>

---

## <span style="color:#4A86C5;">1.2. Who trades FX and why?</span>

---
- International transactions (firms, investors, tourists)
    - Pay for trade, investment, travel
    - E.g., importer buys USD to pay supplier

- Hedging (firms, investors)
    - Reduce FX risk
    - This may result in a reduction in the expected return
        - E.g: An importer or exporter can use forward FX contracts to lock in future exchange rates, for when the transaction will be settled.

- Speculation (traders, investors)
    - Profit from expected FX moves 
    - E.g., buy currency expected to appreciate

- Arbitrage (banks, traders)
    - Exploit price differences across markets 
    - E.g., buy where cheaper, sell where higher

- Policy-driven (central banks, governments)
    - Manage exchange rate / reserves
    - E.g., intervene to smooth volatility/ test the market to ensure currency is accurately priced

---

## <span style="color:#4A86C5;">1.3. Transaction Types</span>

---

### <span style="color:#5b9bd5;">Spot transactions: 
  - An immediate exchange of currencies with settlement typically two business days later; 
  - E.g., EUR/USD = 1.1000 indicate the price of one euro in U.S. dollars.
  - In recent years, the use of spot trading has declined.

### <span style="color:#5b9bd5;">Forward contracts: 
  - Agreements today to exchange currencies at a specified future date and rate.
  - Forwards shield participants from future spot‑rate movements and quote maturities of one, three or six months.

### <span style="color:#5b9bd5;">Swaps: 
  - FX swaps account for over half of daily turnover (mostly for funding and hedging currency risk). 
  - A swap combines a spot and a forward transaction.
  - Dealers buy a currency spot and sell it forward (or vice versa), allowing them to adjust positions without incurring FX risk. 
  - Forward–forward swaps exchange currencies for two different future dates, with the rate difference reflecting interest differentials.


<center>
<img src="../Images/Lecture_3/pic3.png" alt="pic3" width="850">
</center>

### <span style="color:#5b9bd5;">Non‑deliverable forwards (NDFs): 
  - Cash‑settled forwards, typically quoted against the U.S. dollar, used to hedge currencies subject to capital controls.
  - Settlement is based on the difference between the agreed forward rate and the prevailing spot rate.

---

## <span style="color:#4A86C5;">1.4. Quotation Conventions</span>

### <span style="color:#5b9bd5;">Terminology
#### Common terminology for currency pairs:
$$
\boxed{\frac{\text{Base currency}}{\text{Quote currency}}}
$$

- **Base currency:** is the unit being priced.

- **Domestic currency:** The domestic currency is the one relevant to the analyst or investor. 
  - A direct quote uses the domestic currency as the quote currency
  - An indirect quote uses the domestic currency as the base.


#### Another way to express a currency pair:
$$ 
\boxed{\frac{\text{Commodity currency}}{\text{Terms currency}}}
$$

- E.g., USD/VND means the price of the USD in terms of the VND.
---



### <span style="color:#5b9bd5;">Direct vs. Indirect Quotes</span>

#### Direct quote: 
- Value of one unit of foreign currency (FC) in terms of domestic currency.
* Ask “How many units of my home currency (e.g., VND) do I need to buy one USD?”
      - USD/VND = 24,850			1 USD = 24,850 VND

<center>
<img src="../Images/Lecture_3/pic4.png" alt="pic4" width="500">
</center>

  
#### Indirect quote:
- Is the reciprocal: how many units of FC are needed to purchase one unit of the domestic currency.  
  $$
  \boxed{\text{Indirect quote} = \frac{1}{\text{Direct quote}}}
  $$

- E.g., if USD/VND = 24,850, then VND/USD = 1/24,850 ≈ 0.0000402.
- A decline in an indirect quote signals depreciation of the domestic currency.
  
---


### <span style="color:#5b9bd5;">Bid–Ask Spreads
- When dealers quote exchange rates, they provide:
  - a bid price (rate at which they will buy the base currency) and 
  - an ask price (rate at which they will sell it).

- The difference is the **bid–ask spread**, which compensates dealers for providing liquidity.  

<center>
<img src="../Images/Lecture_3/pic9.png" alt="pic9" width="600">
</center>

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
The bid price is always lower than the ask price.

- A simple measure of the percentage spread is:
$$
\boxed{\text{Percentage spread} = \frac{\text{Ask} - \text{Bid}}{\text{Ask}} \times 100}
$$

---


### <span style="color:#5b9bd5;"> The Two-way Pricing Process:

- Price makers (dealers) quote two-way prices to buyers and sellers.
    - If a buyer accepts the ask price, the dealer sells at that price.
    - If a seller accepts the bid price, the dealer buys at that price.

- Price takers (clients) can either accept the quote or negotiate for a better price. 
    - If they accept, the transaction occurs at the quoted price.
    - If they negotiate, the dealer may adjust the bid and ask prices to reflect the new terms.

<center>
<img src="../Images/Lecture_3/pic5.png" alt="pic5" width="600">
</center>

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
If the quote is accepted, the price maker must go ahead with the transaction.

# 2. Deriving FX Relationships

---



## <span style="color:#4A86C5;">2.1 Cross Rates and Triangular Arbitrage</span>
### <span style="color:#5b9bd5;">Cross Rates (Chain Rule)
- When two currencies lack a direct quote, their **cross rate** can be derived from their prices against a common currency. 
  - If both rates are direct quotes, we have standard cross rates as:
    $$
    \text{AUD/JPY} = \frac{\text{AUD/USD}}{\text{JPY/USD}}
    $$

  - If one rate is quoted in direct form and the other in indirect form, multiply rather than divide.  
    $$
    \text{AUD/JPY} = {\text{AUD/USD}} \times {\text{JPY/USD}}
    $$

- **Example:** Cross Rate with Bid–Ask
  Given:
  - USD/JPY = 100.00/100.50 (yen per dollar)  
  - USD/CAD = 1.2700/1.2720 (Canadian dollars per dollar)

  - To find CAD/JPY:
    - Bid CAD/JPY:
      $$
      \text{Bid (CAD/JPY)} = \frac{100.00}{1.2720} = 78.64\ \text{JPY per CAD}.
      $$
    - Ask CAD/JPY:
      $$
      \text{Ask (CAD/JPY)} = \frac{100.50}{1.2700} = 79.13\ \text{JPY per CAD}.
      $$
      
### <span style="color:#5b9bd5;">Triangular Arbitrage
- Triangular arbitrage arises when quoted cross rates deviate from implied cross rates.
- Traders can profit by buying the undervalued currency and selling the overvalued one through a series of three trades, though such opportunities vanish quickly in a liquid market.

---


## <span style="color:#4A86C5;">2.2 Interest‑Rate Parity and Forward Pricing</span>
### <span style="color:#5b9bd5;">Covered Interest‑Rate Parity (CIP)
The CIP condition states that a forward premium/discount offsets IR differentials across countries.
$$
\boxed{F = S \times \frac{1 + i_d}{1 + i_f}}
$$
where:
- $F$ is the forward rate (domestic currency per unit of FC), 
- $S$ is the current spot rate, and 
- $i_d$ and $i_f$ are domestic and foreign interest rates respectively.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
A currency with a lower interest rate trades at a forward **premium**; one with a higher rate trades at a forward **discount**.

---


### <span style="color:#5b9bd5;">CIP: Determining Arbitrage Opportunities Using CIP
Under CIP the relationship is roughly linear (slope ≈ 1),
  - which is a no-arbitrage condition ($\frac{F−S}{S}$) ≈ ($i_d - i_f$).

<center>
<img src="../Images/Lecture_3/pic7.png" alt="pic7" width="600">
</center>
CIP must hold because:
- The forward contract eliminates exchange rate risk
- Cash flows are known with certainty (It uses actual prices (S and F))
- Large institutions (banks, hedge funds) can execute this at scale
- Any deviation gets arbitraged away quickly.

---


#### Example: Determining Arbitrage Opportunities Using CIP
Suppose:
  - the spot rate is EUR/USD = 1.1000,
  - the one‑year euro interest rate is 3 % and 
  - the one‑year U.S. rate is 1 %. 

The one‑year forward rate is
$$
F = S \times \frac{1 + i_d}{1 + i_f} = 1.1000 \times \frac{1 + 0.01}{1 + 0.03} \approx 1.0796.
$$
Thus:
- The EUR sells at a **forward discount** because $i_{EUR}$ > $i_{USD}$;
- Any deviation from 1.0796 would create profit opportunities.

---




### <span style="color:#5b9bd5;">Uncovered Interest‑Rate Parity (UIP)

UIP replaces the forward rate with the expected future spot rate:
$$
\boxed{E[\Delta s]  \approx i_d - i_f}
$$
UIP predicts:
- Higher interest rate → expected depreciation.

Empirically, we often observe the opposite:
- High interest rate currencies appreciate
- This creates the famous carry trade profit


---


## <span style="color:#4A86C5;">2.3 Purchasing-Power Parity (PPP)</span>

### <span style="color:#5b9bd5;">The Law of One Price (LOP)

PPP is based on the LOP:
Identical goods should sell for the same price across countries once expressed in a common currency.

The **Big Mac Index**, published by *The Economist*, is a practical illustration of the LOP and PPP .

<center>
<img src="../Images/Lecture_3/pic20.png" alt="pic20" width="400">
</center>

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
A Big Mac is a nearly identical product (same size, same ingredients) sold worldwide.  
If LOP holds, it should cost the same everywhere once converted into a common currency.

### <span style="color:#5b9bd5;">Absolute PPP

$$
\text{Implied Spot Rate} = \frac{P_d}{P_f}
$$

Where:
- $\text{Implied Spot Rate}$ = spot exchange rate (domestic price of FC)
- $P_d$ = domestic price level
- $P_f$ = foreign price level

Example: If a Big Mac costs $5 in the U.S. and 500 yen in Japan, PPP implies:
$$
S = \frac{500\ \text{JPY}}{5\ \text{USD}} = 100\ \text{JPY/USD}.
$$ 

---



### <span style="color:#5b9bd5;">Relative PPP
- If the assumptions of absolute PPP are relaxed a bit, we observe what is termed relative PPP. 

- Relative PPP holds that PPP is not helpful in determining what the spot rate is today, but that the relative change in prices (inflation) between two countries over a period of time determines the change in the exchange rate over that period.

- In other words, *"if the spot exchange rate between two countries starts in equilibrium, any change in the differential rate of inflation between them tends to be offset over the long run by an equal but opposite change in the spot exchange rate."*

Therefore, higher domestic inflation can lead to domestic currency depreciates, and vice versa.

$$
\boxed{\frac{\Delta S}{S} \approx \pi_d - \pi_f}
$$

Where:
- $ \pi_d $ = domestic inflation
- $ \pi_f $ = foreign inflation

Example: 
If U.S. inflation = 6% and, Japan inflation = 1%, PPP predicts:
    - USD should depreciate ~5% against JPY.

---


###  <span style="color:#5b9bd5;">PPP Fails in the Short Run and Matters in the Long Run
The Big Mac Index is imperfect because:

- Big Macs include non-tradable inputs (rent, wages)
- Taxes differ across countries
- Transport and regulatory costs exist
- Market structures differ
- Prices are sticky

So PPP works better in the **long run** than in the short run.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
Short-run FX = driven by capital flows and interest rates  
Long-run FX = anchored by inflation differentials

# 3. Determinants of Exchange Rates

---

## <span style="color:#4A86C5;">3.1. Economic Fundamentals

Exchange rates are determined by foreign exchange market equilibrium, where:

$$\boxed{\text{Demand for domestic currency} = \text{Supply of domestic currency}}$$

They are fundamentally driven by:
- Relative inflation, CPI (PPP chananel)
- Relative GDP growth
- Relative interest rates
- Commodity prices (for commodity exporters)

Each factor operates through trade flows, capital flows, or both.

---


### <span style="color:#5b9bd5;">Relative Inflation
If Vietnam has higher inflation than its trading partners, it will experience:
- Less demand for exports, and therefore less demand for VND
- More demand for imports, and therefore more supply of the VND
=> VND depreciates against trading partners' currencies.


<center>
<img src="../Images/Lecture_3/pic12.png" alt="pic12" width="600">
</center>

---


### <span style="color:#5b9bd5;">Relative GDP Growth Rate

Economic growth affects exchange rates through **two opposing channels**:


#### Import Channel (Depreciation Pressure)

Higher income → Higher demand for imports  
→ More domestic currency supplied to FX markets  
→ Depreciation pressure  

#### Capital-Flow Channel (Appreciation Pressure)

Stronger growth → Higher expected returns  
→ More foreign investment (FDI & portfolio inflows)  
→ Increased demand for domestic currency  
→ Appreciation pressure  


> <span style="color:#4A86C5;">**Key intuition:**</span><br>
Net effect depends on which force dominates.

Higher GDP growth may cause

- **Supply curve shifts right** (more imports → more currency supplied)  
- **Demand curve shifts right** (more capital inflows → more currency demanded)


If capital inflows dominate → Appreciation
If import effect dominates → Depreciation

---

### <span style="color:#5b9bd5;">Relative Interest Rates</span>

Relative interest rates are a short-run determinant of exchange rates because they influence international capital flows.

When a country offers higher interest rates relative to others:

1. Domestic assets offer higher returns, it becomes more attractive to global investors 
2. Foreign investors buy domestic bonds/assets  
3. Demand for domestic currency increases  
4. Exchange rate appreciates  

Interest rate changes often move exchange rates immediately — even before real economic variables adjust.

---

### <span style="color:#5b9bd5;">Commodity Prices (1)

For commodity exporters (e.g., Australia, Canada, Norway, Chile), commodity price shocks affect exchange rates via:

#### Channel 1: Terms-of-trade (trade prices)
ToT indicates how many imports a country can buy per unit of exports:
$$
\text{ToT Index} = 100 \times \frac{P_X}{P_M}
$$

If export prices rise faster than import prices:
- the ToT index increases, implying stronger export purchasing power.

<center>
<img src="../Images/Lecture_3/pic13.png" alt="pic13" width="800">
</center>


**Example:**
- Because Australia is a major commodity exporter, commodity price changes materially affect export prices and ToT.
- When commodity export prices rise (e.g., iron ore, coal), ToT improves, export receipts increase, and demand for AUD tends to rise—so the AUD often appreciates.
<center>
<img src="../Images/Lecture_3/pic14.png" alt="pic14" width="800">
</center>

A historical illustration (RBA narrative):
- during the mid-2000s–2013 mining investment boom, commodity prices surged, foreign capital flowed into the resources sector, and the AUD appreciated, reaching ~A$1.10 per US$ in 2011.

---


### <span style="color:#5b9bd5;">Commodity Prices (2)
#### Channel 2: Trade-flow (FX demand/supply)
Higher export receipts increase FX inflows; exporters often convert part into local currency (↑ demand). 
Higher imports do the opposite by increasing local currency supply in FX markets (↓ exchange rate).

**Trade Weighted Index (TWI) / Broad Index** measures a currency’s value against a **basket of trading-partner currencies**, weighted by each partner’s share of trade.

Example: Both charts measure the external value of the Australian dollar:
- AUD/USD → Value of AUD against the US dollar
- TWI → Value of AUD against a basket of trading partner currencies

When both rise together, AUD is strengthening broadly, not just against the USD.

<center>
<img src="../Images/Lecture_3/pic17.png" alt="pic17" width="800">
</center>

---


### <span style="color:#5b9bd5;">Commodity Prices (3)
#### Channel 3: Investment/Capital-flow

Commodity booms often trigger large investment cycles.

When commodity prices remain high:

1. Mining firms expand production capacity.  
2. New projects are approved (mines, LNG plants, infrastructure).  
3. These projects are frequently financed by:
   - FDI
   - Foreign portfolio investment  

Foreign investors must **purchase AUD** to invest in Australian assets.

This leads to:
- Capital inflows ↑  
- Demand for AUD ↑  
- Exchange rate appreciation ↑  

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
Trade flows influence exchange rates gradually through export receipts.  
Capital flows can move exchange rates quickly and often dominate in the short run.

---

## <span style="color:#4A86C5;">3.2. Other Economic Factors

### <span style="color:#5b9bd5;">3.2.2. International Speculation

- Capital flows toward economies that are perceived as strong and away from those seen as weak.
- Large institutions, hedge funds, and banks actively trade currencies based on macroeconomic outlooks.  
- Because global FX markets are highly liquid, speculative flows can cause **large short-term exchange rate volatility**.

---

### <span style="color:#5b9bd5;">3.2.2. Exchange Rate Expectation

Expectations about future exchange rates influence current exchange rates.

If investors expect a currency to appreciate:
- They buy the currency now  
- Demand increases  
- The currency appreciates  

If depreciation is expected:
- Investors sell  
- Supply increases  
- The currency weakens  

This creates a **self-fulfilling mechanism**, especially in highly liquid markets.

Example: 
If markets expect the central bank to raise interest rates next month, the currency may appreciate today — before the rate change actually occurs.

---


### <span style="color:#5b9bd5;">3.2.3. Central Bank Intervention

Central banks can influence exchange rates directly by buying or selling currencies.

**Floating Exchange Rate – “Dirty Float”**

- The exchange rate is market-determined.
- The central bank may intervene occasionally to:
  - Smooth excessive volatility  
  - Counter disorderly market conditions  

This is sometimes called a **“dirty float”** because it is not a perfectly free float.

Intervention is typically temporary and limited in scale.

**Managed / Fixed Exchange Rate**

- The central bank targets a specific exchange rate level or band.
- To maintain the target:
  - It buys domestic currency to prevent depreciation  
  - It sells domestic currency to prevent appreciation  

This may require **large-scale and sustained intervention**, affecting foreign reserves.

## <span style="color:#4A86C5;">3.3. Market Psychology and Speculation</span>

- **Safe‑haven flows**: In times of global uncertainty, investors flock to “safe‑haven” currencies like USD and CHF.

- **Herd behaviour and technical trading**: Traders often follow price trends and react to popular economic indicators, reinforcing momentum.

- **Unexpected events and news**: Natural disasters, geopolitical tensions or surprise policy announcements can cause abrupt shifts in exchange rates.

# Case Study </span>

---
### <span style="color:#4A86C5;">Case 1: The Famous Yen Carry Trade
- Borrow JPY at a low interest rate  $i_{JPY}$
- Convert JPY → USD at the spot rate  $S_t \text{(USD/JPY)}$
- Invest in higher-yield USD assets at  $i_{USD}$
- Convert back to JPY at the future spot rate  $S_{t+1}$

A simple approximation of the carry trade return:
$$
R \approx (i_{USD} - i_{JPY}) + \Delta s
$$

where:
- $i_{USD} - i_{JPY}$ = **interest differential (carry)**
- $\Delta s \approx \frac{S_{t+1} - S_t}{S_t}$ = **exchange rate change (USD/JPY)**

However, risk and shocks exist in practice. [Koijen et al. (2018)](https://www.sciencedirect.com/science/article/abs/pii/S0304405X17302908) show that actual returns include unexpected shocks:

$$
R \approx (i_{USD} - i_{JPY}) + \Delta s + \varepsilon
$$

where:

- $\varepsilon$ = unexpected shocks (risk-off events, funding stress, sudden JPY demand)

When global risk rises:
- Investors unwind carry trades  
- They buy JPY to repay borrowing  
- JPY appreciates sharply  
- Losses can accelerate quickly  

Carry trades often generate steady gains but are exposed to **crash risk**.

> <span style="color:#4A86C5;">**Key intuition:**</span><br>
> - UIP = Carry trade = **Uncovered position (takes FX risk)**
> - CIP = **Covered position (no FX risk, no arbitrage profit)**
