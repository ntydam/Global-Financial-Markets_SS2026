> ** Quote-direction discipline**
> Always read the quote first. "25,000 VND/USD" means 25,000 VND per 1 USD — VND is the HC, USD is the FC. Most common exam errors trace back to a quote mistake.

# 1. Purchasing Power Parity & Exchange Rates

- Section 1 of the formula sheet links domestic and foreign price levels to exchange rates.

- Three ideas matter: 
    - prices should equalise after currency conversion (LOP and PPP);
    - the forward rate is pinned down by interest-rate differentials so arbitrage cannot occur (CIP); 
    - and deviations from PPP, captured by the real exchange rate, drive long-run currency adjustment.

---


## Absolute PPP

$$
\boxed{S = P_\text{domestic basket} / P_\text{foreign basket}}
$$

**Meaning:** 
- The exchange rate equals the ratio of the two countries' price levels, so a unit of HC buys the same basket of goods at home as abroad after conversion.
- If absolute PPP holds, the real exchange rate q equals 1.

**Use when:**
- Long-run benchmarks (e.g., Big Mac index, IMF GDP comparisons). 
- Almost never holds in practice because of non-tradables and price stickiness.

---


## Relative PPP

$$      
\boxed{S₁ = S_0 × (1 + π_\text{domestic}) / (1 + π_\text{foreign})}
$$

**Meaning:** 
- Forecasts how much the spot rate should change over a period given each country's inflation.
- If domestic inflation is higher, the home currency depreciates — S rises (more HC per FC).

**Use when:**
- Forecasting the FUTURE spot rate from CPI data. The version most often tested on the exam.

**Why it fails in the short run:** 
- Prices are sticky (wages, contracts, menus adjust slowly), capital flows and risk sentiment dominate over goods-market arbitrage on horizons of months to a few years, and structural factors like the Balassa–Samuelson effect distort the inflation differential.

**Why it still matters:** empirically, real exchange rates revert to PPP with a half-life of 3–5 years (Rogoff 1996), so PPP is a long-run anchor, not a short-run predictor.

---


## Law of One Price

$$\boxed{S_\text{LOP} = P_\text{domestic} / P_\text{foreign}}$$
<center><i>(single tradable good, in HC and FC)</i></center>


**Meaning:** 
- Same good, same price after conversion — otherwise arbitrageurs would buy where it is cheap and sell where it is dear. 
- Aggregating LOP across all goods gives Absolute PPP.

**Use when:** 
- Single-good comparisons (MacBook Pro, Big Mac).
- The LOP-implied rate vs. the actual spot tells you where the arbitrage flows would go.

**Why arbitrage is incomplete in practice:** 
- Tariffs and quotas, transport costs, taxes, non-tradable services (haircuts, rent), grey-market controls, and brand pricing strategies (Apple deliberately prices regionally). 
- The exam expects you to NAME at least one of these frictions when explaining persistent LOP gaps.

---



## Real exchange rate

$$\boxed{q = S \times \frac{P_\text{foreign}}{P_\text{domestic}}}$$
<center><i>[q = 1 ⇔ absolute PPP holds]</i></center>


**Meaning:** q tells you the relative purchasing power of HC abroad after currency conversion.
- $q < 1$ → HC buys MORE goods abroad → home currency is **overvalued** in real terms.
- $q > 1$ → HC buys fewer goods abroad → home currency is **undervalued**.
- Mean-reverts to 1 in the long run.

**Use when:** Diagnosing currency misalignment. "Is the VND overvalued?" → compute q and compare to 1.

---


## CIP / Forward Rate

$$\boxed{F = S_0 \times \frac{1 + r_{\text{domestic}} \times T}{1 + r_{\text{foreign}} \times T}}$$


**Meaning:** The no-arbitrage forward rate. If you can borrow domestic, convert at S₀, invest abroad at r_FC, and lock in the conversion back at F, the riskless return must equal the domestic rate. Any other forward rate creates a riskless arbitrage in deep markets.

**Use when:** Pricing forwards, sanity-checking quoted forwards, and any hedging question that compares forward vs. money-market hedge.

**Direction rule:** when r_domestic > r_foreign, the forward F is ABOVE the spot — the FC is at a forward **discount** (buys fewer HC forward). When r_domestic < r_foreign, the FC is at a forward **premium**. The MMH and forward must produce identical hedged outcomes if CIP holds exactly.

## Forward premium / discount (annualised)


$$\boxed{\text{Premium/discount} = \frac {(F − S₀)}{S₀} \times (360 / \text{days})}$$

**Meaning:** Converts the percentage forward gap into an annualised rate so it is directly comparable to interest-rate differentials. Positive = forward premium on FC; negative = forward discount.

**Use when:** Reporting hedging cost in p.a. terms; comparing forward signals across maturities.

## Frameworks not on the formula sheet

### Exchange-rate regimes

| Regime | What it means | Example |
| --- | --- | --- |
| Pure float | Rate set entirely by market; no central-bank intervention. | USD |
| Managed (dirty) float | Market-determined but central bank intervenes at extremes. | JPY (BOJ FX interventions in 2022) |
| Crawling peg | Rate adjusted by a pre-announced rule (often inflation-driven). | CNY against USD basket |
| Fixed peg | Central bank commits to a band; uses reserves to defend. | HKD/USD |
| Currency board / dollarisation | Full or near-full surrender of monetary autonomy. | Ecuador, Hong Kong |

---


### Balance of Payments (BOP) intuition

$$\boxed{\text{BOP} = \text{Current Account} + \text{Capital Account} + \text{Financial Account} = 0 \text{(net)}}$$

- A country running a current-account deficit must finance it with:
    - a capital-and-financial account surplus — or 
    - its currency depreciates.
- Twin deficits (current AND financial outflows) put the strongest downward pressure on a currency, as Japan illustrated in 2022–2024.

---


### The Impossible Trinity

A country can choose any TWO of:
1. independent monetary policy
2. free capital mobility
3. a fixed exchange rate

— never all three. Japan picks (1)+(2), so it sacrifices (3) and lets the yen float. China picks (1)+(3) and uses capital controls. The Eurozone members pick (2)+(3) and surrender independent monetary policy to the ECB.

---


### Uncovered Interest Rate Parity (UIP) and the carry trade

$$\boxed{E[\Delta_S] = i_\text{domestic} − i_\text{foreign}}$$
<center><i>(UIP, in HC per FC quote)</i></center>

- UIP says the high-rate currency must, on average, depreciate enough to wipe out the interest gain.
- Empirically this fails — high-rate currencies tend to **appreciate**, generating profits for the carry trade (borrow JPY at 0%, invest in MXN at 10%).
- The reward is compensation for crash risk: the carry trade pays steady gains punctuated by violent unwinds, as seen on 5 August 2024 when the BOJ rate hike triggered a ~$500 bn yen-carry unwind. CIP holds tightly (it is enforced by arbitrage); UIP holds only on average, and even then with large persistent deviations.

# 2. Foreign Direct Investment

FDI is when a firm acquires lasting control over a productive asset abroad (≥10% equity, OECD definition). The exam tests two skills: (1) explaining WHY a firm chooses FDI rather than exporting or licensing, using the OLI framework; and (2) valuing the project with NPV, including host-country tax and currency conversion.

## OLI eclectic paradigm (Dunning)

| Pillar | Question it answers | Examples |
| --- | --- | --- |
| **Ownership (O)** | What firm-specific advantage can be deployed abroad? | Patents, brand, scale, proprietary tech, management know-how |
| **Location (L)** | Why this country, not another, and not exports? | Skilled labour, supplier network, market size, tariffs avoided, subsidies |
| **Internalisation (I)** | Why FDI rather than licensing? | Protect IP from leakage, capture full margin, control quality, avoid hold-up |

All three pillars must be present for FDI to dominate. Strong O without L → license abroad. Strong O+L without I → license rather than build. The exam answer always pairs each pillar with ONE concrete advantage from the case scenario.

---


## Net Present Value (general)

$$\boxed{NPV = Σ [CF_t / (1 + r)^t] − \text{Initial Outlay}}$$

**Meaning:** Discounts each future cash flow back to today and nets out the upfront investment. Accept the project if NPV > 0; reject if NPV < 0.

**Use when:** Any project with non-constant cash flows.

---


## Annuity factor

$$
\boxed{AF(r, n) = \frac{1 - (1 + r)^{-n}}{r}}
$$

**Meaning:** Present value of receiving 1 unit per year for n years, discounted at r. Multiplying by the constant cash flow gives the PV of the whole stream in one step.

**Use when:** Cash flows that are constant each period (the assumption made in the exam's FDI question).

> Round AF to **4 decimal places** — this is what the exam rubric specifies. The formula on the sheet may render the exponent ambiguously; the correct exponent is **NEGATIVE n** (so that AF is positive for r > 0).

---


## NPV using the annuity factor

$$\boxed{NPV = Annual CF × AF(r, n) − Initial Outlay}$$

**Meaning:** Compact form when cash flows are constant. The same accept/reject rule applies.

**Use when:** FDI projects with steady annual cash flows; exam "should the firm proceed" questions.

---


## Tax adjustments and repatriation

$$
\boxed{CF_{\text{after-tax}} = CF_\text{pre-tax} \times (1 - \text{tax rate})}
$$

Apply the host-country corporate tax to the operating profit BEFORE you discount.

$$      
\boxed{CF_{\text{repatriated}} = CF_{\text{after-tax}} \times (1 − \text{withholding rate})}
$$

Most host countries impose a withholding tax when dividends are paid to a foreign parent. Apply this on top of the corporate tax.

$$\boxed{HC_received= CF_{\text{repatriated}} (\text{in FC}) \times \text{Forward rate}}$$

Convert the repatriated FC dividend back to home currency using the forward rate to remove FX risk on the conversion. Always state which forward rate you used (e.g., the 1-year forward).

---


## Country risk

| Type | What it covers | Concrete mitigation |
| --- | --- | --- |
| **Political** | Expropriation, sanctions, regime change, contract repudiation | Singapore-SPV holding company, MIGA political-risk insurance, host-government relations |
| **Regulatory** | New tax rules, content requirements, FX controls, ESG/battery rules | Lock in tax stability clause, build local content early, monitor via local advisor |
| **Economic** | Recession, inflation spike, sovereign default | Scenario analysis in NPV, FX hedge on dividend stream, diversify across markets |
| **Operational** | Supply-chain disruption, infrastructure, labour disputes | Dual sourcing, local partner, business-continuity planning |

> **Exam discipline:** every risk you name MUST be paired with a concrete mitigation. Generic risks like "political instability" without a specific mitigation lose marks.

# 3. Transaction Exposure & Hedging
---


## Types of FX exposure — not on the formula sheet

| Exposure | Definition | Time horizon | Standard hedges |
| --- | --- | --- | --- |
| **Transaction** | Known FC cash flow already contracted: receivables, payables, FC debt service. | Days to months | Forward, money-market hedge, option, currency swap |
| **Translation** | Accounting consolidation of an FC subsidiary's balance sheet into parent currency. Non-cash; lives in OCI as a cumulative translation adjustment. | Reporting period | FC-denominated debt as a natural hedge; many MNCs leave it unhedged |
| **Economic (operating)** | PV of future operating cash flows when the firm's competitive position shifts because of a currency move. | Multi-year | Operational levers (sourcing, pricing, factory location); long-dated options |

> **Diagnostic flow**
>
> - Is it a contracted FC cash flow with a known date? → **TRANSACTION**.
> - Is it on the balance sheet of an FC subsidiary being consolidated? → **TRANSLATION**.
> - Is it about future revenues, costs, or margins shifting because of competitive dynamics? → **ECONOMIC**.

---


## Net exposure (per currency)
$$\boxed{\text{Net Exposure} = \sum_{i} \text{Receivables}_i − \sum_{j} \text{Payables}_j}$$
<center><i>(per currency)</center></i>

**Meaning:** Same-currency receivables and payables offset each other. Hedge only the NET position. If receivables = payables, you have a natural hedge and zero exposure (e.g., the CHF position in the MediCorp question).

**Use when:** Building the exposure table — list each currency, sum its receivables and payables, take the difference, then convert to USD at spot to find the largest absolute exposure.

$$\boxed{\text{HC Exposure} = \text{Net FC position} × \text{Expected spot rate}}$$

Restates the FC exposure in HC terms so you can rank currencies by the size of risk. Use the current spot rate unless the question specifies otherwise.

---


## Forward hedge

$$\boxed{\text{Locked-in proceeds} = \text{Notional} × \text{Forward rate}}$$

**Meaning:** Buy or sell the FC forward at F to lock in the HC equivalent today. Zero upfront cost, no basis risk if amount and date are known. Forfeits any favourable spot move at maturity.

**Use when:** Known FC cash flow, known date — the cleanest hedge available.

---


## Money market hedge — receivable (exporter)

Four mechanical steps that synthesise a forward using the spot and money markets:

- Step 1: Borrow PV in FC:                    $PV = \frac{\text{Notional}}{1 + r_{FC} \times T}$
- Step 2: Convert FC borrowing to HC at spot: $HC = PV \times S_0$
- Step 3: Invest the HC for period T:         $HC \times (1 + r_{HC} \times T)$
- Step 4: At maturity, receive the FC inflow and use it to repay the FC loan.

**Concept:** the firm replicates a forward sale of the FC by borrowing in FC against the receivable, swapping the borrowed FC into HC today, and investing the HC. Under exact CIP, the HC outcome matches the forward hedge to the cent. Differences reveal the cross-currency basis or transaction costs.

---


## Money market hedge — payable (importer)

- Step 1: Borrow HC, convert to FC at spot today.
- Step 2: Invest the FC for period T:  $PV = Notional / (1 + r_FC × T)$
- Step 3: At maturity, the FC investment matures and pays the supplier.
- Step 4: Repay the HC loan.

Mirror image of the exporter case. The firm replicates a forward purchase of the FC by pre-funding the FC payable today. Net cost in HC is the original HC borrowing plus interest.

---


## Futures profit and loss

$$\boxed{\text{Long futures P/L} = \text{Notional}\times(S_T − F_0)} $$

<center><i>(long pays F_0, receives S_T)</i></center>

$$\boxed{\text{Short futures P/L} = \text{Notional}\times (F_0 − S_T)}$$
<center><i>(short pays F_0, receives S_T)</i></center>

- **Long futures** profits when the spot at maturity exceeds the agreed futures price. Used to hedge an FC payable or speculate on FC appreciation.
- **Short futures** profits when spot ends below the futures price. Used to hedge an FC receivable or speculate on FC depreciation.

**Forwards vs. futures:** same payoff structure, but futures are exchange-traded, standardised, and marked to market daily (margin calls); forwards are OTC, customisable, and have counterparty credit risk. The exam often asks "which is preferable here?" — the answer hinges on whether the firm needs the exact amount and date (forward) or values liquidity and standardisation (futures).

# 4. Currency Options

- An option gives the **right but not the obligation** to exchange currency at the strike K, in exchange for an upfront premium. 

- Calls give the right to BUY the underlying currency; puts give the right to SELL. 

- The premium is non-recoverable, so options are more expensive than forwards but preserve favourable spot moves.

---


## Payoff and profit formulas

$$\boxed{\text{Long call profit}  = max(S_T − K, 0) − \text{premium}}$$

Buyer of a call exercises if S_T > K, pocketing the difference; loses only the premium otherwise. Convex payoff (limited downside, unlimited upside). Use to **hedge an FC payable** (cap on HC cost) or speculate on FC appreciation.

$$\boxed{\text{Long put profit}   = max(K − S_T, 0) − \text{premium}}$$

Buyer of a put exercises if S_T < K, locking in a sale price K. Convex payoff: limited downside, upside capped at K · notional. Use to **hedge an FC receivable** (floor on HC inflow) or speculate on FC depreciation.

$$\boxed{\text{Short call profit} = \text{premium} − \max(S_T − K, 0)}$$

Seller (writer) collects the premium upfront, accepts unlimited downside if the spot rises above K. Concave payoff. Income strategy when you believe S_T will stay below K; covered-call hedging.

$$\boxed{\text{Short put profit}  = \text{premium} − \max(K − S_T, 0)}$$

Seller of a put collects the premium, accepts loss if S_T < K (must buy at K when market is lower). Income strategy when you believe S_T will stay above K; willingness to acquire at K.

---


## Break-even rates

* Long call break-even:   $S_T* = K + \text{premium}$
* Long put break-even:    $S_T* = K − \text{premium}$

The call buyer recoups the premium only if the spot exceeds strike + premium. 

The put buyer profits only if S_T falls below K − premium. 

Use these to compare options vs. forwards on cost.

---


## Effective ceiling and floor (exam shorthand)

Effective ceiling (long call) = K + premium     (per unit FC)
Effective floor   (long put)  = K − premium     (per unit FC)

- **Ceiling** = importer's worst-case HC cost per FC when hedging a payable with a call.
- **Floor** = exporter's worst-case HC inflow per FC when hedging a receivable with a put.

**Indifference vs. forward** (mock-exam Q4): the option matches the forward when the spot at maturity differs from the forward rate by exactly the option premium. For an exporter put, indifference at $S_T = F + premium$. Above this, the option captures upside. Below this, the forward is cheaper.

## Premium drivers (sensitivity)

| Driver | Call premium | Put premium |
| --- | :---: | :---: |
| Spot S₀ rises | ↑ | ↓ |
| Strike K rises | ↓ | ↑ |
| Volatility σ rises | ↑ | ↑ |
| Time T rises | ↑ | ↑ |
| Domestic rate r_HC rises | ↑ | ↓ |
| Foreign rate r_FC rises | ↓ | ↑ |

**Volatility drives optionality:** more uncertainty about S_T raises the value of the right to choose whether to exercise. Time has the same effect (more chances for the option to end in-the-money). These drivers are routinely asked as conceptual sub-parts — memorise the direction of each.

# 5. Interest Rate Instruments
---


Interest-rate instruments hedge the cost of debt rather than currency. Section 5 of the formula sheet covers the mechanics of debt pricing, futures, FRAs, swaps, and rate caps/floors, plus day-count conventions. The exam's hedging questions sometimes spill into this section if the scenario involves variable-rate borrowing.

## Cost of debt

$$\boxed{\text{Cost of Debt} = \text{Risk-Free Rate} + \text{Credit Risk Premium}}$$

**Meaning:** The required yield on a debt instrument breaks into two parts: the time value of money (risk-free) and compensation for default risk (credit spread). For an MNC, an additional country risk premium may apply when borrowing abroad.

**Use when:** Estimating WACC components, comparing financing options, building EFR-style calculations.

---


## Interest-rate futures price
$$\boxed{Price = 100 − Implied interest rate}$$
<center><i>(e.g., 96.50 implies 3.50% rate)</i></center>

**Meaning:** Convention used by Eurodollar / SOFR futures. Price moves inversely to the implied rate: a 1 bp drop in rates raises the futures price by 1 bp.

**Use when:** Reading a futures quote, computing P&L from a rate move.

---


## Forward Rate Agreement (FRA) — long payoff

$$\boxed{\text{FRA payoff} = \frac{\text{Notional} \times (\text{Market rate} − \text{FRA rate}) \times (\text{Days}/360)}{1 + \text{Market rate} \times (\text{Days}/360)}}$$


**Meaning:** The long FRA receives the difference between the market reference rate and the agreed FRA rate, discounted because settlement is at the START of the underlying period (whereas the rate applies OVER the period). Long FRA gains when rates **rise**.

**Use when:** Hedging a future borrowing against rate increases; short hedges expected falls.

---


## Interest-Rate Swap — net payment from the fixed-payer


$$\boxed{Net = (Fixed rate − Floating rate) \times Notional \times T}$$
<center><i>[payer of fixed]</i></center>


The fixed-rate payer pays fixed and receives floating. Net positive when fixed > floating (the payer is losing); negative when floating > fixed (payer gaining).

Effective rate after swap = Loan spread + Fixed rate on swap        (floating cancels)

If a firm has a floating-rate loan at LIBOR + 1.5% and enters a swap to pay 4% fixed and receive LIBOR, the floating cash flows cancel and the firm's effective cost is $4% + 1.5% = 5.5%$ fixed.

---
## Interest-rate caps and floors

$$\boxed{Cap payoff   = max(Market rate − Cap rate,   0) \times Notional \times T}$$
$$\boxed{Floor payoff = max(Floor rate − Market rate, 0) \times Notional \times T}$$

- **Cap** = series of call options on the reference rate. Pays out only when rates exceed the cap rate, putting a maximum on the borrower's effective cost. Useful for variable-rate borrowers hedging against rate spikes while preserving the benefit of low rates.
- **Floor** = series of put options on the reference rate. Pays out only when rates fall below the floor rate, putting a minimum on the lender's effective income.

Combining a cap (long) and a floor (short) at different strikes creates a **collar** — the premium received from selling the floor offsets some or all of the cost of buying the cap, fixing the effective rate within a band.

---


## Day-count conventions

* Act/360:  $\boxed{Interest = Principal \times Rate \times (Actual days / 360)}$
* Act/365:  $\boxed{Interest = Principal \times Rate \times (Actual days / 365)}$

- **Act/360** — used for USD money-market instruments, EUR money-market, most LIBOR-style rates. Gives slightly more interest per day than Act/365.
- **Act/365** — used for GBP, AUD, JPY money markets, most government bond markets.

> Always check the convention before computing. Mixing Act/360 and Act/365 is a classic exam mistake worth several marks. The mock exam used Act/360 throughout (T = days/360).
